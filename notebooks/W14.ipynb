### 발표 시 반드시 지킬 3가지

1. **인과 표현 금지**: "A가 B를 일으킨다" (X) → "A와 B 사이에 통계적으로 유의한 관련이 있다" (O)

2. **한계를 솔직히**: 데이터의 제약, 표본의 한계, 분석 기법의 제한을 숨기지 않는다. 한계를 인정하는 것은 약점이 아니라 학술적 성실성이다.

3. **"모르겠다"의 용기**: Q&A에서 모르는 질문에 "잘 모르겠지만, ___를 추가로 분석하면 답할 수 있을 것이다"로 대응한다. 근거 없는 추측은 하지 않는다.

### Q&A 예상 질문 대비

교수와 동료가 흔히 하는 질문 유형:

| 질문 | 준비할 답변 방향 |
|---|---|
| 왜 이 분석 기법을 선택했는가? | 변수 척도 + 연구 질문의 유형으로 설명 |
| 표본 크기가 충분한가? | N과 검정력 고려, 일반화 가능 범위 |
| 이 결과를 일반화할 수 있는가? | 표본 특성 + 조사 시점 + 지역 한계 명시 |
| 다른 변수를 통제하면 달라질 수 있지 않은가? | 통제한 변수 + 누락된 변수 솔직 인정 |
| 이분형인데 왜 OLS인가? (또는 역) | 종속변수 분포 근거로 설명 |
