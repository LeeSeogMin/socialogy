### 발표 시 반드시 지킬 3가지

1. **인과 표현 금지**: "A가 B를 일으킨다", "A 때문에 B가 증가한다"는 사용하지 않는다. 대신 "A와 B 사이에 통계적으로 유의한 관련이 있다", "A 집단과 B 집단의 평균 차이가 유의하다"로 표현한다.

2. **한계를 솔직히**: 데이터의 제약, 표본의 한계, 분석 기법의 제한 등을 숨기지 않는다. 한계를 인정하는 것은 약점이 아니라 학술적 성실성의 표현이다.

3. **"모르겠다"의 용기**: 질의응답에서 모르는 질문을 받았을 때 "잘 모르겠지만, ___를 추가로 분석하면 답할 수 있을 것이다"로 대응한다. 근거 없는 추측은 하지 않는다.
