# W07A. 한 변수가 다른 변수를 얼마나 예측하는가? — 단순회귀

> 한 변수가 다른 변수를 얼마나 '예측'하는가? (주의: 예측 ≠ 원인)

**학습 목표**
1. **단순회귀**(simple linear regression)의 모형 구조(Y = β₀ + β₁X + ε)를 이해하고, 계수(β₀, β₁)를 구체적인 문장으로 해석할 수 있다.
2. **적합도(R²)**의 의미를 설명하고, R²가 높아도 인과가 아니며 R²가 낮아도 관계가 무의미하지 않다는 점을 구별할 수 있다.
3. 회귀 결과를 **인과가 아닌 예측/연관**으로 해석하되, 역인과·교란변수·허위상관의 가능성을 명시하는 해석문을 작성할 수 있다.

## 오늘의 질문/목표/진단

### 오늘의 질문

지난 6주차에서는 집단 간 평균 차이를 t-검정과 ANOVA로 검정했다. 오늘은 한 걸음 더 나아가, **한 연속형 변수가 다른 연속형 변수를 얼마나 예측하는지**를 묻는다. 예를 들어, "교육 연수가 길수록 소득이 높은가?"와 같은 질문이다. 이 질문에 답하기 위해서는 두 변수의 관계를 **직선으로 요약**하고, 그 직선이 데이터를 얼마나 잘 설명하는지를 평가하는 도구가 필요하다.

### 빠른 진단 (1문항)

> **Q.** "교육 연수가 1년 증가하면 소득이 50만 원 증가한다"는 회귀 분석 결과이다. 이 문장이 의미하는 것으로 올바른 것은?
>
> ① 교육이 소득의 원인이다
> ② 교육 연수가 1년 증가할 때 소득이 평균적으로 50만 원 증가하는 경향이 관찰된다
> ③ 모든 사람이 교육 1년당 정확히 50만 원을 더 번다

## 핵심 개념 학습

### 왜 필요한가

6주차까지 우리는 두 변수 사이의 관계를 "차이가 있는가?"(t-검정, ANOVA) 또는 "연관이 있는가?"(카이자승)로 물었다. 그러나 사회과학에서 자주 등장하는 질문은 **"X가 변할 때 Y가 얼마나 변하는가?"**이다. 이 질문에 답하기 위한 도구가 **단순회귀(simple linear regression)**이다. 산점도에 가장 잘 맞는 직선을 긋고, 그 직선의 기울기(β₁)와 적합도(R²)를 해석하며, "이 결과가 인과를 의미하지 않는다"는 점을 명확히 하는 흐름을 배운다.

### 핵심 개념 1: 단순회귀(simple linear regression)

**단순회귀(simple linear regression)**란 하나의 독립변수(X)로 하나의 종속변수(Y)를 예측하는 선형 모형이다. 모형의 구조는 다음과 같다.

> Y = β₀ + β₁X + ε

각 요소의 의미는 다음과 같다.

- **β₀(절편, intercept)**: X = 0일 때 Y의 기대값이다.
- **β₁(기울기, slope)**: X가 1단위 증가할 때 Y의 **평균 변화량**이다. 회귀 분석에서 가장 핵심적인 숫자이다.
- **ε(오차항, error)**: 모형이 설명하지 못하는 개인별 차이이다.

**계수 해석 연습**: β₁의 해석은 반드시 구체적인 문장으로 작성한다. 단순히 "β₁ = 45.8"이라고 쓰는 것이 아니라, 다음과 같이 서술한다.

> "교육 연수가 1년 증가하면, 월 소득은 평균 약 45.8만 원 증가하는 것으로 추정된다."

#### 산점도 위의 직선 — 예측 투표

> **투표 질문**: "교육 연수(X)와 소득(Y)의 산점도에 직선을 하나 그었을 때, 교육 연수가 1년 증가하면 소득은 대략 얼마 증가할까?"
>
> - ① 약 10만 원  ② 약 50만 원  ③ 약 100만 원
>
> 투표하라. (10초)

실제로 회귀 분석을 실행하면 β₁ ≈ 45.8이다. 이 직선은 **최소자승법(OLS, Ordinary Least Squares)**으로 구한 것으로, "모든 점에서 직선까지의 수직 거리(잔차)의 제곱 합이 최소가 되는 직선"이다.

### 핵심 개념 2: 적합도(R²)

**적합도(R², coefficient of determination)**란 "회귀 모형이 Y의 전체 변동 중 몇 퍼센트를 설명하는가?"를 나타내는 지표이다. R² = 0.285이면, "이 모형이 소득 변동의 28.5%를 교육 연수로 설명한다"는 뜻이다. R²는 0(전혀 설명 못 함)에서 1(완벽히 설명) 사이의 값을 가진다. 반드시 주의할 점이 두 가지 있다.

1. **R²가 높아도 인과가 아니다**: R² = 0.90이라 해도 X가 Y를 "일으킨다"는 뜻이 아니다.
2. **R²가 낮아도 관계가 무의미하지 않다**: 사회과학에서 R² = 0.10~0.30은 흔한 값이다. 인간 행동은 다양한 요인에 의해 결정되므로, 단일 변수의 설명력이 낮은 것은 자연스러운 일이다.

| R² 범위 | 사회과학에서의 일반적 해석 |
|---:|---|
| < 0.10 | 낮은 설명력 (그러나 이론적으로 의미 있을 수 있음) |
| 0.10 – 0.30 | 보통 수준의 설명력 |
| > 0.30 | 높은 설명력 (단일 변수 기준으로는 상당히 높음) |

### 핵심 개념 3: 인과 오해 방지

이 과목 전체에서 **가장 중요한 메시지**는 다음이다.

> **회귀 분석 = 연관/예측이지, 인과가 아니다.**

인과라고 말하려면 다음 세 가지 조건이 필요하다.

**"인과라고 말할 수 있는 조건" 체크리스트:**
1. **시간적 선후관계**: 원인(X)이 결과(Y)보다 시간적으로 먼저 발생해야 한다.
2. **교란변수 통제**: X와 Y 모두에 관련된 제3의 변수를 통제해야 한다.
3. **실험 설계(무작위 배정)**: X를 무작위로 배정하여 다른 조건을 동일하게 만들어야 한다.

관찰 데이터(설문 등)에서는 이 조건을 충족하기 매우 어렵다. 따라서 회귀 결과를 서술할 때는 "연관", "예측", "추정" 등의 표현을 사용하고, "원인", "영향", "결정" 등의 인과 표현을 피해야 한다.

#### 인과 오해 사례 3가지

| 사례 | 관찰된 상관 | 실제 구조 |
|---|---|---|
| 아이스크림 판매량과 익사 사고 수 | 양의 상관 | **교란변수**(기온)가 둘 다에 관련 |
| 병원에 자주 가는 사람이 더 아프다 | 양의 상관 | **역인과**: 아프기 때문에 병원에 가는 것 |
| 국가별 초콜릿 소비와 노벨상 수 | 양의 상관 | **허위상관**: 국가의 경제 수준이 교란변수 |

#### 인과 표현 찾기 — 30초 짝 토론

> **짝 토론**: 아래 문장에서 인과 표현을 찾아 밑줄을 긋고, 올바른 표현으로 고쳐 보자. (30초)
>
> "회귀 분석 결과, 교육 연수가 소득에 **영향을 미치는 것**으로 나타났다. 교육을 오래 받으면 **소득이 높아진다**."
>
> *(수정 예시: "회귀 분석 결과, 교육 연수와 소득 사이에 통계적으로 유의한 양의 연관이 관찰되었다. 교육 연수가 1년 증가할 때 소득은 평균적으로 약 46만 원 높은 것으로 추정된다.")*

### 예시: 교육 연수 → 소득 단순회귀

아래는 social_survey_w07.csv(200명)를 이용하여 교육 연수로 소득을 예측하는 단순회귀 분석의 핵심 코드이다.

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("social_survey_w07.csv")
X = sm.add_constant(df["교육연수"])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
```

_전체 코드는 practice/W07/code/W07-1-simple-regression.py 참고_

**검정 결과:**

- **β₀(절편) = 85.2**: 교육 연수가 0년일 때 소득의 기대값은 85.2만 원이다.
- **β₁(기울기) = 45.8, p < 0.001**: 교육 연수가 1년 증가할 때, 소득은 평균 약 45.8만 원 증가하는 것으로 추정된다.
- **R² = 0.285**: 이 모형은 소득 변동의 28.5%를 교육 연수로 설명한다.

**잔차 확인**: 잔차(실제값 - 예측값)가 특정 패턴 없이 무작위로 흩어져 있으면, 선형 모형의 가정이 대체로 충족된 것이다. 잔차가 깔때기 모양이면 **이분산(heteroscedasticity)** 문제가 있을 수 있다.

**결과 해석:**

**(분석목적)** 교육 연수가 소득을 얼마나 예측하는지 확인하기 위해 단순회귀 분석을 실행했다. **(핵심근거)** β₁ = 45.8(p < 0.001)이며, R² = 0.285이다. 교육 연수가 1년 증가할 때 소득은 평균 약 45.8만 원 높은 것으로 추정된다. **(결론)** 유의수준 0.05에서 교육 연수는 소득의 통계적으로 유의한 예측 변수이며, 소득 변동의 약 28.5%를 설명한다. **(주의)** 이 결과는 교육이 소득을 "결정"한다는 의미가 아니다. 부모의 경제적 배경, 거주 지역, 직종 등 교란변수를 통제하지 않았으며, 역인과의 가능성도 배제할 수 없다.

## 실습 과제 (수업 내 완결형)

### 과제 개요

오늘 실습에서는 교육 연수 → 소득 단순회귀를 직접 실행하고, 해석문 구조(분석목적 → 핵심근거 → 결론 → 주의)를 회귀 분석에 적용한다.

### 기본(Basic) 트랙

#### ① 실행

교육 연수 → 소득 단순회귀를 실행하고 산점도를 그린다.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 1. 데이터 로드
df = pd.read_csv("social_survey_w07.csv")

# 2. 산점도 그리기
plt.scatter(df["교육연수"], df["소득"], alpha=0.5)
plt.xlabel("교육연수")
plt.ylabel("소득(만원)")
plt.title("교육연수 vs 소득")
plt.show()

# 3. 단순회귀 실행
X = sm.add_constant(df["교육연수"])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
```

**완료 체크**: 산점도가 출력되었는가? summary()에서 β₀, β₁, R², p값을 확인할 수 있는가?

#### ② 확인

회귀 결과를 확인하고, 잔차 플롯을 그린다.

**2-1. 핵심 수치 기록**: summary() 출력에서 다음을 찾아 기록한다.

| 항목 | 기록 |
|---|---|
| β₀(절편) | ___ |
| β₁(기울기) | ___ |
| β₁의 p값 | ___ |
| R² | ___ |

**2-2. 잔차 플롯 확인**:

```python
# 잔차 플롯
residuals = model.resid
plt.scatter(model.fittedvalues, residuals, alpha=0.5)
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("예측값")
plt.ylabel("잔차")
plt.title("잔차 플롯")
plt.show()
```

잔차가 0을 중심으로 패턴 없이 흩어져 있는지 확인한다.

**2-3. Copilot 프롬프트 (필수 1)**: Copilot에게 다음을 질문한다.

> "이 회귀 결과에서 β₁ = 45.8, p < 0.001이다. 이것을 '인과'처럼 읽히지 않게 해석문을 써줘."

Copilot의 답변을 읽고, **1가지를 채택 또는 거절**한다. 채택/거절 이유를 결과 근거로 2문장 이내로 작성한다.

**완료 체크**: 핵심 수치 4개를 기록했는가? 잔차 플롯에 뚜렷한 패턴이 있는가? Copilot 채택/거절 기록이 있는가?

#### ③ 해석

A–E 형식으로 해석문을 작성한다.

**제출물 구성:**

- **(A) 분석 질문** (1–2문장): "교육 연수는 소득을 통계적으로 유의하게 예측하는가?"
- **(B) 사용 변수/수준**: 교육연수(비율척도, 연속형), 소득(비율척도, 연속형, 단위: 만 원)
- **(C) 결과 근거**: 산점도 + β₀ = 85.2, β₁ = 45.8(p < 0.001) + R² = 0.285 + 잔차 플롯
- **(D) 해석문** (5–7문장): 아래 모범 구조를 참고하여 직접 작성한다.
- **(E) Copilot 사용 요약** (2–3줄): 어떤 질문을 했고, 어떤 답변을 받았으며, 무엇을 채택/거절했는가?

### 확장(Plus) 트랙

기본 트랙을 완료한 학생은 추가로 다음을 수행한다.

> 동일 데이터에서 **연령 → 생활만족도** 단순회귀를 실행한다. 산점도 + 회귀 계수 + R² + 잔차 플롯까지 수행한 뒤, **교육 연수 → 소득 모형과 R²를 비교**하여 다음 질문에 2문장으로 답한다.
>
> "두 모형 중 어느 쪽의 설명력이 더 높은가? R²를 근거로 서술하고, R²가 낮은 모형은 의미가 없는지 판단하라."

### Copilot 프롬프트

| 구분 | 프롬프트 |
|---|---|
| 필수 1 | "이 회귀 결과에서 β₁ = 45.8, p < 0.001이다. 이것을 '인과'처럼 읽히지 않게 해석문을 써줘." |
| 필수 2 | "R² = 0.285이면 모델이 쓸모없는 건가? 사회과학에서 R² 기준을 알려줘." |
| 선택 | "이 잔차 플롯을 보고 회귀 가정이 충족되는지 판단해줘." |

- 필수 프롬프트 2개는 실습 중 반드시 사용하고, 결과를 (E)항에 기록한다.
- 선택 프롬프트는 잔차 플롯 해석에 도움이 필요할 때 활용한다.

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 해석문에 해당하는 항목이 없는지 점검해 보자.

**오류 1: β₁을 인과로 해석**

가장 흔하고 가장 심각한 오류이다. "교육 연수가 소득에 영향을 미친다"는 인과 표현이다. 회귀 분석의 β₁은 "X가 1단위 증가할 때 Y가 평균적으로 얼마 변하는가"라는 **연관의 크기**를 보여줄 뿐이다.

| 잘못된 표현 (X) | 올바른 표현 (O) |
|---|---|
| "교육이 소득에 영향을 미친다" | "교육 연수와 소득 사이에 유의한 양의 연관이 관찰된다" |
| "교육을 받으면 소득이 높아진다" | "교육 연수가 높을수록 소득이 높은 경향이 추정된다" |
| "교육이 소득을 결정한다" | "교육 연수는 소득의 유의한 예측 변수이다" |

**오류 2: R²만 보고 모형의 가치를 판단**

"R² = 0.285라서 모형이 별로 좋지 않다"고 결론짓는 학생이 많다. 사회과학에서 단일 변수로 R² = 0.285는 보통 수준의 설명력이다.

**확인법**: "낮으므로 의미 없다"가 아니라 "소득 변동의 28.5%를 설명하며, 나머지는 다른 요인(경력, 직종, 지역 등)에 의해 설명될 수 있다"로 서술한다.

**오류 3: 잔차 플롯 확인을 건너뜀**

계수와 R²만 확인하고 잔차 플롯을 그리지 않는 경우가 많다. 잔차에 패턴이 있으면 선형 모형의 가정이 위반된 것이다.

**확인법**: `model.resid`를 이용하여 잔차 플롯을 반드시 그리고, 패턴 유무를 눈으로 확인한다.

### 해석문 모범 구조

오늘 과제의 (D)항 해석문은 다음 4단계 구조를 따른다.

>
> **(핵심근거)** β₁ = 45.8(p < 0.001)이며, R² = 0.285이다. 교육 연수가 1년 증가할 때 소득은 평균 약 45.8만 원 높은 것으로 추정된다. 잔차 플롯에서 뚜렷한 패턴은 관찰되지 않았다.
>
> **(결론)** 유의수준 0.05에서 교육 연수는 소득의 통계적으로 유의한 예측 변수이다. 이 모형은 소득 변동의 약 28.5%를 설명하며, 사회과학에서 단일 변수 기준으로 보통 수준의 설명력에 해당한다.
>
> **(주의)** 이 결과는 교육이 소득을 "결정"한다는 의미가 아니다. 부모의 경제적 배경, 거주 지역, 직종 등 교란변수를 통제하지 않았으며, 역인과(소득이 높은 가정에서 교육 기간이 더 긴 경향)의 가능성도 존재한다.

## A회차 빠른 진단 정답

**Q.** "교육 연수가 1년 증가하면 소득이 50만 원 증가한다"는 회귀 분석 결과이다. 이 문장이 의미하는 것으로 올바른 것은?

**정답**: ② 교육 연수가 1년 증가할 때 소득이 평균적으로 50만 원 증가하는 경향이 관찰된다. 회귀 계수는 인과가 아닌 연관/예측 관계를 나타내며, "평균적으로" 증가하는 경향이라는 표현이 적절하다.

## B회차 빠른 진단 정답

**Q.** "회귀 분석 결과 β₁ = 45.8, p < 0.001, R² = 0.285이다. '교육이 소득을 높인다'고 해석해도 되는가?"

**정답**: 해석할 수 없다. 회귀 분석은 두 변수의 연관/예측 관계를 보여줄 뿐, 인과를 증명하지 않는다. "교육 연수가 높을수록 소득이 높은 경향이 관찰된다"가 올바른 표현이다. 부모의 경제적 배경, 거주 지역, 직종 등 교란변수를 통제하지 않았으며, 역인과(소득이 높은 가정에서 교육 기간이 더 긴 경향)의 가능성도 존재한다.

## 과제 1 기대 결과: 교육연수 → 소득

### 기대 코드

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv("social_survey_w07.csv")

# 산점도
plt.scatter(df["교육연수"], df["소득"], alpha=0.5)
plt.xlabel("교육연수")
plt.ylabel("소득(만원)")
plt.title("교육연수 vs 소득")
plt.show()

# 단순회귀
X = sm.add_constant(df["교육연수"])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())

# 잔차 플롯
plt.scatter(model.fittedvalues, model.resid, alpha=0.5)
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("예측값")
plt.ylabel("잔차")
plt.title("잔차 플롯")
plt.show()
```

### 기대 수치

**산점도**: 교육 연수가 높을수록 소득이 높아지는 양의 경향이 관찰된다. 점들이 오른쪽 위를 향해 흩어져 있다.

**회귀 결과**:

| 항목 | 값 |
|---|---|
| β₀(절편) | 85.2 |
| β₁(기울기) | 45.8 |
| β₁의 p값 | < 0.001 |
| R² | 0.285 |
| 잔차 패턴 | 패턴 없음 (무작위 분포) |
| 귀무가설 기각 여부 | 기각 (alpha = 0.05) |

**계수 해석**: 교육 연수가 1년 증가할 때, 소득은 평균 약 45.8만 원 증가하는 것으로 추정된다. 이 모형은 소득 변동의 약 28.5%를 교육 연수로 설명한다.

**잔차 해석**: 잔차 플롯에서 뚜렷한 패턴(깔때기 모양, U자 모양 등)은 관찰되지 않으며, 선형성 가정이 대체로 충족된 것으로 판단된다.

## 흔한 오류 3개 + 확인법

### 오류 1: β₁을 인과로 해석

"교육이 소득에 영향을 미친다", "교육을 받으면 소득이 올라간다" 등의 인과 표현이 해석문에 포함되는 경우가 가장 빈번하다.

**확인법**: 해석문(D항)에서 아래 단어를 검색하여 수정한다.
- "영향" -> "연관"
- "높인다/올린다" -> "높은 것으로 추정된다"
- "때문에" -> "에 따라"
- "결정/원인" -> "예측 변수"

### 오류 2: R²만으로 모형의 가치를 단정

"R² = 0.285이라서 좋지 않다"고 결론짓거나, "R² = 0.008이라서 이 분석은 의미 없다"고 판단하는 경우가 많다.

**확인법**: R²는 반드시 맥락과 함께 해석한다. 사회과학에서 단일 변수 기준으로 R² = 0.285는 보통 수준의 설명력이다. R²가 낮더라도, "유의한 관계가 없다"는 결과 자체가 의미 있는 발견이다.

### 오류 3: 유의하지 않은 결과를 "실패"로 간주

과제 2에서 p = 0.215로 유의하지 않은 결과가 나오면, "분석을 잘못했다" 또는 "다른 변수를 찾아야 한다"고 반응하는 학생이 있다.

**확인법**: 유의하지 않은 결과는 "이 표본에서 연령과 생활만족도 사이에 통계적으로 유의한 선형 관계의 증거를 찾지 못했다"는 의미이며, 이 자체가 보고할 가치가 있는 결과이다. "관계가 없다"고 단정하는 것은 오류이다.

## 비판적 읽기 + 토론 가이드

### 비판적 읽기 자료

아래 가상 뉴스 기사에서 문제점을 확인한다.

> **<OO경제> 2026.03.10**
>
> *"한국노동패널 분석 결과, 교육 연수가 1년 증가할 때 월 소득은 평균 48만 원 증가하는 것으로 나타났다(R² = 0.31). 이는 **교육이 소득 향상의 핵심 원인**임을 보여주며, **학력을 높이면 확실히 소득이 올라간다**는 점을 입증한다."*

**문제점 3가지**:

1. **인과 표현**: "교육이 소득 향상의 핵심 **원인**"이라고 서술했다. 관찰 데이터의 회귀 분석으로는 연관/예측만 확인할 수 있다. "교육 연수와 소득 사이에 통계적으로 유의한 양의 연관이 **관찰되었다**"가 적절하다.

2. **역인과 가능성 무시**: 소득이 높은 가정에서 교육 기간이 더 길어질 수 있다. 이 가능성이 전혀 언급되지 않았다.
