> 한 변수가 다른 변수를 얼마나 '예측'하는가? (주의: 예측 ≠ 원인)

**학습 목표**
1. **단순회귀**(simple linear regression)의 모형 구조(Y = β₀ + β₁X + ε)를 이해하고, 계수(β₀, β₁)를 구체적인 문장으로 해석할 수 있다.
2. **적합도(R²)**의 의미를 설명하고, R²가 높아도 인과가 아니며 R²가 낮아도 관계가 무의미하지 않다는 점을 구별할 수 있다.
3. 회귀 결과를 **인과가 아닌 예측/연관**으로 해석하되, 역인과·교란변수·허위상관의 가능성을 명시하는 해석문을 작성할 수 있다.

## 핵심 개념 학습

### 왜 필요한가

- **6주차까지 학습한 질문**:
  - "차이가 있는가?" (t-검정, ANOVA)
  - "연관이 있는가?" (카이자승)
- **이번 주 질문**: **"X가 변할 때 Y가 얼마나 변하는가?"**
  - 사회과학에서 자주 등장하는 질문 유형
- **이번 주 학습 도구**: **단순회귀(simple linear regression)**
  - 산점도에 가장 잘 맞는 직선 긋기
  - 직선의 기울기(β₁)와 적합도(R²) 해석
  - "이 결과가 인과를 의미하지 않는다"는 점 명확히 하기

---

### 핵심 개념 1: 단순회귀(simple linear regression)

- **단순회귀(simple linear regression)** 정의
  - 하나의 독립변수(X)로 하나의 종속변수(Y)를 예측하는 선형 모형

- **모형 구조**:

> Y = β₀ + β₁X + ε

- **각 요소의 의미**:
  - **β₀(절편, intercept)**: X = 0일 때 Y의 기대값
  - **β₁(기울기, slope)**: X가 1단위 증가할 때 Y의 **평균 변화량** ← 회귀 분석의 핵심
  - **ε(오차항, error)**: 모형이 설명하지 못하는 개인별 차이

- **계수 해석 연습**:
  - 단순히 "β₁ = 45.8" (✕)
  - 구체적 문장으로 작성:

> "교육 연수가 1년 증가하면, 월 소득은 평균 약 45.8만 원 증가하는 것으로 추정된다." (✓)

---

#### 산점도 위의 직선 — 예측 투표

> **투표 질문**: "교육 연수(X)와 소득(Y)의 산점도에 직선을 하나 그었을 때, 교육 연수가 1년 증가하면 소득은 대략 얼마 증가할까?"
>
> - ① 약 10만 원  ② 약 40만 원  ③ 약 100만 원
>
> *(10초)*

- **정답**: 회귀 분석 실행 시 β₁ ≈ 45.8 → **② 약 40만 원**
- **직선 산출 방법**: **최소자승법(OLS, Ordinary Least Squares)**
  - "모든 점에서 직선까지의 수직 거리(잔차)의 제곱 합이 최소가 되는 직선"

---

### 핵심 개념 2: 적합도(R²)

- **적합도(R², coefficient of determination)** 정의
  - "회귀 모형이 Y의 전체 변동 중 몇 퍼센트를 설명하는가?"를 나타내는 지표
  - 예: R² = 0.293 → "이 모형이 소득 변동의 29.3%를 교육 연수로 설명한다"
  - R² 범위: 0(전혀 설명 못 함) ~ 1(완벽히 설명)

- **⚠️ 반드시 주의할 점 2가지**:

1. **R²가 높아도 인과가 아니다**
   - R² = 0.90이라 해도 X가 Y를 "일으킨다"는 뜻 아님

2. **R²가 낮아도 관계가 무의미하지 않다**
   - 사회과학에서 R² = 0.10~0.30은 흔한 값
   - 인간 행동 = 다양한 요인에 의해 결정
   - ∴ 단일 변수의 설명력이 낮은 것은 자연스러운 일

| R² 범위 | 사회과학에서의 일반적 해석 |
|---:|---|
| < 0.10 | 낮은 설명력 (그러나 이론적으로 의미 있을 수 있음) |
| 0.10 – 0.30 | 보통 수준의 설명력 |
| > 0.30 | 높은 설명력 (단일 변수 기준으로는 상당히 높음) |

---

### 핵심 개념 3: 인과 오해 방지

- 이 과목 전체에서 **가장 중요한 메시지**:

> **회귀 분석 = 연관/예측이지, 인과가 아니다.**

- **"인과라고 말할 수 있는 조건" 체크리스트** (3가지):
  1. **시간적 선후관계**: 원인(X)이 결과(Y)보다 시간적으로 먼저 발생
  2. **교란변수 통제**: X와 Y 모두에 관련된 제3의 변수 통제
  3. **실험 설계(무작위 배정)**: X를 무작위로 배정하여 다른 조건 동일하게

- **관찰 데이터(설문 등)의 한계**:
  - 위 조건 충족 매우 어려움
  - ∴ 회귀 결과 서술 시:
    - 사용 권장: "연관", "예측", "추정"
    - 사용 금지: "원인", "영향", "결정"

---

#### 인과 오해 사례 3가지

| 사례 | 관찰된 상관 | 실제 구조 |
|---|---|---|
| 아이스크림 판매량과 익사 사고 수 | 양의 상관 | **교란변수**(기온)가 둘 다에 관련 |
| 병원에 자주 가는 사람이 더 아프다 | 양의 상관 | **역인과**: 아프기 때문에 병원에 가는 것 |
| 국가별 초콜릿 소비와 노벨상 수 | 양의 상관 | **허위상관**: 국가의 경제 수준이 교란변수 |

### 예시: 교육 연수 → 소득 단순회귀

- 아래는 social_survey_w07.csv(200명)를 이용하여 교육 연수로 소득을 예측하는 단순회귀 분석의 핵심 코드이다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("data/social_survey_w07.csv", encoding='utf-8')
X = sm.add_constant(df["교육연수"])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
```

- _전체 코드는 practice/W07/code/W07-1-simple-regression.py 참고_

**검정 결과:**

- **β₀(절편) = 85.6**: 교육 연수가 0년일 때 소득의 기대값 = 85.6만 원
- **β₁(기울기) = 45.8, p < 0.001**: 교육 연수 1년 증가 시 → 소득 평균 약 45.8만 원 증가 추정
- **R² = 0.293**: 이 모형이 소득 변동의 29.3%를 교육 연수로 설명

- **잔차 확인**:
  - 잔차(실제값 - 예측값)가 특정 패턴 없이 무작위로 흩어져 있으면 → 선형 모형 가정 대체로 충족
  - 잔차가 깔때기 모양이면 → **이분산(heteroscedasticity)** 문제 가능성

---

**결과 해석:**

- **(분석목적)** 교육 연수가 소득을 얼마나 예측하는지 확인하기 위해 단순회귀 분석 실행
- **(핵심근거)**
  - β₁ = 45.8 (p < 0.001)
  - R² = 0.293
  - 교육 연수 1년 증가 시 소득은 평균 약 45.8만 원 높음
- **(결론)** 유의수준 0.05에서 교육 연수는 소득의 통계적으로 유의한 예측 변수, 소득 변동의 약 29.3% 설명
- **(주의)**
  - 이 결과 ≠ 교육이 소득을 "결정"
  - 부모의 경제적 배경, 거주 지역, 직종 등 교란변수 미통제
  - 역인과 가능성 배제 불가

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

- 실습 중 자주 나타나는 오류 정리
- 자신의 해석문에 해당 항목이 없는지 점검 필요

**오류 1: β₁을 인과로 해석**

- 가장 흔하고 가장 심각한 오류
- "교육 연수가 소득에 영향을 미친다" = 인과 표현
- 회귀 분석의 β₁ = "X가 1단위 증가할 때 Y가 평균적으로 얼마 변하는가"라는 **연관의 크기**

| 잘못된 표현 (✕) | 올바른 표현 (✓) |
|---|---|
| "교육이 소득에 영향을 미친다" | "교육 연수와 소득 사이에 유의한 양의 연관이 관찰된다" |
| "교육을 받으면 소득이 높아진다" | "교육 연수가 높을수록 소득이 높은 경향이 추정된다" |
| "교육이 소득을 결정한다" | "교육 연수는 소득의 유의한 예측 변수이다" |

**오류 2: R²만 보고 모형의 가치를 판단**

- 문제: "R² = 0.293이라서 모형이 별로 좋지 않다"고 결론
- 사실: 사회과학에서 단일 변수로 R² = 0.293 = 보통 수준의 설명력
- **해결**: "낮으므로 의미 없다" (✕) → "소득 변동의 29.3%를 설명하며, 나머지는 다른 요인(경력, 직종, 지역 등)에 의해 설명될 수 있다" (✓)

**오류 3: 잔차 플롯 확인을 건너뜀**

- 문제: 계수와 R²만 확인, 잔차 플롯 미작성
- 사실: 잔차에 패턴 있으면 → 선형 모형 가정 위반
- **해결**: `model.resid` 이용하여 잔차 플롯 반드시 작성 → 패턴 유무 눈으로 확인
