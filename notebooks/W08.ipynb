# W08A. 여러 변수를 동시에 고려하면 해석이 어떻게 달라지는가? — 다중회귀

> 여러 변수를 동시에 고려하면 해석이 어떻게 달라지는가?

**학습 목표**
1. **다중회귀(multiple regression)**의 모형 구조(Y = β₀ + β₁X₁ + β₂X₂ + ... + ε)를 이해하고, 각 계수를 "다른 변수를 통제한 상태에서"라는 조건과 함께 해석할 수 있다.
2. **통제(control)**의 의미가 "통계적 조정"일 뿐 "인과 해결"이 아님을 설명하고, 관찰 데이터에서 통제의 한계를 서술할 수 있다.
3. **다중공선성(multicollinearity)**을 VIF로 진단하고, 단순회귀 대비 다중회귀에서 계수가 변하는 이유를 설명할 수 있다.

## 오늘의 질문/목표/진단

### 오늘의 질문

7주차에서는 하나의 독립변수로 종속변수를 예측하는 **단순회귀**를 배웠다. 그런데 현실에서 소득은 교육 연수 하나만으로 결정되지 않는다. 경력, 성별, 거주 지역 등 여러 요인이 동시에 작용한다. 단순회귀에서 교육 연수의 계수(β₁ = 45.8)는 이런 다른 변수를 고려하지 않은 값이다. 만약 경력 연수를 함께 모형에 넣으면, 교육 연수의 계수는 어떻게 변할까? 오늘은 독립변수를 **여러 개** 동시에 투입하는 **다중회귀(multiple regression)**를 배우고, "통제"라는 표현이 정확히 무엇을 의미하며 무엇을 의미하지 않는지를 정리한다.

### 빠른 진단 (1문항)

> **Q.** 단순회귀에서 교육연수의 β₁ = 45.8이었다. 경력연수를 추가한 다중회귀에서 교육연수의 β₁ = 38.4로 줄었다. 이 변화의 가장 적절한 해석은?
>
> ① 교육연수의 효과가 실제로 줄어들었다
> ② 경력연수가 교육연수의 효과를 일부 "빼앗아" 설명하고 있다
> ③ 다중회귀의 결과가 더 부정확하다

## 핵심 개념 학습

### 왜 필요한가

7주차 단순회귀에서 "교육 연수가 1년 증가하면 소득이 평균 45.8만 원 높다"는 결과를 얻었다. 그런데 이 45.8에는 경력 연수의 효과가 섞여 있을 수 있다. 교육 연수가 긴 사람은 경력도 함께 쌓이는 경향이 있기 때문이다. 이처럼 하나의 독립변수만 넣으면, 그 변수에 다른 변수의 효과까지 합산되어 계수가 **과대 추정**될 수 있다. 여러 독립변수를 동시에 모형에 넣으면, 각 변수의 "순수한" 연관(다른 변수를 고정한 상태에서의 연관)을 분리할 수 있다. 이것이 **다중회귀(multiple regression)**가 필요한 이유이다.

### 핵심 개념 1: 다중회귀(multiple regression)

**다중회귀(multiple regression)**란 둘 이상의 독립변수(X₁, X₂, ...)로 하나의 종속변수(Y)를 예측하는 선형 모형이다. 모형의 구조는 다음과 같다.

> Y = β₀ + β₁X₁ + β₂X₂ + ... + ε

각 계수의 의미는 단순회귀와 미묘하지만 결정적으로 다르다.

- **β₁**: "**다른 모든 독립변수를 일정하게 유지한 상태에서** X₁이 1단위 증가할 때 Y의 평균 변화량"
- 이를 **편회귀계수(partial regression coefficient)**라고 한다.

쉽게 말해서, 단순회귀의 β₁은 "교육 연수만 보았을 때"의 효과이고, 다중회귀의 β₁은 "경력 연수를 같은 수준으로 맞춘 뒤 교육 연수만 비교했을 때"의 효과이다.

**표 W08.1** 단순회귀 vs 다중회귀 계수 비교 (동일 데이터, social_survey_w08.csv)

| 모형 | 교육연수 β₁ | 경력연수 β₂ | 성별 β₃ | R² |
|---|---:|---:|---:|---:|
| 단순회귀 (교육연수만) | 54.1 | — | — | 0.550 |
| 다중회귀 (교육연수 + 경력연수) | 38.4 | 12.3 | — | 0.617 |
| 다중회귀 (교육연수 + 경력연수 + 성별) | 36.8 | 12.0 | 41.8 | 0.625 |

교육연수의 β₁이 54.1 → 38.4 → 36.8로 줄어든다. 이는 단순회귀에서 교육연수의 계수에 경력연수와 성별의 효과가 섞여 있었다는 뜻이다. 변수를 추가할수록 교육연수의 "순수한" 연관만 남게 된다.

### 핵심 개념 2: 통제(control)의 의미와 한계

**통제(control)**란 "다른 변수를 일정하게 유지한 상태에서"라는 뜻이다. 다중회귀에서 "경력연수를 통제한 상태에서 교육연수의 효과"라고 말할 때, 이는 **통계적 조정(statistical adjustment)**을 의미한다. 실제로 경력연수를 고정한 실험을 한 것이 아니다.

이 구분이 왜 중요한가? 많은 연구 보고서에서 "통제했다"는 표현을 마치 인과를 증명한 것처럼 사용하기 때문이다.

**통제 ≠ 인과 해결**인 이유:

1. **측정되지 않은 교란변수(omitted variable bias)**: 모형에 포함하지 않은 변수(예: 부모의 경제적 배경, 거주 지역, 직종)가 여전히 결과에 영향을 줄 수 있다. 아무리 많은 변수를 통제해도, 빠진 변수가 있다면 계수는 편향될 수 있다.
2. **통계적 조정 ≠ 실험적 통제**: 실험에서는 무작위 배정으로 다른 조건을 동일하게 만들지만, 관찰 데이터에서는 통계적으로 "조정"할 뿐이다. 이 조정은 측정된 변수에 한해서만 가능하다.

| 표현 | 의미 | 인과 증명? |
|---|---|---|
| "경력을 통제한 상태에서 교육의 효과" | 경력이 같은 사람끼리 비교했을 때 교육의 연관 | 아니다 |
| "경력을 통제했으므로 교육이 원인이다" | 잘못된 해석 | — |

### 핵심 개념 3: 다중공선성(multicollinearity)

**다중공선성(multicollinearity)**이란 독립변수들끼리 강하게 상관되어 있는 상태를 말한다. 예를 들어 "교육연수"와 "최종학력(고졸/대졸/대학원졸)"을 동시에 모형에 넣으면, 두 변수가 거의 같은 정보를 담고 있어 계수 추정이 불안정해진다.

**VIF(Variance Inflation Factor, 분산팽창인자)**로 진단한다.

- **VIF < 5**: 공선성 문제 없음
- **VIF 5–10**: 주의 필요
- **VIF > 10**: 심각한 공선성 문제, 변수 제거 또는 결합 고려

오늘 데이터(social_survey_w08.csv)의 VIF를 확인하면 다음과 같다.

| 변수 | VIF |
|---|---:|
| 교육연수 | 1.75 |
| 경력연수 | 1.70 |
| 성별_더미 | 1.08 |

모두 10 미만이므로 다중공선성 문제는 없다.

### 예시: 소득 ~ 교육연수 + 경력연수 + 성별 다중회귀

아래는 social_survey_w08.csv(200명)를 이용하여 교육연수, 경력연수, 성별로 소득을 예측하는 다중회귀의 핵심 코드이다.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("social_survey_w08.csv")
df["성별_더미"] = (df["성별"] == "남성").astype(int)
X = sm.add_constant(df[["교육연수", "경력연수", "성별_더미"]])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
for i, col in enumerate(X.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X.values, i+1):.2f}")
```

_전체 코드는 practice/W08/code/W08-1-multiple-regression.py 참고_

**검정 결과:**

- **교육연수 β₁ = 36.8, p < 0.001**: 경력연수와 성별을 통제한 상태에서, 교육연수가 1년 증가할 때 소득은 평균 약 36.8만 원 높은 것으로 추정된다.
- **경력연수 β₂ = 12.0, p < 0.01**: 교육연수와 성별을 통제한 상태에서, 경력연수가 1년 증가할 때 소득은 평균 약 12.0만 원 높은 것으로 추정된다.
- **성별 β₃ = 41.8, p < 0.05**: 교육연수와 경력연수를 통제한 상태에서, 남성의 소득이 여성보다 평균 약 41.8만 원 높은 것으로 추정된다.
- **R² = 0.625**: 이 모형은 소득 변동의 62.5%를 설명한다.

**결과 해석:**

**(분석목적)** 교육연수, 경력연수, 성별이 소득을 얼마나 예측하는지, 각 변수의 "순수한" 연관을 확인하기 위해 다중회귀 분석을 실행했다. **(핵심근거)** 교육연수 β₁ = 36.8(p < 0.001), 경력연수 β₂ = 12.0(p < 0.01), 성별 β₃ = 41.8(p < 0.05)이며, R² = 0.625이다. 단순회귀 대비 교육연수의 계수가 54.1에서 36.8로 감소하였다. **(결론)** 유의수준 0.05에서 세 변수 모두 소득의 통계적으로 유의한 예측 변수이며, 모형은 소득 변동의 약 62.5%를 설명한다. **(주의)** 이 결과는 교육이 소득을 "결정"한다거나 성별이 소득 차이의 "원인"이라는 의미가 아니다. 부모의 경제적 배경, 직종, 거주 지역 등 측정되지 않은 교란변수가 여전히 존재할 수 있다.

## 실습 과제 (수업 내 완결형)

### 과제 개요

오늘 실습에서는 동일 데이터(social_survey_w08.csv)로 단순회귀와 다중회귀를 모두 실행하고, 계수 변화를 관찰하며, VIF를 확인한 뒤 해석문을 작성한다.

### 기본(Basic) 트랙

#### ① 실행

단순회귀(교육연수만)와 다중회귀(교육연수 + 경력연수)를 실행한다.

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("social_survey_w08.csv")

# 1. 단순회귀: 소득 ~ 교육연수
X1 = sm.add_constant(df["교육연수"])
model1 = sm.OLS(df["소득"], X1).fit()
print("=== 단순회귀 ===")
print(model1.summary())

# 2. 다중회귀: 소득 ~ 교육연수 + 경력연수
X2 = sm.add_constant(df[["교육연수", "경력연수"]])
model2 = sm.OLS(df["소득"], X2).fit()
print("\n=== 다중회귀 ===")
print(model2.summary())
```

| 선택 | "이 다중회귀 결과를 '통제했다'는 표현 없이 정확하게 해석하는 문장을 써줘." |

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 해석문에 해당하는 항목이 없는지 점검해 보자.

**오류 1: "통제했다 = 인과를 증명했다"**

가장 흔하고 가장 심각한 오류이다. "경력을 통제했으므로 교육이 소득의 원인이다"라고 쓰는 학생이 많다. 관찰 데이터에서의 통제는 통계적 조정일 뿐, 측정되지 않은 교란변수가 존재할 수 있다.

| 잘못된 표현 (X) | 올바른 표현 (O) |
|---|---|
| "경력을 통제했으므로 교육이 원인이다" | "경력을 통계적으로 조정한 상태에서 교육과 소득의 양의 연관이 관찰된다" |
| "변수를 통제하면 인과를 알 수 있다" | "통제는 통계적 조정이며, 인과 증명에는 실험 설계가 필요하다" |

**오류 2: 단순회귀와 다중회귀 계수를 비교하지 않음**

다중회귀만 실행하고, 단순회귀 대비 계수가 어떻게 변했는지 확인하지 않는 경우가 많다. 계수 변화는 "변수를 추가하면 해석이 어떻게 달라지는가?"를 보여주는 핵심이다.

**확인법**: 반드시 단순회귀와 다중회귀를 모두 실행하고, 교육연수의 β₁ 변화를 기록한다.

**오류 3: VIF 확인을 건너뜀**

계수와 R²만 확인하고 VIF를 계산하지 않는 경우가 많다. VIF가 높으면 계수 추정이 불안정하여 해석이 신뢰할 수 없게 된다.

**확인법**: `variance_inflation_factor`를 사용하여 각 독립변수의 VIF를 반드시 확인하고, 10 미만인지 점검한다.

### 해석문 모범 구조

오늘 과제의 (D)항 해석문은 다음 4단계 구조를 따른다.

> **(분석목적)** 경력연수를 통제했을 때 교육연수와 소득의 연관이 어떻게 달라지는지 확인하기 위해, 단순회귀와 다중회귀를 비교 실행했다.
>
> **(핵심근거)** 단순회귀에서 교육연수 β₁ = 54.1이었으나, 경력연수를 추가한 다중회귀에서 β₁ = 38.4로 감소했다. 경력연수 β₂ = 12.3(p < 0.01)이며, R²는 0.550에서 0.617로 증가했다. VIF는 두 변수 모두 2 미만으로 다중공선성 문제는 없다.
>
> **(결론)** 유의수준 0.05에서 교육연수와 경력연수는 모두 소득의 유의한 예측 변수이다. 경력연수를 함께 고려하면 교육연수의 계수가 줄어들며, 이는 단순회귀에서 교육연수의 계수에 경력연수의 효과가 일부 섞여 있었음을 의미한다.
>
> **(주의)** 이 결과는 교육이나 경력이 소득을 "결정"한다는 인과적 의미가 아니다. 부모의 경제적 배경, 직종, 거주 지역 등 측정되지 않은 교란변수가 여전히 존재할 수 있으며, 관찰 데이터에서 "통제"는 통계적 조정에 불과하다.

````python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
````

````python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("data/social_survey_w08.csv", encoding='utf-8')
df["성별_더미"] = (df["성별"] == "남성").astype(int)
X = sm.add_constant(df[["교육연수", "경력연수", "성별_더미"]])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
for i, col in enumerate(X.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X.values, i+1):.2f}")
````

**정답**: ② 경력연수가 교육연수의 효과를 일부 "빼앗아" 설명하고 있다. 단순회귀에서 교육연수의 계수에는 경력연수의 효과가 일부 포함되어 있었으며, 경력을 모형에 추가하면 교육연수의 "순수한" 연관만 남게 되어 계수가 감소한다.

## B회차 빠른 진단 정답

**Q.** "단순회귀에서 교육연수 β₁ = 54.1이고, 경력연수를 추가한 다중회귀에서 β₁ = 38.4로 줄었다. 이 변화가 의미하는 바를 1문장으로 설명하라."

**정답**: 단순회귀에서 교육연수의 계수에는 경력연수의 효과가 일부 포함되어 있었으며, 경력연수를 모형에 추가하여 통계적으로 조정하면 교육연수의 "순수한" 연관만 남게 되므로 계수가 줄어든다. 이는 교육연수가 긴 사람은 경력도 함께 쌓이는 경향이 있기 때문이다.

## 과제 1 기대 결과: 단순회귀 → 다중회귀 계수 변화

### 기대 코드

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("social_survey_w08.csv")

# 단순회귀: 소득 ~ 교육연수
X1 = sm.add_constant(df["교육연수"])
model1 = sm.OLS(df["소득"], X1).fit()
print("=== 단순회귀 ===")
print(model1.summary())

# 다중회귀: 소득 ~ 교육연수 + 경력연수
X2 = sm.add_constant(df[["교육연수", "경력연수"]])
model2 = sm.OLS(df["소득"], X2).fit()
print("\n=== 다중회귀 (2변수) ===")
print(model2.summary())
```

### 기대 수치

**단순회귀 (소득 ~ 교육연수)**:

| 항목 | 값 |
|---|---|
| β₀(절편) | 18.9 (참고값) |
| 교육연수 β₁ | 54.1 |
| 교육연수 p값 | < 0.001 |
| R² | 0.550 |

**다중회귀 (소득 ~ 교육연수 + 경력연수)**:

| 항목 | 값 |
|---|---|
| 교육연수 β₁ | 38.4 |
| 경력연수 β₂ | 12.3 |
| 교육연수 p값 | < 0.001 |
| 경력연수 p값 | < 0.001 |
| R² | 0.617 |

**계수 변화 해석**: 교육연수의 β₁이 54.1에서 38.4로 감소했다. 이는 단순회귀에서 교육연수의 계수에 경력연수의 효과가 포함되어 있었기 때문이다. 경력연수를 추가하여 통계적으로 조정하면, 교육연수의 "순수한" 연관(경력이 같은 수준인 사람끼리 비교했을 때의 교육 효과)만 남게 된다. R²도 0.550에서 0.617로 증가하여, 경력연수가 소득 변동의 추가적인 부분을 설명하고 있다.

## 과제 2 기대 결과: 3변수 다중회귀 + VIF

### 기대 코드

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 성별 더미변수 생성
df["성별_더미"] = (df["성별"] == "남성").astype(int)

# 다중회귀: 소득 ~ 교육연수 + 경력연수 + 성별_더미
X3 = sm.add_constant(df[["교육연수", "경력연수", "성별_더미"]])
model3 = sm.OLS(df["소득"], X3).fit()
print("=== 다중회귀 (3변수) ===")
print(model3.summary())

# VIF 확인
print("\n=== VIF ===")
for i, col in enumerate(X3.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X3.values, i+1):.2f}")
```

### 기대 수치

**다중회귀 (소득 ~ 교육연수 + 경력연수 + 성별_더미)**:

| 항목 | 값 |
|---|---|
| 교육연수 β₁ | 36.8 |
| 경력연수 β₂ | 12.0 |
| 성별_더미 β₃ | 41.8 |
| 교육연수 p값 | < 0.001 |
| 경력연수 p값 | < 0.001 |
| 성별_더미 p값 | < 0.05 |
| R² | 0.625 |

**VIF 결과**:

| 변수 | VIF | 판단 |
|---|---:|---|
| 교육연수 | 1.75 | 문제 없음 (< 10) |
| 경력연수 | 1.70 | 문제 없음 (< 10) |
| 성별_더미 | 1.08 | 문제 없음 (< 10) |

**VIF 해석**: 세 변수의 VIF가 모두 2 미만으로, 다중공선성 문제는 없다. 각 독립변수가 서로 독립적인 정보를 제공하고 있으며, 계수 추정이 안정적이다.

**성별 계수 해석**: β₃ = 41.8은 "교육연수와 경력연수를 통계적으로 조정한 상태에서, 남성의 소득이 여성보다 평균 약 41.8만 원 높은 것으로 추정된다"는 의미이다. 이는 성별이 소득의 유의한 예측 변수임을 시사하지만, 성별이 소득 차이의 "원인"이라는 뜻이 아니다. 직종, 근무 형태 등 측정되지 않은 교란변수가 존재할 수 있다.

## 모범 보고서 (A-E)

아래는 과제 1, 과제 2, 과제 3을 모두 포함한 모범 보고서이다.

### (A) 분석 질문

경력연수와 성별을 통계적으로 조정하면 교육연수와 소득의 연관은 어떻게 달라지는가?

### (B) 사용 변수/척도

- 교육연수: 비율척도 (6-18년, 연속형)
- 경력연수: 비율척도 (0-40년, 연속형)
- 소득: 비율척도 (만 원, 연속형)
- 성별: 명목척도 (남성/여성 → 더미변수 변환)

### (C) 결과 근거

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("social_survey_w08.csv")
df["성별_더미"] = (df["성별"] == "남성").astype(int)

# 단순회귀
X1 = sm.add_constant(df["교육연수"])
model1 = sm.OLS(df["소득"], X1).fit()

# 다중회귀 (2변수)
X2 = sm.add_constant(df[["교육연수", "경력연수"]])
model2 = sm.OLS(df["소득"], X2).fit()

# 다중회귀 (3변수)
X3 = sm.add_constant(df[["교육연수", "경력연수", "성별_더미"]])
model3 = sm.OLS(df["소득"], X3).fit()

print("단순회귀:", model1.summary())
print("다중회귀(2변수):", model2.summary())
print("다중회귀(3변수):", model3.summary())

for i, col in enumerate(X3.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X3.values, i+1):.2f}")
```

- 단순회귀: 교육연수 β₁ = 54.1(p < 0.001), R² = 0.550
- 다중회귀(2변수): 교육연수 β₁ = 38.4(p < 0.001), 경력연수 β₂ = 12.3(p < 0.001), R² = 0.617
- 다중회귀(3변수): 교육연수 β₁ = 36.8(p < 0.001), 경력연수 β₂ = 12.0(p < 0.001), 성별 β₃ = 41.8(p < 0.05), R² = 0.625
- VIF: 교육연수 1.75, 경력연수 1.70, 성별 1.08 (모두 10 미만)

### (D) 해석문

**(분석목적)** 경력연수와 성별을 통계적으로 조정했을 때 교육연수와 소득의 연관이 어떻게 달라지는지 확인하기 위해, 단순회귀와 다중회귀를 비교 실행했다.

**(핵심근거)** 단순회귀에서 교육연수 β₁ = 54.1이었으나, 경력연수를 추가한 다중회귀에서 β₁ = 38.4로, 성별까지 추가하면 β₁ = 36.8로 감소했다. R²는 0.550 → 0.617 → 0.625로 증가했다. VIF는 세 변수 모두 2 미만으로 다중공선성 문제는 없다.

**(결론)** 유의수준 0.05에서 교육연수, 경력연수, 성별은 모두 소득의 통계적으로 유의한 예측 변수이다. 변수를 추가할수록 교육연수의 계수가 감소하며, 이는 단순회귀에서 교육연수의 계수에 다른 변수의 효과가 섞여 있었음을 의미한다.

**(주의)** 이 결과는 교육이나 성별이 소득을 "결정"한다는 인과적 의미가 아니다. 관찰 데이터에서 "통제"는 통계적 조정에 불과하며, 부모의 경제적 배경, 직종, 거주 지역 등 측정되지 않은 교란변수가 여전히 존재할 수 있다.

### (E) Copilot 사용 요약

Copilot에게 "소득을 종속변수로, 교육연수를 독립변수로 단순회귀를 실행해줘"라고 요청하여 기본 코드를 생성했다. 다중회귀는 변수를 추가하여 동일한 방식으로 요청했다. VIF 계산 코드는 Copilot이 `variance_inflation_factor` 함수를 제안하여 활용했다. Copilot이 생성한 해석문에서 "교육이 소득에 영향을 미친다"는 인과 표현을 "교육연수와 소득 사이에 양의 연관이 관찰된다"로 직접 수정했다.

## 확장(Plus) 트랙 기대 결과

**표준화 계수 비교** (3변수 다중회귀):

표준화 계수(beta)는 각 변수를 표준화(평균 0, 표준편차 1)한 뒤 회귀를 실행하여 구한다. 변수의 단위가 다를 때 "어떤 변수가 Y에 더 강하게 연관되는가"를 비교할 수 있다.

| 변수 | 비표준화 β | 표준화 β (참고) |
|---|---:|---:|
| 교육연수 | 36.8 | 약 0.50 |
| 경력연수 | 12.0 | 약 0.33 |
| 성별_더미 | 41.8 | 약 0.10 |

**참고**: 비표준화 계수는 단위(교육: 년, 성별: 0/1)가 다르므로 직접 비교하기 어렵다. 표준화 계수로 비교하면 교육연수의 상대적 연관이 가장 큰 것으로 추정된다.

## 흔한 오류 3개 + 확인법

### 오류 1: "통제했다 = 인과를 증명했다"

"경력을 통제했으므로 교육이 소득의 원인이다"라고 해석하는 경우가 가장 빈번하다. 관찰 데이터에서의 통제는 통계적 조정일 뿐이며, 측정되지 않은 교란변수가 여전히 존재할 수 있다.

**확인법**: 해석문(D항)에서 아래 표현을 검색하여 수정한다.
- "통제했으므로 원인이다" → "통계적으로 조정한 상태에서 연관이 관찰된다"
- "인과를 확인했다" → "예측 관계를 확인했다"
- "영향을 미친다" → "유의한 예측 변수이다"

### 오류 2: 단순회귀 대비 계수 변화를 보고하지 않음

다중회귀만 실행하고, 단순회귀와 비교하지 않는 경우가 많다. 계수 변화는 "변수를 추가하면 해석이 어떻게 달라지는가"를 보여주는 핵심이다.

**확인법**: 단순회귀와 다중회귀의 교육연수 β₁을 반드시 나란히 기록하고, 변화 방향(증가/감소)과 그 이유를 1문장으로 서술한다.

### 오류 3: VIF 확인을 생략

다중공선성을 확인하지 않으면 계수의 신뢰성을 판단할 수 없다. VIF가 10 이상인 변수가 있으면 해당 변수를 제거하거나 결합하는 것을 고려해야 한다.

**확인법**: `variance_inflation_factor`를 사용하여 모든 독립변수의 VIF를 계산하고, 10 미만인지 확인한다.

## 비판적 읽기 + 토론 가이드

### 비판적 읽기 자료

아래 가상 연구 보고서에서 문제점을 확인한다.

> **<사회연구> 2026년 제42권**
>
> *"다중회귀 분석 결과, 교육 연수(β = 0.35, p < 0.001), 경력 연수(β = 0.22, p < 0.01), 성별(β = 0.18, p < 0.05)이 소득에 유의한 **영향을 미치는 것**으로 나타났다. 경력과 성별을 **통제함으로써** 교육이 소득의 **독립적 원인**임을 **확인**하였다."*

**문제점 3가지**:

1. **인과 표현**: "영향을 미치는 것", "독립적 원인"은 인과 표현이다. 관찰 데이터의 다중회귀로는 연관/예측만 확인할 수 있다. "교육연수는 소득의 유의한 예측 변수로 관찰되었다"가 적절하다.

2. **통제 = 인과 증명으로 오해**: "통제함으로써 원인임을 확인하였다"는 통제의 의미를 과대해석한 것이다. 통계적 조정은 측정된 변수에 한해서만 가능하며, 직종, 지역, 가정 배경 등 빠진 변수의 효과는 제거되지 않았다.

3. **측정되지 않은 교란변수 미언급**: 보고서에 모형의 한계가 전혀 서술되지 않았다. "부모의 경제적 배경, 직종 등 측정되지 않은 교란변수가 존재할 수 있다"는 한계를 반드시 명시해야 한다.

### 토론 진행 가이드

아래 질문에 대한 답을 확인한다.

**질문 1**: "단순회귀에서 교육연수 β₁ = 54.1인데 다중회귀에서 36.8로 줄었다. 단순회귀의 결과가 '틀린' 것인가?"
- 기대 답변: 틀린 것이 아니다. 단순회귀의 β₁은 "교육연수만 고려했을 때"의 연관이고, 다중회귀의 β₁은 "다른 변수를 통계적으로 조정한 후"의 연관이다. 질문의 성격이 다르며, 어느 쪽이 옳고 그른 것이 아니다.

**질문 2**: "변수를 많이 넣을수록 좋은 모형인가?"
- 기대 답변: 반드시 그렇지 않다. 이론적 근거 없이 변수를 무작위로 추가하면 과적합(overfitting)이 발생하거나, 다중공선성이 심해질 수 있다. 변수 선택은 이론 기반으로 하는 것이 원칙이다.

**질문 3**: "VIF가 3이면 문제인가?"
- 기대 답변: VIF 3은 일반적으로 문제로 보지 않는다. VIF > 5이면 주의가 필요하고, VIF > 10이면 심각한 공선성 문제로 판단한다. VIF 3 정도는 독립변수 간 약한 상관이 있다는 뜻이며, 계수 추정에 큰 영향을 주지 않는다.
