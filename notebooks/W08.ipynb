# W08A. 여러 변수를 동시에 고려하면 해석이 어떻게 달라지는가? — 중다회귀

> 여러 변수를 동시에 고려하면 해석이 어떻게 달라지는가?

**학습 목표**
1. **중다회귀(multiple regression)**의 모형 구조(Y = β₀ + β₁X₁ + β₂X₂ + ... + ε)를 이해하고, 각 계수를 "다른 변수를 통제한 상태에서"라는 조건과 함께 해석할 수 있다.
2. **통제(control)**의 의미가 "통계적 조정"일 뿐 "인과 해결"이 아님을 설명하고, 관찰 데이터에서 통제의 한계를 서술할 수 있다.
3. **다중공선성(multicollinearity)**을 VIF로 진단하고, 단순회귀 대비 중다회귀에서 계수가 변하는 이유를 설명할 수 있다.

## 핵심 개념 학습

### 왜 필요한가

- **7주차 복습**: 단순회귀에서 "교육 연수가 1년 증가하면 소득이 평균 54.1만 원 높다"는 결과
- **문제점**: 54.1에는 **경력 연수의 효과가 섞여 있을 수 있음**
  - 이유: 교육 연수가 긴 사람은 경력도 함께 쌓이는 경향
  - 결과: 하나의 독립변수만 넣으면 → 다른 변수의 효과까지 합산 → 계수 **과대 추정**
- **해결책**: **여러 독립변수를 동시에 모형에 넣기**
  - 각 변수의 "순수한" 연관(다른 변수를 고정한 상태에서의 연관) 분리 가능
  - → 이것이 **중다회귀(multiple regression)**가 필요한 이유

---

### 핵심 개념 1: 중다회귀(multiple regression)

- **중다회귀(multiple regression)** 정의
  - 둘 이상의 독립변수(X₁, X₂, ...)로 하나의 종속변수(Y)를 예측하는 선형 모형

- **모형 구조**:

> Y = β₀ + β₁X₁ + β₂X₂ + ... + ε

- **각 계수의 의미** — 단순회귀와 미묘하지만 결정적으로 다름
  - **β₁**: "**다른 모든 독립변수를 일정하게 유지한 상태에서** X₁이 1단위 증가할 때 Y의 평균 변화량"
  - = **편회귀계수(partial regression coefficient)**

- **단순회귀 vs 중다회귀 차이**:
  - 단순회귀 β₁: "교육 연수만 보았을 때"의 효과
  - 중다회귀 β₁: "경력 연수를 같은 수준으로 맞춘 뒤 교육 연수만 비교했을 때"의 효과

- **표 W08.1** 단순회귀 vs 중다회귀 계수 비교 (동일 데이터, social_survey_w08.csv)

| 모형 | 교육연수 β₁ | 경력연수 β₂ | 성별 β₃ | R² |
|---|---:|---:|---:|---:|
| 단순회귀 (교육연수만) | 54.1 | — | — | 0.550 |
| 중다회귀 (교육연수 + 경력연수) | 38.4 | 12.3 | — | 0.617 |
| 중다회귀 (교육연수 + 경력연수 + 성별) | 36.8 | 12.0 | 41.8 | 0.625 |

- **해석**: 교육연수 β₁이 54.1 → 38.4 → 36.8로 감소
  - 의미: 단순회귀에서 교육연수 계수에 경력연수와 성별의 효과가 섞여 있었음
  - 변수 추가할수록 → 교육연수의 **"순수한" 연관**만 남음

---

#### 계수 변화 예측 투표

> **투표 질문**: "단순회귀에서 교육연수 β₁ = 54.1이다. 경력연수를 추가하면 교육연수의 β₁은 어떻게 될까?"
>
> - ① 그대로 54.1  ② 줄어든다  ③ 늘어난다
>
> *(10초)*

- **정답**: β₁ = 38.4로 **② 줄어든다**
- **이유**: 교육연수와 경력연수 사이에 **양의 상관** 존재
  - 교육 연수가 긴 사람 = 경력도 함께 쌓이는 경향
  - ∴ 단순회귀의 교육연수 계수에 = 경력의 효과 일부 포함

---

### 핵심 개념 2: 통제(control)의 의미와 한계

- **통제(control)** 정의
  - "다른 변수를 일정하게 유지한 상태에서"라는 뜻
  - 예: "경력연수를 통제한 상태에서 교육연수의 효과" = **통계적 조정(statistical adjustment)**
  - ⚠️ 실제로 경력연수를 고정한 실험을 한 것이 **아님**

- **왜 중요한가**: 많은 연구 보고서에서 "통제했다"를 마치 인과 증명처럼 사용

- **통제 ≠ 인과 해결인 이유** (2가지):

1. **측정되지 않은 교란변수(omitted variable bias)**
   - 모형에 포함하지 않은 변수(예: 부모의 경제적 배경, 거주 지역, 직종)가 여전히 영향
   - 아무리 많은 변수 통제해도 → 빠진 변수 있으면 → 계수 편향 가능

2. **통계적 조정 ≠ 실험적 통제**
   - 실험: 무작위 배정으로 다른 조건 동일하게
   - 관찰 데이터: 통계적으로 "조정"할 뿐 (측정된 변수에 한해서만 가능)

| 표현 | 의미 | 인과 증명? |
|---|---|---|
| "경력을 통제한 상태에서 교육의 효과" | 경력이 같은 사람끼리 비교했을 때 교육의 연관 | 아니다 |
| "경력을 통제했으므로 교육이 원인이다" | 잘못된 해석 | — |

### 핵심 개념 3: 다중공선성(multicollinearity)

- **다중공선성(multicollinearity)** 정의
  - 독립변수들끼리 **강하게 상관**되어 있는 상태
  - 예: "교육연수"와 "최종학력(고졸/대졸/대학원졸)"을 동시에 모형에 넣는 경우
    - 두 변수가 거의 같은 정보 담고 있음 → 계수 추정 불안정

- **진단 방법**: **VIF(Variance Inflation Factor, 분산팽창인자)**

  | VIF 값 | 해석 |
  |---:|---|
  | < 5 | 공선성 문제 없음 |
  | 5–10 | 주의 필요 |
  | > 10 | 심각한 공선성 문제 → 변수 제거 또는 결합 고려 |

- **오늘 데이터(social_survey_w08.csv) VIF 결과**:

| 변수 | VIF |
|---|---:|
| 교육연수 | 1.75 |
| 경력연수 | 1.70 |
| 성별_더미 | 1.08 |

- 모두 10 미만 → 다중공선성 문제 **없음**

### 예시: 소득 ~ 교육연수 + 경력연수 + 성별 중다회귀

- 아래는 social_survey_w08.csv(200명)를 이용하여 교육연수, 경력연수, 성별로 소득을 예측하는 중다회귀의 핵심 코드이다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("data/social_survey_w08.csv", encoding='utf-8')
df["성별_더미"] = (df["성별"] == "남성").astype(int)
X = sm.add_constant(df[["교육연수", "경력연수", "성별_더미"]])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
for i, col in enumerate(X.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X.values, i+1):.2f}")
```

- _전체 코드는 practice/W08/code/W08-1-multiple-regression.py 참고_

**검정 결과:**

- **교육연수 β₁ = 36.8, p < 0.001**
  - 해석: 경력연수와 성별 통제 상태에서, 교육연수 1년 증가 시 소득 평균 약 36.8만 원 높음
- **경력연수 β₂ = 12.0, p < 0.001**
  - 해석: 교육연수와 성별 통제 상태에서, 경력연수 1년 증가 시 소득 평균 약 12.0만 원 높음
- **성별 β₃ = 41.8, p < 0.001**
  - 해석: 교육연수와 경력연수 통제 상태에서, 남성 소득이 여성보다 평균 약 41.8만 원 높음
- **R² = 0.625**: 이 모형이 소득 변동의 62.5%를 설명

---

**결과 해석:**

- **(분석목적)** 교육연수, 경력연수, 성별이 소득을 얼마나 예측하는지, 각 변수의 "순수한" 연관 확인을 위해 중다회귀 분석 실행
- **(핵심근거)**
  - 교육연수 β₁ = 36.8 (p < 0.001)
  - 경력연수 β₂ = 12.0 (p < 0.001)
  - 성별 β₃ = 41.8 (p < 0.001)
  - R² = 0.625
  - 단순회귀 대비 교육연수 계수: 54.1 → 36.8로 감소
- **(결론)** 유의수준 0.05에서 세 변수 모두 소득의 통계적으로 유의한 예측 변수, 모형이 소득 변동의 약 62.5% 설명
- **(주의)**
  - 이 결과 ≠ 교육이 소득을 "결정" 또는 성별이 소득 차이의 "원인"
  - 부모의 경제적 배경, 직종, 거주 지역 등 **측정되지 않은 교란변수** 존재 가능
