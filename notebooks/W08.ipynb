# W08A. 여러 변수를 동시에 고려하면 해석이 어떻게 달라지는가? — 중다회귀

> 여러 변수를 동시에 고려하면 해석이 어떻게 달라지는가?

**학습 목표**
1. **중다회귀(multiple regression)**의 모형 구조(Y = β₀ + β₁X₁ + β₂X₂ + ... + ε)를 이해하고, 각 계수를 "다른 변수를 통제한 상태에서"라는 조건과 함께 해석할 수 있다.
2. **통제(control)**의 의미가 "통계적 조정"일 뿐 "인과 해결"이 아님을 설명하고, 관찰 데이터에서 통제의 한계를 서술할 수 있다.
3. **다중공선성(multicollinearity)**을 VIF로 진단하고, 단순회귀 대비 중다회귀에서 계수가 변하는 이유를 설명할 수 있다.

## 핵심 개념 학습

### 왜 필요한가

7주차 단순회귀에서 "교육 연수가 1년 증가하면 소득이 평균 45.8만 원 높다"는 결과를 얻었다. 그런데 이 45.8에는 경력 연수의 효과가 섞여 있을 수 있다. 교육 연수가 긴 사람은 경력도 함께 쌓이는 경향이 있기 때문이다. 이처럼 하나의 독립변수만 넣으면, 그 변수에 다른 변수의 효과까지 합산되어 계수가 **과대 추정**될 수 있다. 여러 독립변수를 동시에 모형에 넣으면, 각 변수의 "순수한" 연관(다른 변수를 고정한 상태에서의 연관)을 분리할 수 있다. 이것이 **중다회귀(multiple regression)**가 필요한 이유이다.

---

### 핵심 개념 1: 중다회귀(multiple regression)

**중다회귀(multiple regression)**란 둘 이상의 독립변수(X₁, X₂, ...)로 하나의 종속변수(Y)를 예측하는 선형 모형이다. 모형의 구조는 다음과 같다.

> Y = β₀ + β₁X₁ + β₂X₂ + ... + ε

각 계수의 의미는 단순회귀와 미묘하지만 결정적으로 다르다.

- **β₁**: "**다른 모든 독립변수를 일정하게 유지한 상태에서** X₁이 1단위 증가할 때 Y의 평균 변화량"
- 이를 **편회귀계수(partial regression coefficient)**라고 한다.

쉽게 말해서, 단순회귀의 β₁은 "교육 연수만 보았을 때"의 효과이고, 중다회귀의 β₁은 "경력 연수를 같은 수준으로 맞춘 뒤 교육 연수만 비교했을 때"의 효과이다.

**표 W08.1** 단순회귀 vs 중다회귀 계수 비교 (동일 데이터, social_survey_w08.csv)

| 모형 | 교육연수 β₁ | 경력연수 β₂ | 성별 β₃ | R² |
|---|---:|---:|---:|---:|
| 단순회귀 (교육연수만) | 54.1 | — | — | 0.550 |
| 중다회귀 (교육연수 + 경력연수) | 38.4 | 12.3 | — | 0.617 |
| 중다회귀 (교육연수 + 경력연수 + 성별) | 36.8 | 12.0 | 41.8 | 0.625 |

교육연수의 β₁이 54.1 → 38.4 → 36.8로 줄어든다. 이는 단순회귀에서 교육연수의 계수에 경력연수와 성별의 효과가 섞여 있었다는 뜻이다. 변수를 추가할수록 교육연수의 "순수한" 연관만 남게 된다.

### 핵심 개념 2: 통제(control)의 의미와 한계

**통제(control)**란 "다른 변수를 일정하게 유지한 상태에서"라는 뜻이다. 중다회귀에서 "경력연수를 통제한 상태에서 교육연수의 효과"라고 말할 때, 이는 **통계적 조정(statistical adjustment)**을 의미한다. 실제로 경력연수를 고정한 실험을 한 것이 아니다.

이 구분이 왜 중요한가? 많은 연구 보고서에서 "통제했다"는 표현을 마치 인과를 증명한 것처럼 사용하기 때문이다.

**통제 ≠ 인과 해결**인 이유:

1. **측정되지 않은 교란변수(omitted variable bias)**: 모형에 포함하지 않은 변수(예: 부모의 경제적 배경, 거주 지역, 직종)가 여전히 결과에 영향을 줄 수 있다. 아무리 많은 변수를 통제해도, 빠진 변수가 있다면 계수는 편향될 수 있다.
2. **통계적 조정 ≠ 실험적 통제**: 실험에서는 무작위 배정으로 다른 조건을 동일하게 만들지만, 관찰 데이터에서는 통계적으로 "조정"할 뿐이다. 이 조정은 측정된 변수에 한해서만 가능하다.

| 표현 | 의미 | 인과 증명? |
|---|---|---|
| "경력을 통제한 상태에서 교육의 효과" | 경력이 같은 사람끼리 비교했을 때 교육의 연관 | 아니다 |
| "경력을 통제했으므로 교육이 원인이다" | 잘못된 해석 | — |

### 핵심 개념 3: 다중공선성(multicollinearity)

**다중공선성(multicollinearity)**이란 독립변수들끼리 강하게 상관되어 있는 상태를 말한다. 예를 들어 "교육연수"와 "최종학력(고졸/대졸/대학원졸)"을 동시에 모형에 넣으면, 두 변수가 거의 같은 정보를 담고 있어 계수 추정이 불안정해진다.

**VIF(Variance Inflation Factor, 분산팽창인자)**로 진단한다.

- **VIF < 5**: 공선성 문제 없음
- **VIF 5–10**: 주의 필요
- **VIF > 10**: 심각한 공선성 문제, 변수 제거 또는 결합 고려

오늘 데이터(social_survey_w08.csv)의 VIF를 확인하면 다음과 같다.

| 변수 | VIF |
|---|---:|
| 교육연수 | 1.75 |
| 경력연수 | 1.70 |
| 성별_더미 | 1.08 |

모두 10 미만이므로 다중공선성 문제는 없다.

### 예시: 소득 ~ 교육연수 + 경력연수 + 성별 중다회귀

아래는 social_survey_w08.csv(200명)를 이용하여 교육연수, 경력연수, 성별로 소득을 예측하는 중다회귀의 핵심 코드이다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("data/social_survey_w08.csv", encoding='utf-8')
df["성별_더미"] = (df["성별"] == "남성").astype(int)
X = sm.add_constant(df[["교육연수", "경력연수", "성별_더미"]])
model = sm.OLS(df["소득"], X).fit()
print(model.summary())
for i, col in enumerate(X.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X.values, i+1):.2f}")
```

_전체 코드는 practice/W08/code/W08-1-multiple-regression.py 참고_

**검정 결과:**

- **교육연수 β₁ = 36.8, p < 0.001**: 경력연수와 성별을 통제한 상태에서, 교육연수가 1년 증가할 때 소득은 평균 약 36.8만 원 높은 것으로 추정된다.
- **경력연수 β₂ = 12.0, p < 0.01**: 교육연수와 성별을 통제한 상태에서, 경력연수가 1년 증가할 때 소득은 평균 약 12.0만 원 높은 것으로 추정된다.
- **성별 β₃ = 41.8, p < 0.05**: 교육연수와 경력연수를 통제한 상태에서, 남성의 소득이 여성보다 평균 약 41.8만 원 높은 것으로 추정된다.
- **R² = 0.625**: 이 모형은 소득 변동의 62.5%를 설명한다.

---

**결과 해석:**

**(분석목적)** 교육연수, 경력연수, 성별이 소득을 얼마나 예측하는지, 각 변수의 "순수한" 연관을 확인하기 위해 중다회귀 분석을 실행했다. **(핵심근거)** 교육연수 β₁ = 36.8(p < 0.001), 경력연수 β₂ = 12.0(p < 0.01), 성별 β₃ = 41.8(p < 0.05)이며, R² = 0.625이다. 단순회귀 대비 교육연수의 계수가 54.1에서 36.8로 감소하였다. **(결론)** 유의수준 0.05에서 세 변수 모두 소득의 통계적으로 유의한 예측 변수이며, 모형은 소득 변동의 약 62.5%를 설명한다. **(주의)** 이 결과는 교육이 소득을 "결정"한다거나 성별이 소득 차이의 "원인"이라는 의미가 아니다. 부모의 경제적 배경, 직종, 거주 지역 등 측정되지 않은 교란변수가 여전히 존재할 수 있다.

## 실습 과제 (수업 내 완결형)

### 과제 개요

오늘 실습에서는 동일 데이터(social_survey_w08.csv)로 단순회귀와 중다회귀를 모두 실행하고, 계수 변화를 관찰하며, VIF를 확인한 뒤 해석문을 작성한다.

---

### 기본(Basic) 트랙

#### ① 실행

단순회귀(교육연수만)와 중다회귀(교육연수 + 경력연수)를 실행한다.

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("data/social_survey_w08.csv", encoding='utf-8')

# 1. 단순회귀: 소득 ~ 교육연수
X1 = sm.add_constant(df["교육연수"])
model1 = sm.OLS(df["소득"], X1).fit()
print("=== 단순회귀 ===")
print(model1.summary())

# 2. 중다회귀: 소득 ~ 교육연수 + 경력연수
X2 = sm.add_constant(df[["교육연수", "경력연수"]])
model2 = sm.OLS(df["소득"], X2).fit()
print("\n=== 중다회귀 ===")
print(model2.summary())
```

**완료 체크**: 두 모형의 summary()가 모두 출력되었는가? 교육연수의 β₁이 단순회귀와 중다회귀에서 다른 값인가?

#### ② 확인

VIF를 계산하고, 계수 변화를 기록한다.

**2-1. 계수 비교표 작성**: 두 모형의 핵심 수치를 기록한다.

| 항목 | 단순회귀 | 중다회귀 |
|---|---|---|
| 교육연수 β₁ | ___ | ___ |
| 경력연수 β₂ | — | ___ |
| 교육연수 p값 | ___ | ___ |
| R² | ___ | ___ |

**2-2. VIF 확인**:

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X2 = sm.add_constant(df[["교육연수", "경력연수"]])
for i, col in enumerate(X2.columns[1:]):
    print(f"{col}: VIF = {variance_inflation_factor(X2.values, i+1):.2f}")
```

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 해석문에 해당하는 항목이 없는지 점검해 보자.

**오류 1: "통제했다 = 인과를 증명했다"**

가장 흔하고 가장 심각한 오류이다. "경력을 통제했으므로 교육이 소득의 원인이다"라고 쓰는 학생이 많다. 관찰 데이터에서의 통제는 통계적 조정일 뿐, 측정되지 않은 교란변수가 존재할 수 있다.

| 잘못된 표현 (X) | 올바른 표현 (O) |
|---|---|
| "경력을 통제했으므로 교육이 원인이다" | "경력을 통계적으로 조정한 상태에서 교육과 소득의 양의 연관이 관찰된다" |
| "변수를 통제하면 인과를 알 수 있다" | "통제는 통계적 조정이며, 인과 증명에는 실험 설계가 필요하다" |

**오류 2: 단순회귀와 중다회귀 계수를 비교하지 않음**

중다회귀만 실행하고, 단순회귀 대비 계수가 어떻게 변했는지 확인하지 않는 경우가 많다. 계수 변화는 "변수를 추가하면 해석이 어떻게 달라지는가?"를 보여주는 핵심이다.

**확인법**: 반드시 단순회귀와 중다회귀를 모두 실행하고, 교육연수의 β₁ 변화를 기록한다.

**오류 3: VIF 확인을 건너뜀**

계수와 R²만 확인하고 VIF를 계산하지 않는 경우가 많다. VIF가 높으면 계수 추정이 불안정하여 해석이 신뢰할 수 없게 된다.

**확인법**: `variance_inflation_factor`를 사용하여 각 독립변수의 VIF를 반드시 확인하고, 10 미만인지 점검한다.

### 해석문 모범 구조

오늘 과제의 (D)항 해석문은 다음 4단계 구조를 따른다.

> **(분석목적)** 경력연수를 통제했을 때 교육연수와 소득의 연관이 어떻게 달라지는지 확인하기 위해, 단순회귀와 중다회귀를 비교 실행했다.
>
> **(핵심근거)** 단순회귀에서 교육연수 β₁ = 54.1이었으나, 경력연수를 추가한 중다회귀에서 β₁ = 38.4로 감소했다. 경력연수 β₂ = 12.3(p < 0.01)이며, R²는 0.550에서 0.617로 증가했다. VIF는 두 변수 모두 2 미만으로 다중공선성 문제는 없다.
>
> **(결론)** 유의수준 0.05에서 교육연수와 경력연수는 모두 소득의 유의한 예측 변수이다. 경력연수를 함께 고려하면 교육연수의 계수가 줄어들며, 이는 단순회귀에서 교육연수의 계수에 경력연수의 효과가 일부 섞여 있었음을 의미한다.
>
> **(주의)** 이 결과는 교육이나 경력이 소득을 "결정"한다는 인과적 의미가 아니다. 부모의 경제적 배경, 직종, 거주 지역 등 측정되지 않은 교란변수가 여전히 존재할 수 있으며, 관찰 데이터에서 "통제"는 통계적 조정에 불과하다.

3. **측정되지 않은 교란변수 미언급**: 보고서에 모형의 한계가 전혀 서술되지 않았다.
