# W04A. 두 범주형 변수가 관련 있는가? — 분할표와 카이자승 검정

> 두 범주형 변수 사이의 관계가 우연인지(독립인지), 아니면 통계적으로 유의한 연관인지 어떻게 판단할까?

**학습 목표**
1. **분할표**(교차표)를 작성하고, 관찰빈도와 기대빈도를 비교하여 두 범주형 변수의 관계를 탐색할 수 있다.
2. **카이자승 검정**의 논리(기대빈도와 관찰빈도의 차이)를 설명하고, 가정(기대빈도 ≥ 5)을 확인할 수 있다.
3. 검정 결과를 **효과 크기**(Cramér's V)와 함께 해석하되, 연관과 인과를 구별하는 해석문을 작성할 수 있다.

## 오늘의 질문/목표/진단

### 오늘의 질문

지난 3주차에서는 데이터의 전체 모습을 파악하는 기술통계와 해석문 구조를 배웠다. 오늘은 한 걸음 더 나아가, **두 범주형 변수 사이에 관계가 있는지**를 묻는다. 예를 들어, "남성과 여성은 지지하는 정당의 분포가 다른가?"와 같은 질문이다. 이 질문에 답하기 위해서는 눈으로 보이는 패턴이 **우연인지, 아니면 통계적으로 의미 있는 차이인지**를 구별하는 도구가 필요하다.

### 빠른 진단 (1문항)

## 핵심 개념 학습

### 왜 필요한가

3주차에서 우리는 변수 하나의 분포를 요약하는 방법(평균, 중앙값, 표준편차, 히스토그램 등)을 배웠다. 그러나 사회과학에서 관심 있는 질문은 대부분 **변수와 변수 사이의 관계**이다. "성별에 따라 정치 성향이 다른가?", "연령대에 따라 SNS 이용 여부가 다른가?"처럼, 두 범주형 변수가 서로 관련이 있는지를 묻는 것이다.

이 질문에 답하기 위한 도구가 바로 **분할표**와 **카이자승 검정**이다. 먼저 두 변수의 빈도를 표로 정리하고, 그 패턴이 우연인지 판단하는 검정을 실행하며, 마지막으로 관계의 강도를 효과 크기로 확인하는 흐름을 배운다.

### 핵심 개념 1: 분할표(contingency table/crosstab)

**분할표(contingency table)**, 또는 **교차표(crosstab)**란 두 범주형 변수의 빈도를 행(row)과 열(column)으로 교차 정리한 표이다. 각 칸(셀)에는 해당 조합에 속하는 관찰 대상의 수, 즉 **관찰빈도(observed frequency)**가 들어간다.

분할표에서 핵심적인 개념은 **기대빈도(expected frequency)**이다. 쉽게 말해서, "만약 성별과 지지정당이 **완전히 무관**하다면, 각 칸의 빈도는 얼마일까?"를 계산한 것이 기대빈도이다. 기대빈도는 다음과 같이 구한다.

> 기대빈도 = (해당 행의 합계 × 해당 열의 합계) ÷ 전체 합계

관찰빈도와 기대빈도를 비교하면, 두 변수 사이에 관계가 있는지 탐색할 수 있다. 관찰빈도가 기대빈도보다 눈에 띄게 크거나 작은 칸이 있다면, 두 변수가 독립이 아닐 수 있다는 단서가 된다.

#### 예측 투표 — 기대빈도 직관 체험

아래 분할표를 보자. 성별(남성/여성)과 지지정당(A당/B당/C당/무당파) 200명의 응답 결과이다.

**표 W04.1** 성별 × 지지정당 분할표 (관찰빈도)

| | A당 | B당 | C당 | 무당파 | 합계 |
|---|---:|---:|---:|---:|---:|
| 남성 | 45 | 16 | 19 | 24 | 104 |
| 여성 | 18 | 27 | 29 | 22 | 96 |
| 합계 | 63 | 43 | 48 | 46 | 200 |

> **투표 질문**: "성별과 지지정당이 완전히 무관하다면, **남성 + A당** 칸의 빈도(기대빈도)는 대략 얼마일까?"
>
> - 손으로 숫자를 써보자. (10초)

정답을 계산해 보면, 기대빈도 = (104 × 63) ÷ 200 = **32.76**이다. 그런데 실제 관찰빈도는 45이다. 기대빈도(약 33명)보다 12명이나 더 많다. 이것이 단순한 우연일까, 아니면 성별과 지지정당 사이에 실제로 관계가 있기 때문일까? 이 질문에 답하는 것이 카이자승 검정이다.

### 핵심 개념 2: 카이자승 검정(Chi-square test)

**카이자승 검정(Chi-square test)**은 관찰빈도와 기대빈도의 차이가 **우연(표본 변동)으로 설명될 수 있는지**를 검정하는 방법이다.

검정통계량 **χ²(카이자승)**는 모든 칸에서 (관찰빈도 − 기대빈도)²를 기대빈도로 나눈 값을 모두 더한 것이다. 쉽게 말해서, "관찰과 기대의 차이가 클수록 χ²값이 커진다." χ²값이 충분히 크면, "이 정도 차이가 우연히 나올 가능성은 매우 낮다"고 판단하여 **독립이라는 귀무가설을 기각**한다.

**카이자승 검정의 가정 두 가지:**

1. **독립 표본**: 각 관찰(응답자)은 서로 독립적이어야 한다. 동일한 사람이 두 번 응답하거나, 한 응답이 다른 응답에 영향을 주면 안 된다.
2. **기대빈도 ≥ 5**: 모든 칸의 기대빈도가 5 이상이어야 한다. 기대빈도가 5 미만인 칸이 있으면 카이자승 근사(approximation)가 부정확해진다. 이 경우에는 **Fisher 정확 검정(Fisher's exact test)**을 대안으로 사용한다.

### 핵심 개념 3: 독립/연관과 효과 크기

**"독립(independence)"**이란 한 변수의 분포가 다른 변수의 범주와 무관하다는 뜻이다. 예를 들어, 남성이든 여성이든 지지정당의 비율이 동일하면 성별과 지지정당은 독립이다.

카이자승 검정에서 귀무가설을 기각하면, 두 변수 사이에 **통계적 연관(association)**이 있다고 판단한다. 그런데 여기서 반드시 주의할 점이 있다.

> **연관 ≠ 인과(causation)**

"성별과 지지정당 사이에 연관이 있다"는 것은 "성별에 따라 지지정당의 분포가 다르다"는 사실을 말할 뿐이다. "성별이 정당 선택을 **결정한다**" 또는 "성별 **때문에** 정치 성향이 달라진다"고 말할 수는 없다. 연령, 교육수준, 지역 등 다양한 **교란변수(confounding variable)**가 성별과 정치 성향 모두에 관련되어 있을 수 있기 때문이다.

또한, 카이자승 검정의 p값만으로는 관계의 **강도**를 알 수 없다. 표본 크기가 크면 아주 약한 연관도 통계적으로 유의하게 나올 수 있다. 따라서 효과 크기 지표인 **Cramér's V**를 반드시 함께 보고해야 한다.

**Cramér's V**는 0(완전 독립)에서 1(완전 연관) 사이의 값을 가진다. 일반적인 해석 기준은 다음과 같다.

| Cramér's V | 연관의 강도 |
|---:|---|
| < 0.1 | 약한 연관 |
| 0.1 – 0.3 | 중간 수준의 연관 |
| > 0.3 | 강한 연관 |

### 예시: 성별 × 지지정당

위의 분할표(표 W04.1)를 이용하여 카이자승 검정을 실행한다. 아래는 핵심 코드이다.

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

ct = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct)
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"χ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")
print(f"Cramér's V = {V:.3f}")
```

_전체 코드는 practice/W04/code/W04-1-chi-square-intro.py 참고_

**검정 결과:**

- **χ²(3) = 16.262, p = 0.0010**
- **Cramér's V = 0.285**
- 기대빈도: 모두 5 이상 (최솟값 20.64) → 가정 충족

**표 W04.2** 성별 × 지지정당 기대빈도

| | A당 | B당 | C당 | 무당파 |
|---|---:|---:|---:|---:|
| 남성 | 32.76 | 22.36 | 24.96 | 23.92 |
| 여성 | 30.24 | 20.64 | 23.04 | 22.08 |

기대빈도(표 W04.2)와 관찰빈도(표 W04.1)를 비교하면, 남성은 A당에서 관찰빈도(45)가 기대빈도(32.76)보다 크고, 여성은 B당(관찰 27 vs 기대 20.64)과 C당(관찰 29 vs 기대 23.04)에서 관찰빈도가 기대빈도보다 크다. 즉, 남성은 A당(43.3%)을 더 지지하는 경향이 있고, 여성은 B당(28.1%)과 C당(30.2%)을 더 지지하는 경향이 관찰된다.

> **짝 토론**: 표 W04.1과 표 W04.2를 비교하면서, "관찰빈도와 기대빈도의 차이가 **가장 눈에 띄는 칸**은 어디인가?"를 옆 사람과 30초간 이야기해 보자.
>
> *(남성+A당 칸에서 45 vs 32.76으로 차이가 약 12로 가장 크다는 점을 확인)*

**결과 해석:**

유의수준 0.05에서, 성별과 지지정당은 독립이 아닌 것으로 나타났다(χ²(3) = 16.262, p = 0.001). Cramér's V = 0.285로, 중간 수준의 연관이 관찰된다. 남성은 A당을, 여성은 B당과 C당을 상대적으로 더 지지하는 경향이 있다.

**주의**: 이 결과는 성별과 지지정당 사이에 통계적 연관이 있다는 것을 보여줄 뿐이다. "성별이 정당 선택을 결정한다"거나 "성별 때문에 정치 성향이 달라진다"고 말할 수 **없다**. 연령, 교육수준, 거주지역 등 다양한 교란변수를 고려하지 않았기 때문이다.

```python
from scipy.stats import chi2_contingency
import numpy as np

# 1. 분할표 생성
ct = pd.crosstab(df["성별"], df["지지정당"])

print(ct)

# 2. 카이자승 검정 실행
chi2, p, dof, expected = chi2_contingency(ct)
print(f"\nχ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")

# 3. 기대빈도 출력
print("\n=== 기대빈도 ===")
print(pd.DataFrame(expected, index=ct.index, columns=ct.columns).round(2))
```

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 해석문에 해당하는 항목이 없는지 점검해 보자.

**오류 1: 기대빈도 확인을 건너뜀**

카이자승 검정을 실행한 뒤 p값만 확인하고, 기대빈도를 출력하지 않는 경우가 많다. `chi2_contingency(ct)`는 4개의 값을 반환하며, 네 번째 값이 기대빈도 배열(`expected`)이다. 반드시 다음을 실행하여 확인한다.

```python
chi2, p, dof, expected = chi2_contingency(ct)
print(expected)  # 모든 값이 5 이상인지 눈으로 확인
```

기대빈도가 5 미만인 칸이 있으면, Fisher 정확 검정(`scipy.stats.fisher_exact`, 2×2 표의 경우)을 사용하거나 빈도가 적은 범주를 통합하여 재분석해야 한다.

**오류 2: "유의하다 = 효과가 크다"로 해석**

p값이 작다고 해서 효과가 크다는 뜻이 아니다. 표본이 크면 매우 약한 연관도 통계적으로 유의하게 나올 수 있다. 반드시 Cramér's V를 함께 보고하고, 아래 기준으로 실질적 의미를 판단한다.

- V < 0.1: 약한 연관 (통계적으로 유의하더라도 실질적 의미가 작다)
- 0.1 ≤ V ≤ 0.3: 중간 수준의 연관
- V > 0.3: 강한 연관

**오류 3: "카이자승 결과 = 인과"로 해석**

가장 흔하고 가장 심각한 오류이다. 카이자승 검정은 두 변수 간 **통계적 연관의 유무**만 판단한다. 인과의 방향이나 메커니즘에 대해서는 아무것도 말하지 않는다.

| 잘못된 표현 (X) | 올바른 표현 (O) |
|---|---|
| "성별이 정당 선택에 영향을 미친다" | "성별과 지지정당 사이에 통계적으로 유의한 연관이 관찰된다" |
| "성별 때문에 정치 성향이 달라진다" | "성별에 따라 지지정당의 분포에 차이가 나타난다" |
| "남성이라서 A당을 지지한다" | "남성에서 A당 지지 비율이 상대적으로 높게 나타난다" |

````python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv("data/social_survey_w04.csv", encoding="utf-8")
print(f"데이터 크기: {df.shape}")
df.head()
````

````python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

ct = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct)
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"χ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")
print(f"Cramér's V = {V:.3f}")
````

**Q.** "카이자승 검정에서 p값이 0.03이고 Cramer's V가 0.08이면, 이 결과를 어떻게 해석해야 하는가?"

**정답**: 통계적으로 유의(p = 0.03 < 0.05)하지만, Cramer's V = 0.08으로 효과 크기가 매우 약하다. 유의수준을 통과했더라도 실질적 의미는 크지 않다. 이는 표본 크기가 충분히 크면 아주 약한 연관도 유의하게 나올 수 있음을 보여주는 사례이다.

## 과제 1 기대 결과: 성별 x 지지정당

### 기대 코드

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

df = pd.read_csv("social_survey_w04.csv")

# 분할표 생성
ct = pd.crosstab(df["성별"], df["지지정당"], margins=True)
print("=== 분할표 (관찰빈도) ===")
print(ct)

# 카이자승 검정
ct_core = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct_core)
print(f"\nχ²({dof}) = {chi2:.3f}, p = {p:.4f}")

# 기대빈도 확인
print("\n=== 기대빈도 ===")
print(pd.DataFrame(expected, index=ct_core.index,
                    columns=ct_core.columns).round(2))
print(f"최소 기대빈도: {expected.min():.2f}")

# Cramér's V
n = ct_core.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct_core.shape) - 1)))
print(f"Cramér's V = {V:.3f}")
```

**분할표 (관찰빈도)**:

| | A당 | B당 | C당 | 무당파 | 합계 |
|---|---:|---:|---:|---:|---:|
| 남성 | 45 | 16 | 19 | 24 | 104 |
| 여성 | 18 | 27 | 29 | 22 | 96 |
| 합계 | 63 | 43 | 48 | 46 | 200 |

**기대빈도**:

| | A당 | B당 | C당 | 무당파 |
|---|---:|---:|---:|---:|
| 남성 | 32.76 | 22.36 | 24.96 | 23.92 |
| 여성 | 30.24 | 20.64 | 23.04 | 22.08 |

**검정 결과**:

|---|---|
| chi-sq(df) | chi-sq(3) = 16.262 |

## 모범 보고서 (A-E)

아래는 과제 1(성별 x 지지정당)을 선택한 경우의 모범 보고서이다.

### (A) 분석 질문

성별과 지지정당 사이에 통계적으로 유의한 연관이 있는가?

### (B) 사용 변수/척도

- 성별: 명목척도 (남성, 여성)
- 지지정당: 명목척도 (A당, B당, C당, 무당파)

### (C) 결과 근거

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

df = pd.read_csv("social_survey_w04.csv")
ct = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct)
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"χ²({dof}) = {chi2:.3f}, p = {p:.4f}")
print(f"Cramér's V = {V:.3f}")
print(pd.DataFrame(expected, index=ct.index, columns=ct.columns).round(2))
```

- 분할표: 남성 A당 45 / B당 16 / C당 19 / 무당파 24, 여성 A당 18 / B당 27 / C당 29 / 무당파 22
- chi-sq(3) = 16.262, p = 0.0010
- 기대빈도 최솟값 20.64 (모든 셀 >= 5 충족)
- Cramer's V = 0.285

### (D) 해석문

**(분석목적)** 성별과 지지정당 사이에 통계적으로 유의한 연관이 있는지 확인하기 위해 카이자승 검정을 실시했다.

**(핵심근거)** 분할표에서 남성은 A당(43.3%), 여성은 B당(28.1%)과 C당(30.2%)을 상대적으로 더 지지하는 경향이 나타났다. chi-sq(3) = 16.262, p = 0.001이며, Cramer's V = 0.285이다. 기대빈도는 모든 셀에서 5 이상(최솟값 20.64)으로 검정 가정을 충족했다.

**(결론)** 유의수준 0.05에서 귀무가설(성별과 지지정당은 독립이다)을 기각한다. 두 변수 사이에 중간 수준의 연관이 관찰된다.

**(주의)** 이 결과는 성별이 정당 선택을 "결정"하거나 "원인"이라는 의미가 아니다. 연령, 교육수준, 거주지역 등 교란변수를 고려하지 않았으며, 설문 표본(n = 200)의 대표성에도 한계가 있다.

### (E) Copilot 사용 요약

Copilot에게 "성별과 지지정당의 분할표를 만들고 카이자승 검정을 실행해줘"라고 요청하여 기본 코드를 생성했다. Cramer's V 계산은 추가 프롬프트로 요청했다. 변수명 오타를 1회 수정했으며, Copilot이 생성한 해석문에서 "영향을 미친다"는 인과 표현을 "연관이 관찰된다"로 직접 수정했다.

## 흔한 오류 3개 + 확인법

### 오류 1: 기대빈도 확인 누락

`chi2_contingency()`는 4개의 값을 반환하며, 네 번째 값이 기대빈도 배열(`expected`)이다. 기대빈도를 출력하지 않고 p값만 보고하는 조가 많다.

**확인법**: `print(expected)`를 실행하여 모든 셀이 5 이상인지 눈으로 확인한다. 오늘 데이터에서는 과제 1 최솟값 20.64, 과제 2 최솟값 16.00으로 모두 충족한다.

### 오류 2: Cramer's V 미보고

p값만 보고하고 효과 크기를 보고하지 않는 경우가 흔하다. p값은 연관의 유무만 말해 줄 뿐 강도는 알려주지 않는다.

**확인법**: 결과 기록표에 Cramer's V가 비어 있다면 코드를 추가한다. 해석 기준: V < 0.1(약함), 0.1-0.3(중간), > 0.3(강함).

### 오류 3: 인과 표현 사용

"영향을 미친다", "결정한다", "~때문에" 등의 인과 표현이 해석문에 포함되는 경우가 가장 빈번하다.

### 비판적 읽기 자료

아래 가상 뉴스 기사에서 문제점을 확인한다.

> **<OO일보> 2026.01.15**
>
> *"전국 성인 1,200명을 대상으로 한 설문 결과, 남녀의 정당 지지율이 다른 것으로 나타나, **성별이 정치 성향을 결정하는 주요 요인으로 확인되었다**. 이번 조사에서 남성의 52%가 A당을, 여성의 48%가 B당을 지지하는 것으로 드러났다."*

**문제점 3가지**:

1. **인과 표현**: "성별이 정치 성향을 **결정**하는 주요 요인"이라고 서술했다. 설문 데이터(횡단면 조사)로는 연관만 확인할 수 있으며, 인과를 주장하려면 실험 설계나 교란변수 통제가 필요하다. "성별에 따라 정당 지지율에 차이가 **관찰되었다**"가 적절하다.

2. **효과 크기 미보고**: 지지율 차이의 통계적 유의성(p값)이나 효과 크기(Cramer's V)가 전혀 보고되지 않았다. 1,200명이라는 큰 표본에서는 사소한 차이도 유의하게 나올 수 있으므로, 효과 크기 없이는 실질적 의미를 판단할 수 없다.

3. **교란변수 무시**: 연령, 교육수준, 소득, 거주지역 등 성별과 정치 성향 모두에 관련될 수 있는 변수가 고려되지 않았다. 이러한 교란변수를 통제하지 않은 상태에서 성별을 "주요 요인"이라고 단정하는 것은 과잉 해석이다.
