## 핵심 개념 학습

### 왜 필요한가

3주차에서 우리는 변수 하나의 분포를 요약하는 방법(평균, 중앙값, 표준편차, 히스토그램 등)을 배웠다. 그러나 사회과학에서 관심 있는 질문은 대부분 **변수와 변수 사이의 관계**이다. "성별에 따라 정치 성향이 다른가?", "연령대에 따라 SNS 이용 여부가 다른가?"처럼, 두 범주형 변수가 서로 관련이 있는지를 묻는 것이다.

이 질문에 답하기 위한 도구가 바로 **분할표**와 **카이자승 검정**이다. 전체 흐름은 3단계로 이루어진다:

1. 먼저 두 변수의 빈도를 **표로 정리**하고 (분할표)
2. 그 패턴이 **우연인지 판단**하는 검정을 실행하며 (카이자승 검정)
3. 마지막으로 관계의 **강도**를 효과 크기로 확인한다 (Cramér's V)

---

### 핵심 개념 1: 분할표(contingency table/crosstab)

**분할표(contingency table)**, 또는 **교차표(crosstab)**란 두 범주형 변수의 빈도를 행(row)과 열(column)으로 교차 정리한 표이다. 각 칸(셀)에는 해당 조합에 속하는 관찰 대상의 수, 즉 **관찰빈도(observed frequency)**가 들어간다.

분할표에서 핵심적인 개념은 **기대빈도(expected frequency)**이다. 쉽게 말해서, "만약 성별과 지지정당이 **완전히 무관**하다면, 각 칸의 빈도는 얼마일까?"를 계산한 것이 기대빈도이다. 기대빈도는 다음과 같이 구한다.

> 기대빈도 = (해당 행의 합계 × 해당 열의 합계) ÷ 전체 합계

관찰빈도와 기대빈도를 비교하면, 두 변수 사이에 관계가 있는지 탐색할 수 있다. 관찰빈도가 기대빈도보다 눈에 띄게 크거나 작은 칸이 있다면, 두 변수가 독립이 아닐 수 있다는 단서가 된다.

---

#### 예측 투표 — 기대빈도

- **예시 데이터**: 성별(남성/여성)과 지지정당(A당/B당/C당/무당파) 200명 응답 결과

- **표 W04.1** 성별 × 지지정당 분할표 (관찰빈도)

| | A당 | B당 | C당 | 무당파 | 합계 |
|---|---:|---:|---:|---:|---:|
| 남성 | 45 | 16 | 19 | 24 | 104 |
| 여성 | 18 | 27 | 29 | 22 | 96 |
| 합계 | 63 | 43 | 48 | 46 | 200 |

> **투표 질문**: "성별과 지지정당이 완전히 무관하다면, **남성 + A당** 칸의 빈도(기대빈도)는 대략 얼마일까?"
>
> - 손으로 숫자를 써보자.

- **정답 계산**:
  - 기대빈도 = (104 × 63) ÷ 200 = **32.76**
  - 실제 관찰빈도 = 45
  - 차이 = 45 - 32.76 = **약 12명 더 많음**
- **핵심 질문**: 이 차이가 단순한 우연인가, 실제로 성별과 지지정당 사이에 관계가 있는 것인가?
  - → 이 질문에 답하는 것이 **카이자승 검정**

### 핵심 개념 2: 카이자승 검정(Chi-square test)

**카이자승 검정(Chi-square test)**은 관찰빈도와 기대빈도의 차이가 **우연(표본 변동)으로 설명될 수 있는지**를 검정하는 방법이다.

검정통계량 **χ²(카이자승)**는 모든 칸에서 (관찰빈도 − 기대빈도)²를 기대빈도로 나눈 값을 모두 더한 것이다. 쉽게 말해서, "관찰과 기대의 차이가 클수록 χ²값이 커진다." χ²값이 충분히 크면, "이 정도 차이가 우연히 나올 가능성은 매우 낮다"고 판단하여 **독립이라는 귀무가설을 기각**한다.

**카이자승 검정의 가정 두 가지:**

1. **독립 표본**: 각 관찰(응답자)은 서로 독립적이어야 한다. 동일한 사람이 두 번 응답하거나, 한 응답이 다른 응답에 영향을 주면 안 된다.
2. **기대빈도 ≥ 5**: 모든 칸의 기대빈도가 5 이상이어야 한다. 기대빈도가 5 미만인 칸이 있으면 카이자승 근사(approximation)가 부정확해진다. 이 경우에는 **Fisher 정확 검정(Fisher's exact test)**을 대안으로 사용하거나, 빈도가 적은 범주를 통합하여 재분석한다.

### 핵심 개념 3: 독립/연관과 효과 크기

- **"독립(independence)"** 정의
  - 한 변수의 분포가 다른 변수의 범주와 무관한 상태
  - 예: 남성이든 여성이든 지지정당의 비율이 동일 → 성별과 지지정당은 독립

- **통계적 연관(association)**
  - 카이자승 검정에서 귀무가설 기각 시 → 두 변수 사이에 통계적 연관 있다고 판단
  - **⚠️ 중요 주의점**:

> **연관 ≠ 인과(causation)**

- **연관과 인과의 구별**
  - **연관**: "A와 B가 함께 움직인다"는 **패턴의 관찰**. 방향이나 이유는 포함하지 않음
  - **인과**: "A가 B를 일으킨다"는 **메커니즘의 주장**. 실험 설계나 교란변수 통제가 필요
  - 카이자승 검정이 알려주는 것은 **연관(패턴)의 유무**뿐이며, 인과에 대해서는 아무것도 말하지 않음

- **왜 연관에서 인과를 주장할 수 없는가? — 교란변수(confounding variable)**
  - 교란변수란 두 변수 **모두에** 영향을 주어 가짜 연관을 만들어낼 수 있는 제3의 변수
  - 구체적 예시:
    - 데이터에서 "성별 → 지지정당"에 연관이 관찰되었다고 하자
    - 그런데 **연령**이 교란변수일 수 있음: 남성 응답자가 고연령층에 많고, 고연령층이 A당을 지지하는 경향이 있다면 → 실제로는 **연령 → 지지정당**인데 성별과 엮여 보이는 것
    - 이처럼 성별 자체가 정당 선택을 결정하는 것이 아니라, 연령·교육수준·지역 등이 뒤에서 작용할 수 있음

- **올바른 표현 vs 잘못된 표현**

| | 표현 | 판정 |
|---|---|:---:|
| ✅ | "성별에 따라 지지정당의 분포가 **다르게 관찰되었다**" | 연관 (OK) |
| ✅ | "성별과 지지정당 사이에 **통계적으로 유의한 연관**이 있다" | 연관 (OK) |
| ❌ | "성별이 정당 선택을 **결정한다**" | 인과 주장 |
| ❌ | "성별 **때문에** 정치 성향이 달라진다" | 인과 주장 |

- **효과 크기 — Cramér's V**
  - 카이자승 검정의 p값만으로는 관계의 **강도**를 알 수 없음
  - 표본 크기가 크면 → 아주 약한 연관도 통계적으로 유의하게 나옴
  - ∴ 효과 크기 지표 **Cramér's V**를 반드시 함께 보고
  - Cramér's V 범위: 0(완전 독립) ~ 1(완전 연관)

| Cramér's V | 연관의 강도 |
|---:|---|
| < 0.1 | 약한 연관 |
| 0.1 – 0.3 | 중간 수준의 연관 |
| > 0.3 | 강한 연관 |

### 예시: 성별 × 지지정당

위의 분할표(표 W04.1)를 이용하여 카이자승 검정을 실행한다. 코드 흐름은 다음과 같다:

1. `pd.crosstab()`으로 분할표(관찰빈도) 생성
2. `chi2_contingency()`로 검정통계량(χ²), p값, 자유도(df), 기대빈도(expected) 계산
3. Cramér's V 수동 계산

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv("data/social_survey_w04.csv", encoding="utf-8")
print(f"데이터 크기: {df.shape}")
df.head()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

ct = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct)
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"χ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")
print(f"Cramér's V = {V:.3f}")
```

_전체 코드는 practice/W04/code/W04-1-chi-square-intro.py 참고_

**검정 결과:**

- **χ²(3) = 16.26, p = 0.0010**
- **Cramér's V = 0.285**
- **기대빈도**: 모두 5 이상 (최솟값 20.64) → 가정 충족

**오류 1: 기대빈도 확인을 건너뜀**

**오류 2: "유의하다 = 효과가 크다"로 해석**

p값이 작다고 해서 효과가 크다는 뜻이 아니다. 표본이 크면 매우 약한 연관도 통계적으로 유의하게 나올 수 있다. 반드시 Cramér's V를 함께 보고하고, 아래 기준으로 실질적 의미를 판단한다.

- V < 0.1: 약한 연관 (통계적으로 유의하더라도 실질적 의미가 작다)
- 0.1 ≤ V ≤ 0.3: 중간 수준의 연관
- V > 0.3: 강한 연관

**오류 3: "카이자승 결과 = 인과"로 해석**

가장 흔하고 가장 심각한 오류이다. 카이자승 검정은 두 변수 간 **통계적 연관의 유무**만 판단한다. 인과의 방향이나 메커니즘에 대해서는 아무것도 말하지 않는다.

| 잘못된 표현 (✕) | 올바른 표현 (○) |
|---|---|
| "성별이 정당 선택에 영향을 미친다" | "성별과 지지정당 사이에 통계적으로 유의한 연관이 관찰된다" |
| "성별 때문에 정치 성향이 달라진다" | "성별에 따라 지지정당의 분포에 차이가 나타난다" |
| "남성이라서 A당을 지지한다" | "남성에서 A당 지지 비율이 상대적으로 높게 나타난다" |
