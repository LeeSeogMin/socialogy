> 두 범주형 변수 사이의 관계가 우연인지(독립인지), 아니면 통계적으로 유의한 연관인지 어떻게 판단할까?

2. **카이자승 검정**의 논리(기대빈도와 관찰빈도의 차이)를 설명하고, 가정(기대빈도 ≥ 5)을 확인할 수 있다.
3. 검정 결과를 **효과 크기**(Cramér's V)와 함께 해석하되, 연관과 인과를 구별하는 해석문을 작성할 수 있다.

## 핵심 개념 학습

  - 예: "성별에 따라 정치 성향이 다른가?"
  - 예: "연령대에 따라 SNS 이용 여부가 다른가?"

---

### 핵심 개념 1: 분할표(contingency table/crosstab)

- **분할표(contingency table)** 또는 **교차표(crosstab)** 정의
  - 두 범주형 변수의 빈도를 행(row)과 열(column)으로 교차 정리한 표
  - 각 칸(셀)에는 해당 조합에 속하는 관찰 대상의 수 = **관찰빈도(observed frequency)**

- **기대빈도(expected frequency)** — 분할표의 핵심 개념
  - "만약 성별과 지지정당이 **완전히 무관**하다면, 각 칸의 빈도는 얼마일까?"를 계산한 값
  - 계산 공식:

> 기대빈도 = (해당 행의 합계 × 해당 열의 합계) ÷ 전체 합계

- **관찰빈도와 기대빈도 비교**
  - 두 변수 사이의 관계 존재 여부를 탐색하는 방법
  - 관찰빈도가 기대빈도보다 눈에 띄게 크거나 작은 칸이 있다면 → 두 변수가 독립이 아닐 수 있다는 단서

---

#### 예측 투표 — 기대빈도 직관 체험

- **예시 데이터**: 성별(남성/여성)과 지지정당(A당/B당/C당/무당파) 200명 응답 결과

- **표 W04.1** 성별 × 지지정당 분할표 (관찰빈도)

| | A당 | B당 | C당 | 무당파 | 합계 |
|---|---:|---:|---:|---:|---:|
| 남성 | 45 | 16 | 19 | 24 | 104 |
| 여성 | 18 | 27 | 29 | 22 | 96 |
| 합계 | 63 | 43 | 48 | 46 | 200 |

> **투표 질문**: "성별과 지지정당이 완전히 무관하다면, **남성 + A당** 칸의 빈도(기대빈도)는 대략 얼마일까?"
>
> - 손으로 숫자를 써보자. (10초)

- **정답 계산**:
  - 기대빈도 = (104 × 63) ÷ 200 = **32.76**
  - 실제 관찰빈도 = 45
  - 차이 = 45 - 32.76 = **약 12명 더 많음**
- **핵심 질문**: 이 차이가 단순한 우연인가, 실제로 성별과 지지정당 사이에 관계가 있는 것인가?
  - → 이 질문에 답하는 것이 **카이자승 검정**

---

### 핵심 개념 2: 카이자승 검정(Chi-square test)

- **카이자승 검정(Chi-square test)** 정의
  - 관찰빈도와 기대빈도의 차이가 **우연(표본 변동)으로 설명될 수 있는지** 검정하는 방법

- **검정통계량 χ²(카이자승)**
  - 모든 칸에서 (관찰빈도 − 기대빈도)²를 기대빈도로 나눈 값의 합
  - 핵심: "관찰과 기대의 차이가 클수록 χ²값이 커진다"
  - χ²값이 충분히 크면 → "이 정도 차이가 우연히 나올 가능성은 매우 낮다" → **독립이라는 귀무가설 기각**

- **카이자승 검정의 가정 2가지**:
  1. **독립 표본**
     - 각 관찰(응답자)은 서로 독립적이어야 함
     - 동일 사람이 두 번 응답 ✕
     - 한 응답이 다른 응답에 영향 ✕
  2. **기대빈도 ≥ 5**
     - 모든 칸의 기대빈도가 5 이상이어야 함
     - 기대빈도 5 미만인 칸 존재 시 → 카이자승 근사(approximation) 부정확
     - 대안: **Fisher 정확 검정(Fisher's exact test)**

---

### 핵심 개념 3: 독립/연관과 효과 크기

- **"독립(independence)"** 정의
  - 한 변수의 분포가 다른 변수의 범주와 무관한 상태
  - 예: 남성이든 여성이든 지지정당의 비율이 동일 → 성별과 지지정당은 독립

- **통계적 연관(association)**
  - 카이자승 검정에서 귀무가설 기각 시 → 두 변수 사이에 통계적 연관 있다고 판단
  - **⚠️ 중요 주의점**:

> **연관 ≠ 인과(causation)**

- **연관과 인과의 구별**
  - "성별과 지지정당 사이에 연관이 있다" = "성별에 따라 지지정당의 분포가 다르다"는 **사실**
  - "성별이 정당 선택을 **결정한다**" ✕
  - "성별 **때문에** 정치 성향이 달라진다" ✕
  - 이유: 연령, 교육수준, 지역 등 **교란변수(confounding variable)**가 성별과 정치 성향 모두에 관련될 수 있음

- **효과 크기 — Cramér's V**
  - 카이자승 검정의 p값만으로는 관계의 **강도**를 알 수 없음
  - 표본 크기가 크면 → 아주 약한 연관도 통계적으로 유의하게 나옴
  - ∴ 효과 크기 지표 **Cramér's V**를 반드시 함께 보고
  - Cramér's V 범위: 0(완전 독립) ~ 1(완전 연관)

| Cramér's V | 연관의 강도 |
|---:|---|
| < 0.1 | 약한 연관 |
| 0.1 – 0.3 | 중간 수준의 연관 |
| > 0.3 | 강한 연관 |

### 예시: 성별 × 지지정당

- 위의 분할표(표 W04.1)를 이용하여 카이자승 검정을 실행한다. 아래는 핵심 코드이다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv("data/social_survey_w04.csv", encoding="utf-8")
print(f"데이터 크기: {df.shape}")
df.head()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

ct = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct)
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"χ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")
print(f"Cramér's V = {V:.3f}")
```

- _전체 코드는 practice/W04/code/W04-1-chi-square-intro.py 참고_

**검정 결과:**

- **χ²(3) = 16.26, p = 0.0010**
- **Cramér's V = 0.285**
- **기대빈도**: 모두 5 이상 (최솟값 20.64) → 가정 충족

- **표 W04.2** 성별 × 지지정당 기대빈도

| | A당 | B당 | C당 | 무당파 |
|---|---:|---:|---:|---:|
| 남성 | 32.76 | 22.36 | 24.96 | 23.92 |
| 여성 | 30.24 | 20.64 | 23.04 | 22.08 |

- **관찰빈도 vs 기대빈도 비교** (표 W04.1 vs 표 W04.2):
  - **남성 A당**: 관찰빈도(45) > 기대빈도(32.76)
  - **여성 B당**: 관찰빈도(27) > 기대빈도(20.64)
  - **여성 C당**: 관찰빈도(29) > 기대빈도(23.04)
- **패턴 해석**:
  - 남성: A당(43.3%) 더 지지하는 경향
  - 여성: B당(28.1%)과 C당(30.2%) 더 지지하는 경향
