# W04A. 두 범주형 변수가 관련 있는가? — 분할표와 카이자승 검정

> 두 범주형 변수 사이의 관계가 우연인지(독립인지), 아니면 통계적으로 유의한 연관인지 어떻게 판단할까?

**학습 목표**
1. **분할표**(교차표)를 작성하고, 관찰빈도와 기대빈도를 비교하여 두 범주형 변수의 관계를 탐색할 수 있다.
2. **카이자승 검정**의 논리(기대빈도와 관찰빈도의 차이)를 설명하고, 가정(기대빈도 ≥ 5)을 확인할 수 있다.
3. 검정 결과를 **효과 크기**(Cramér's V)와 함께 해석하되, 연관과 인과를 구별하는 해석문을 작성할 수 있다.

## 핵심 개념 학습

### 왜 필요한가

3주차에서 우리는 변수 하나의 분포를 요약하는 방법(평균, 중앙값, 표준편차, 히스토그램 등)을 배웠다. 그러나 사회과학에서 관심 있는 질문은 대부분 **변수와 변수 사이의 관계**이다. "성별에 따라 정치 성향이 다른가?", "연령대에 따라 SNS 이용 여부가 다른가?"처럼, 두 범주형 변수가 서로 관련이 있는지를 묻는 것이다.

이 질문에 답하기 위한 도구가 바로 **분할표**와 **카이자승 검정**이다. 먼저 두 변수의 빈도를 표로 정리하고, 그 패턴이 우연인지 판단하는 검정을 실행하며, 마지막으로 관계의 강도를 효과 크기로 확인하는 흐름을 배운다.

---

### 핵심 개념 1: 분할표(contingency table/crosstab)

**분할표(contingency table)**, 또는 **교차표(crosstab)**란 두 범주형 변수의 빈도를 행(row)과 열(column)으로 교차 정리한 표이다. 각 칸(셀)에는 해당 조합에 속하는 관찰 대상의 수, 즉 **관찰빈도(observed frequency)**가 들어간다.

분할표에서 핵심적인 개념은 **기대빈도(expected frequency)**이다. 쉽게 말해서, "만약 성별과 지지정당이 **완전히 무관**하다면, 각 칸의 빈도는 얼마일까?"를 계산한 것이 기대빈도이다. 기대빈도는 다음과 같이 구한다.

> 기대빈도 = (해당 행의 합계 × 해당 열의 합계) ÷ 전체 합계

관찰빈도와 기대빈도를 비교하면, 두 변수 사이에 관계가 있는지 탐색할 수 있다. 관찰빈도가 기대빈도보다 눈에 띄게 크거나 작은 칸이 있다면, 두 변수가 독립이 아닐 수 있다는 단서가 된다.

---

#### 예측 투표 — 기대빈도 직관 체험

아래 분할표를 보자. 성별(남성/여성)과 지지정당(A당/B당/C당/무당파) 200명의 응답 결과이다.

**표 W04.1** 성별 × 지지정당 분할표 (관찰빈도)

| | A당 | B당 | C당 | 무당파 | 합계 |
|---|---:|---:|---:|---:|---:|
| 남성 | 45 | 16 | 19 | 24 | 104 |
| 여성 | 18 | 27 | 29 | 22 | 96 |
| 합계 | 63 | 43 | 48 | 46 | 200 |

> **투표 질문**: "성별과 지지정당이 완전히 무관하다면, **남성 + A당** 칸의 빈도(기대빈도)는 대략 얼마일까?"
>
> - 손으로 숫자를 써보자. (10초)

정답을 계산해 보면, 기대빈도 = (104 × 63) ÷ 200 = **32.76**이다. 그런데 실제 관찰빈도는 45이다. 기대빈도(약 33명)보다 12명이나 더 많다. 이것이 단순한 우연일까, 아니면 성별과 지지정당 사이에 실제로 관계가 있기 때문일까? 이 질문에 답하는 것이 카이자승 검정이다.

---

### 핵심 개념 2: 카이자승 검정(Chi-square test)

**카이자승 검정(Chi-square test)**은 관찰빈도와 기대빈도의 차이가 **우연(표본 변동)으로 설명될 수 있는지**를 검정하는 방법이다.

검정통계량 **χ²(카이자승)**는 모든 칸에서 (관찰빈도 − 기대빈도)²를 기대빈도로 나눈 값을 모두 더한 것이다. 쉽게 말해서, "관찰과 기대의 차이가 클수록 χ²값이 커진다." χ²값이 충분히 크면, "이 정도 차이가 우연히 나올 가능성은 매우 낮다"고 판단하여 **독립이라는 귀무가설을 기각**한다.

**카이자승 검정의 가정 두 가지:**

1. **독립 표본**: 각 관찰(응답자)은 서로 독립적이어야 한다. 동일한 사람이 두 번 응답하거나, 한 응답이 다른 응답에 영향을 주면 안 된다.
2. **기대빈도 ≥ 5**: 모든 칸의 기대빈도가 5 이상이어야 한다. 기대빈도가 5 미만인 칸이 있으면 카이자승 근사(approximation)가 부정확해진다. 이 경우에는 **Fisher 정확 검정(Fisher's exact test)**을 대안으로 사용한다.

---

### 핵심 개념 3: 독립/연관과 효과 크기

**"독립(independence)"**이란 한 변수의 분포가 다른 변수의 범주와 무관하다는 뜻이다. 예를 들어, 남성이든 여성이든 지지정당의 비율이 동일하면 성별과 지지정당은 독립이다.

카이자승 검정에서 귀무가설을 기각하면, 두 변수 사이에 **통계적 연관(association)**이 있다고 판단한다. 그런데 여기서 반드시 주의할 점이 있다.

> **연관 ≠ 인과(causation)**

"성별과 지지정당 사이에 연관이 있다"는 것은 "성별에 따라 지지정당의 분포가 다르다"는 사실을 말할 뿐이다. "성별이 정당 선택을 **결정한다**" 또는 "성별 **때문에** 정치 성향이 달라진다"고 말할 수는 없다. 연령, 교육수준, 지역 등 다양한 **교란변수(confounding variable)**가 성별과 정치 성향 모두에 관련되어 있을 수 있기 때문이다.

또한, 카이자승 검정의 p값만으로는 관계의 **강도**를 알 수 없다. 표본 크기가 크면 아주 약한 연관도 통계적으로 유의하게 나올 수 있다. 따라서 효과 크기 지표인 **Cramér's V**를 반드시 함께 보고해야 한다.

**Cramér's V**는 0(완전 독립)에서 1(완전 연관) 사이의 값을 가진다. 일반적인 해석 기준은 다음과 같다.

| Cramér's V | 연관의 강도 |
|---:|---|
| < 0.1 | 약한 연관 |
| 0.1 – 0.3 | 중간 수준의 연관 |
| > 0.3 | 강한 연관 |

### 예시: 성별 × 지지정당

위의 분할표(표 W04.1)를 이용하여 카이자승 검정을 실행한다. 아래는 핵심 코드이다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

df = pd.read_csv("data/social_survey_w04.csv", encoding="utf-8")
print(f"데이터 크기: {df.shape}")
df.head()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

ct = pd.crosstab(df["성별"], df["지지정당"])
chi2, p, dof, expected = chi2_contingency(ct)
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"χ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")
print(f"Cramér's V = {V:.3f}")
```

_전체 코드는 practice/W04/code/W04-1-chi-square-intro.py 참고_

**검정 결과:**

- **χ²(3) = 16.262, p = 0.0010**
- **Cramér's V = 0.285**
- 기대빈도: 모두 5 이상 (최솟값 20.64) → 가정 충족

**표 W04.2** 성별 × 지지정당 기대빈도

| | A당 | B당 | C당 | 무당파 |
|---|---:|---:|---:|---:|
| 남성 | 32.76 | 22.36 | 24.96 | 23.92 |
| 여성 | 30.24 | 20.64 | 23.04 | 22.08 |

기대빈도(표 W04.2)와 관찰빈도(표 W04.1)를 비교하면, 남성은 A당에서 관찰빈도(45)가 기대빈도(32.76)보다 크고, 여성은 B당(관찰 27 vs 기대 20.64)과 C당(관찰 29 vs 기대 23.04)에서 관찰빈도가 기대빈도보다 크다. 즉, 남성은 A당(43.3%)을 더 지지하는 경향이 있고, 여성은 B당(28.1%)과 C당(30.2%)을 더 지지하는 경향이 관찰된다.

### 기본(Basic) 트랙

#### ① 실행

성별 × 지지정당 분할표를 생성하고 카이자승 검정을 실행한다.

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

# 1. 분할표 생성
ct = pd.crosstab(df["성별"], df["지지정당"])
print("=== 분할표 (관찰빈도) ===")
print(ct)

# 2. 카이자승 검정 실행
chi2, p, dof, expected = chi2_contingency(ct)
print(f"\nχ² = {chi2:.3f}, df = {dof}, p = {p:.4f}")

# 3. 기대빈도 출력
print("\n=== 기대빈도 ===")
print(pd.DataFrame(expected, index=ct.index, columns=ct.columns).round(2))
```

```python
n = ct.sum().sum()
V = np.sqrt(chi2 / (n * (min(ct.shape) - 1)))
print(f"Cramér's V = {V:.3f}")
```

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 해석문에 해당하는 항목이 없는지 점검해 보자.

**오류 1: 기대빈도 확인을 건너뜀**

카이자승 검정을 실행한 뒤 p값만 확인하고, 기대빈도를 출력하지 않는 경우가 많다. `chi2_contingency(ct)`는 4개의 값을 반환하며, 네 번째 값이 기대빈도 배열(`expected`)이다. 반드시 다음을 실행하여 확인한다.

```python
chi2, p, dof, expected = chi2_contingency(ct)
print(expected)  # 모든 값이 5 이상인지 눈으로 확인
```

기대빈도가 5 미만인 칸이 있으면, Fisher 정확 검정(`scipy.stats.fisher_exact`, 2×2 표의 경우)을 사용하거나 빈도가 적은 범주를 통합하여 재분석해야 한다.

**오류 2: "유의하다 = 효과가 크다"로 해석**

p값이 작다고 해서 효과가 크다는 뜻이 아니다. 표본이 크면 매우 약한 연관도 통계적으로 유의하게 나올 수 있다. 반드시 Cramér's V를 함께 보고하고, 아래 기준으로 실질적 의미를 판단한다.

- V < 0.1: 약한 연관 (통계적으로 유의하더라도 실질적 의미가 작다)
- 0.1 ≤ V ≤ 0.3: 중간 수준의 연관
- V > 0.3: 강한 연관

**오류 3: "카이자승 결과 = 인과"로 해석**

가장 흔하고 가장 심각한 오류이다. 카이자승 검정은 두 변수 간 **통계적 연관의 유무**만 판단한다. 인과의 방향이나 메커니즘에 대해서는 아무것도 말하지 않는다.

| 잘못된 표현 (X) | 올바른 표현 (O) |
|---|---|
| "성별이 정당 선택에 영향을 미친다" | "성별과 지지정당 사이에 통계적으로 유의한 연관이 관찰된다" |
| "성별 때문에 정치 성향이 달라진다" | "성별에 따라 지지정당의 분포에 차이가 나타난다" |
| "남성이라서 A당을 지지한다" | "남성에서 A당 지지 비율이 상대적으로 높게 나타난다" |

### 해석문 모범 구조

오늘 과제의 (D)항 해석문은 다음 4단계 구조를 따른다. 3주차에서 배운 틀을 카이자승 검정에 적용한 것이다.

> **(분석목적)** 성별과 지지정당의 연관을 확인하기 위해 카이자승 검정을 실행했다.
>
> **(핵심근거)** χ²(3) = 16.262, p = 0.001이며, Cramér's V = 0.285이다. 기대빈도는 모든 칸에서 5 이상(최솟값 20.64)으로 검정 가정을 충족했다.
>
> **(결론)** 유의수준 0.05에서 성별과 지지정당은 독립이 아닌 것으로 나타났으며, 중간 수준의 연관이 있다. 남성은 A당(43.3%), 여성은 B당(28.1%)과 C당(30.2%)을 상대적으로 더 지지하는 경향이 관찰된다.
>
> **(주의)** 이 결과는 성별이 정당 선택을 "결정"한다는 의미가 아니며, 연령·교육수준·거주지역 등 교란변수를 고려하지 않은 한계가 있다.
