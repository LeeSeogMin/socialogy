## 핵심 개념 학습

### 왜 통계 수업에서 AI를 쓰는가

통계 실습의 흐름은 대체로 "코드 작성 → 실행 → 에러 수정 → 결과 확인 → 해석 작성"이다. 이 중 코드 작성과 에러 수정에 드는 시간이 전체의 절반 이상을 차지하는 경우가 많다. Copilot은 이 앞단의 시간을 줄여줌으로써, 학생이 **결과 해석과 판단**에 더 많은 시간을 쓸 수 있게 한다.

그렇다고 코딩을 전혀 몰라도 된다는 뜻은 아니다. Copilot의 제안이 맞는지 틀린지를 판단하려면 코드를 읽고 이해하는 능력이 필요하다. Copilot은 **조교**이지, 학생을 대신하는 것이 아니다.

---

### 핵심 개념 1: Copilot의 역할 정의 — "코딩 조교"

**GitHub Copilot**은 코드 편집기(VS Code 등)에서 작동하는 AI 코딩 도우미이다. 크게 두 가지 기능을 제공한다.

1. **코드 완성(Autocomplete)**: 코드를 작성하다가 멈추면, 다음에 올 코드를 회색 글씨(ghost text)로 제안한다. Tab 키를 누르면 수락, Esc를 누르면 거절할 수 있다.
2. **Copilot Chat**: 채팅 창에서 자연어로 질문하면 코드, 설명, 디버깅 절차 등을 답변해준다.

이 수업에서 Copilot의 역할은 **"정답 생성기"가 아니라 "코딩/디버깅/해석문 다듬기 조교"**이다. 학생이 코딩에 쓰는 시간을 줄여, 더 많은 시간을 "해석·판단"에 할당하는 것이 목표이다.

쉽게 말해서, Copilot은 "시험 대신 봐주는 친구"가 아니라, "공부할 때 옆에서 도와주는 과외 선생님"이다. 최종 답안지에 이름을 쓰는 것은 학생 본인이다.

---

####  "Copilot이 틀릴 수 있을까?"

> **질문**: "Copilot이 틀린 답을 줄 수 있다고 생각하는가?"
>
> - 투표해 보자. (예 / 아니오)

Copilot은 대규모 코드 데이터로 학습한 AI 모델이다. 따라서 **틀린 코드를 자신 있게 제안하는 경우가 있다**. 예를 들어, 함수 이름을 잘못 쓰거나, 존재하지 않는 라이브러리를 불러오거나, 논리적으로 잘못된 계산을 하는 경우가 실제로 발생한다. 이것이 바로 "검증 원칙"이 필요한 이유이다.

---

### 핵심 개념 2: 검증 원칙

**검증 원칙**이란, Copilot이 생성한 코드와 설명을 반드시 **코드 실행 결과(출력/표/그래프)**로 확인한다는 규칙이다. 검증 없이 Copilot의 답변을 그대로 제출하면 안 된다.

구체적으로, 이 수업에서 모든 과제는 다음 형식으로 제출한다. **이 형식을 앞으로 매주 사용한다.**

- **(A) 분석 질문** (1-2문장): 무엇을 알고 싶은가?
- **(B) 사용 변수/수준**: 어떤 데이터를 사용했는가?
- **(C) 결과 근거**: 핵심 표/그래프/수치
- **(D) 해석문** (5-7문장): 결과가 무엇을 의미하는가?
- **(E) Copilot 사용 요약** (2-3줄): 어떤 요청을 했고, 무엇을 검증했는가?

특히 **(E)항**이 중요하다. Copilot에게 무엇을 요청했고, 그 답변을 어떻게 검증했는지를 기록하는 것이다. "Copilot이 해줬으니 끝"이 아니라, "Copilot이 해준 것을 내가 확인했다"가 핵심이다.

---

### 핵심 개념 3: 프라이버시 규칙

**프라이버시 규칙**이란, 개인 및 민감 데이터를 Copilot에 절대 입력하지 않는다는 원칙이다.

구체적으로 다음을 Copilot Chat이나 코드에 입력하면 안 된다.

- 이름, 전화번호, 학번, 주소 등 **개인 식별 정보**
- 건강 상태, 정치 성향, 종교, 성적 등 **민감 정보**
- 비공개 자료, 타인의 과제 파일 등

이 수업에서 사용하는 데이터는 모두 **익명화된 공개 데이터 또는 수업용 가상 데이터**이다. 학생 개인의 실제 데이터를 업로드하는 것은 금지한다.

---

### Copilot 사용 정책 요약

| 구분 | 내용 |
|---|---|
| **허용** | 개념 설명 요청, 분석 절차 안내, 디버깅, 결과 해석 초안, 글쓰기 도움 |
| **금지** | 개인정보/민감정보 업로드, 시험 답안 복사 제출, 데이터 조작/위조, 타인 과제 복제 |
| **증빙** | 과제에 결과 근거 + Copilot 사용 요약 첨부 |
| **책임** | Copilot 답변은 참고이며, 최종 결과의 정확성은 **학생 책임** |

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

설치 및 실습 중 자주 나타나는 오류를 정리한다.

**오류 1: Copilot 로그인이 안 됨**

VS Code에서 Copilot 아이콘이 비활성화 상태로 남아 있거나, "Sign In" 이후에도 제안이 나오지 않는 경우이다. 해결 절차는 다음과 같다.

1. VS Code를 완전히 종료한 후 다시 실행한다.
2. Extensions에서 GitHub Copilot 확장이 "Enabled" 상태인지 확인한다.
3. 문제가 지속되면 확장을 삭제하고 재설치한다.

**오류 2: Copilot Chat 창이 보이지 않음**

코드 완성(ghost text)은 작동하지만 Chat 창이 보이지 않는 경우이다. **GitHub Copilot Chat** 확장은 GitHub Copilot과 별도의 확장이므로, Extensions에서 "GitHub Copilot Chat"을 검색하여 설치되어 있는지 확인한다.

**오류 3: 결제 화면이 등장함**

GitHub Student Developer Pack 신청 과정에서 결제 정보를 요구하는 화면이 나올 수 있다. 이 경우 **즉시 멈추고 조교/교수에게 확인**한다. 계정 상태나 학교 정책에 따라 무료 이용 가능 여부가 다르다.

> **토론 질문**: 이 제출물에서 무엇이 문제인가? (2분, 옆 사람과 토론 후 전체 공유)

**문제점 예시:**

1. **결과 근거 누락**: 코드를 실행했는지, 결과가 55인지 확인할 수 없다. "완료했다"는 말만으로는 검증이 된 것이 아니다.

2. **해석 부재**: "Copilot이 짜줘서 완료했다"는 해석이 아니다. Copilot의 제안이 왜 맞는지(또는 틀린지)를 본인의 판단으로 서술해야 한다.

3. **Copilot 사용 요약 불충분**: "코드를 짜달라고 했다"만으로는 어떤 프롬프트를 사용했고, 답변을 어떻게 검증했는지 알 수 없다. 구체적인 요청 내용과 검증 과정을 기록해야 한다.

요약하면, Copilot은 "조교"이다. 조교가 해준 일을 본인이 확인하고, 이해하고, 설명할 수 있어야 한다. 그래야 **학생 본인의 학습**이 된다.

> **참고**: **척도(scale)**란 변수를 측정하는 수준을 말한다. **명목(nominal)**은 이름만 있는 범주(성별, 지역 등), **서열(ordinal)**은 순서가 있는 범주(만족도 1--5점), **비율(ratio)**은 절대 영점이 있는 연속 수치(연령, 소득)이다. 척도에 대해서는 다음 주(2주차)에 본격적으로 배운다. 오늘은 "변수마다 종류가 다르다"는 점만 인식하면 충분하다.

### 바이브코딩이란?

오늘 과제는 학생이 직접 코드를 작성하는 대신, **Copilot에게 데이터를 제시하고 분석을 요청**하는 방식(**바이브코딩**, vibe coding)으로 수행한다.

**바이브코딩 절차**:
1. Copilot Chat에 **구체적인 프롬프트**를 입력한다.
2. Copilot이 생성한 **코드를 복사하여 실행**한다.
3. 에러가 나면 에러 메시지를 Copilot에게 다시 보내 **수정 코드를 요청**한다.
4. 최종 코드와 실행 결과를 **모두 과제물에 포함**한다.

**바이브코딩 프롬프트 예시**:

| 용도 | 프롬프트 |
|---|---|
| 데이터 로드 + 요약 | "social_survey_w01.csv 파일을 pandas로 불러와서, 데이터 크기(행x열)와 기본 요약통계를 출력하는 코드를 짜줘." |
| 결측치 확인 | "이 데이터의 변수별 결측치 수를 확인하는 코드를 짜줘." |
| 에러 수정 | "아래 에러가 발생했다. 원인과 해결 코드를 알려줘: (에러 메시지 붙여넣기)" |

> **주의**: Copilot이 생성한 코드를 그대로 실행하기 전에, **코드가 무엇을 하는지 조원끼리 간단히 확인**한다. 코드의 의미를 전혀 모른 채 실행하는 것은 바이브코딩의 목적이 아니다.

  - **(주의)** 이 데이터는 가상 설문이므로, 결과를 실제 모집단에 일반화할 수 없다.

2주차에서는 **"이 데이터를 믿을 수 있는가?"**라는 질문으로, 데이터의 출처/편향/한계를 짚고, 변수의 **척도 수준**(명목/서열/등간/비율)과 **결측치 처리**를 본격적으로 배운다. 오늘 과제에서 "성별은 명목이다", "생활만족도는 서열이다"처럼 간단히 분류했던 내용을, 왜 그렇게 분류하는지, 분류가 달라지면 분석 방법이 어떻게 바뀌는지를 다음 시간에 깊이 있게 다룬다.
