### 사회통계가 말할 수 있는 것과 없는 것

**말할 수 있는 것**:
- 두 변수 사이에 통계적으로 유의한 관련/차이가 있는가
- 여러 변수를 함께 고려했을 때 각 변수의 고유 기여
- 결과의 불확실성 (신뢰구간, 효과크기)

**말할 수 없는 것**:
- 인과관계 (횡단 자료로는 방향 불명)
- 측정되지 않은 변수의 효과
- 표본을 넘어선 일반화 (표집틀을 벗어난 집단)
- 의미와 해석의 깊이 (질적 연구의 영역)

### 필요시 보강: 다음 단계로 가려면

본 과목 이후 학생이 진학·취업·연구에서 마주할 수 있는 주제를 간략히 안내한다. 시간이 남으면 이 중 1-2개를 30분씩 짧게 시연한다.

| 주제 | 언제 필요한가 | 학습 경로 |

| **인과추론 (IV, DID, RDD)** | 정책 효과, 사회적 개입 평가 | 계량경제학·정책분석 과목 |
| **패널 회귀 / 다층모형** | 종단 자료, 학교-학생 위계 자료 | 사회통계 심화, R 필요 |
| **질적 방법론** | 의미·경험·맥락 연구 | 사회조사방법론, 인터뷰·문화기술지 |
| **텍스트 분석 / 머신러닝** | 대규모 텍스트, 예측 문제 | 데이터사이언스·NLP 입문 |
| **베이즈 통계** | 사전 지식 반영, 작은 표본 | 수리통계 기반 필요 |

**중요**: 본 과목은 양적 분석의 **입구**이다. 이 입구에서 배운 "질문 → 데이터 → 검정 → 해석 → 한계"의 구조는 어떤 고급 방법론에도 그대로 적용된다.
