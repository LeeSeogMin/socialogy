# W03A. 데이터의 전체 모습을 어떻게 파악하고 표현할까? — 요약·분포·산포 + 해석문 구조

> 데이터의 전체적인 모습을 어떻게 한눈에 파악하고, 말로 표현할 수 있을까?

**학습 목표**
1. **중심 경향**(평균, 중앙값, 최빈값)과 **산포**(범위, 분산, 표준편차)를 계산하고, 각 지표의 특성과 한계를 설명할 수 있다.
2. 히스토그램과 박스플롯을 해석하여 분포의 형태(대칭/치우침)와 이상치를 파악할 수 있다.
3. 기술통계 결과를 **해석문 4단계 구조**(분석목적 → 핵심근거 → 결론 → 주의)로 작성할 수 있다.

## 핵심 개념 학습

### 왜 필요한가

- 2주차에서 우리는 데이터를 정리하고 변수의 척도를 판별했다. 그러나 정리된 데이터를 바로 분석에 사용하기 전에, **데이터 전체의 모습을 먼저 파악**하는 것이 중요하다. "응답자의 소득은 대략 얼마인가?", "연령은 고르게 분포하는가, 아니면 특정 연령대에 몰려 있는가?"와 같은 질문에 답하는 과정이 **기술통계(descriptive statistics)**이다.

- 기술통계는 크게 두 축으로 이루어진다. 첫째, 데이터의 "대표값"을 구하는 **중심 경향**, 둘째, 데이터가 얼마나 퍼져 있는지를 보는 **산포**이다. 이 두 가지를 함께 보아야 데이터의 전체 모습을 파악할 수 있다. 오늘은 여기에 더하여, 분석 결과를 체계적으로 서술하는 **해석문 4단계 구조**를 처음 도입한다. 이 구조는 이번 주차부터 14주차까지 모든 과제에서 반복 사용하게 된다.

---

### 핵심 개념 1: 중심 경향(central tendency)

- **중심 경향(central tendency)**이란?
  - 데이터의 "대표값"을 나타내는 지표
  - 데이터 전체를 하나의 숫자로 요약하려 할 때 사용

- **대표적인 세 가지 지표:**
  - **평균(mean)**: 모든 값의 합을 개수로 나눈 것
    - 가장 널리 쓰이지만, **이상치에 민감**
    - 극단적으로 크거나 작은 값이 하나만 있어도 평균이 크게 변함
  - **중앙값(median)**: 값을 크기순으로 나열했을 때 **정가운데**에 위치하는 값
    - 이상치에 **강건(robust)**
    - 분포가 치우쳐 있을 때 평균보다 더 대표적인 값이 됨
  - **최빈값(mode)**: 가장 빈번하게 나타나는 값
    - **범주형 데이터**에서 특히 유용

- **핵심 포인트:**
  - **분포가 치우쳐 있으면 평균과 중앙값이 달라진다**
  - 이때 중앙값이 "전형적인 값"을 더 잘 대표하는 경우가 많음

- **쉬운 예시 — 소득 분포:**
  - 대부분의 사람: 200~400만 원 사이 소득
  - 소수의 고소득자: 월 1,000만 원 이상
  - 소수의 고소득자가 평균을 끌어올림
  - 결과: **평균 소득 > 중앙값 소득**
  - 따라서 중앙값이 "보통 사람의 소득"을 더 잘 보여줌

---

#### 예측 투표 — 평균 vs 중앙값

> **투표 질문**: "한국 가구의 월 평균 소득과 중앙값 소득 중 어느 쪽이 더 클까?"
>
> ① 평균이 더 크다
> ② 중앙값이 더 크다
> ③ 거의 같다
>
> — 투표해 보자. (10초)

- **정답: ① 평균이 더 크다**
- **이유:**
  - 소득 분포는 오른쪽으로 치우쳐 있음 (소수의 고소득자가 평균을 끌어올림)
  - 평균 소득이 중앙값 소득보다 항상 높게 나타남
  - 통계청 자료에서도 가구 평균 소득은 중앙값 소득보다 약 20~30% 높음

---

### 핵심 개념 2: 산포(dispersion)

- **산포(dispersion)**란?
  - 데이터가 중심(대표값)에서 **얼마나 퍼져 있는지**를 나타내는 지표
  - 평균이 같더라도 산포가 다르면 전혀 다른 분포일 수 있음

- **주요 산포 지표:**
  - **범위(range)**: 최댓값 − 최솟값
    - 가장 간단하지만, 극단값(이상치) 하나에 크게 좌우됨
  - **분산(variance)**: 각 값과 평균의 차이를 **제곱**하여 평균 낸 것
    - 제곱 때문에 원래 단위와 다름 (예: 만원² → 해석 어려움)
  - **표준편차(standard deviation)**: 분산의 **제곱근**
    - 원래 단위와 같아 해석이 쉬움
    - 가장 많이 사용하는 산포 지표

- **핵심 포인트:**
  - **"평균만으로는 분포의 퍼짐 정도를 알 수 없다"**
  - 예: 두 학급 시험 평균이 모두 70점이지만
    - A반: 60~80점에 몰려 있음 (표준편차 작음)
    - B반: 30~100점에 퍼져 있음 (표준편차 큼)
    - → 평균만 보고하면 이 차이를 놓치게 됨

- **요약통계 해석 체크리스트:**
  - 평균과 중앙값의 차이가 크면 → **치우침(skew)** 의심
  - 표준편차가 평균에 비해 크면 → **이질적 집단** 가능성 (다양한 하위 그룹이 섞여 있을 수 있음)

### 핵심 개념 3: 해석문 4단계 구조(interpretation template)

- **해석문 4단계 구조**란?
  - 통계 분석 결과를 체계적으로 서술하는 틀
  - **이번 주차에서 처음 도입**, 이후 4주~14주까지 모든 과제에서 반복 사용

- **4단계 구조:**
  1. **(분석목적)** "___를 보기 위해 ___를 사용했다."
  2. **(핵심근거)** "출력에서 ___(수치/표/그래프)를 확인했다."
  3. **(결론)** "따라서 ___이라고 말할 수 있다."
  4. **(주의)** "이 결과는 ___를 의미하지 않으며, ___의 한계가 있다."

- **왜 필요한가?**
  - 숫자만 나열 → 독자는 "그래서 무엇을 말하려는 건가?"를 알 수 없음
  - 결론만 제시 → "근거가 무엇인가?"를 알 수 없음
  - 4단계 구조 → **왜 분석했는지 → 무엇을 봤는지 → 무엇을 말할 수 있는지 → 무엇을 말할 수 없는지**를 빠짐없이 포함

### 예시: social_survey_w03.csv 기술통계

- 오늘 사용하는 데이터: 2주차에서 정리한 설문 데이터의 깨끗한 버전 (n=200)

- **표 W03.1** 주요 변수의 기술통계 (n=200):

| 변수 | 평균 | 중앙값 | 표준편차 | 최솟값 | 최댓값 |
|---|---:|---:|---:|---:|---:|
| 연령 | 37.4 | 37.0 | 11.4 | 20 | 65 |
| 소득(만원) | 351.1 | 306.5 | 189.6 | 80 | 980 |
| 생활만족도(1-5) | 3.30 | 3.0 | 1.1 | 1 | 5 |

- **주목할 패턴:**

  - **소득:**
    - 평균(351.1) > 중앙값(306.5), 차이 = 44.6만 원으로 큼
    - → **오른쪽으로 치우친 분포** (소수의 고소득자가 평균을 끌어올림)
    - 표준편차(189.6)가 평균(351.1)의 절반 이상 → 소득 분포가 매우 넓게 퍼져 있음

  - **연령:**
    - 평균(37.4) ≈ 중앙값(37.0), 차이 = 0.4세로 작음
    - → 대체로 **대칭에 가까운 분포**

  - **생활만족도:**
    - 평균(3.30) ≈ 중앙값(3.0)
    - 5점 척도이므로 중간 수준

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/social_survey_w03.csv", encoding='utf-8')

# 기술통계
print(df[["연령", "소득", "생활만족도"]].describe())

# 히스토그램 + 평균/중앙값 표시
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df["소득"].hist(bins=20, ax=axes[0])
axes[0].set_title("소득 분포")
axes[0].axvline(df["소득"].mean(), color='r', linestyle='--', label=f'평균={df["소득"].mean():.0f}')
axes[0].axvline(df["소득"].median(), color='g', linestyle='-', label=f'중앙값={df["소득"].median():.0f}')
axes[0].legend()

# 박스플롯
df.boxplot(column="소득", ax=axes[1])
axes[1].set_title("소득 박스플롯")
plt.tight_layout()
plt.show()
```

### 그래프 해석 가이드

- 시각화는 숫자만으로는 보이지 않는 패턴을 드러냄
- 오늘은 두 가지 그래프에 집중

- **히스토그램(histogram):**
  - 연속형 변수의 분포를 막대로 표현한 것
  - 막대가 한쪽으로 치우쳐 있으면 → **비대칭(skewed)** 분포
  - 빨간 점선(평균)과 초록 실선(중앙값)의 위치 차이를 확인
  - 막대가 좌우 대칭이면 → 평균과 중앙값이 거의 같음

- **박스플롯(boxplot):**
  - 분포의 요약 정보를 상자와 수염으로 표현한 것
  - 상자 안의 선 = **중앙값**
  - 상자의 아래/위 변 = **1사분위수(Q1)/3사분위수(Q3)**
  - 상자 바깥의 점 = **이상치(outlier)**
  - 상자가 한쪽으로 길면 → 비대칭

- **그래프 선택 가이드:**

| 변수 유형 | 권장 그래프 |
|---|---|
| 연속형(소득, 연령 등) | 히스토그램, 박스플롯 |
| 범주형(성별, 교육수준 등) | 막대그래프 |
| 두 연속형 변수의 관계 | 산점도 |

### 기본(Basic) 트랙

#### ① 실행

> data/social_survey_w03.csv에서 연령/소득/생활만족도의 기술통계를 출력하고, 소득 히스토그램에 평균선과 중앙값선을 함께 표시하는 코드를 작성해줘


```python
# ============================================================
# 1단계: 연령, 소득, 생활만족도의 기술통계 출력
# ============================================================
# 보기 좋은 제목 출력
print("=" * 60)
print("연령/소득/생활만족도 기술통계")
print("=" * 60)

# df에서 3개 열만 선택하고 .describe() 메서드로 기술통계 계산
# describe()는 평균, 표준편차, 최솟값, 사분위수, 최댓값 등을 자동 계산
descriptive_stats = df[['연령', '소득', '생활만족도']].describe()
print(descriptive_stats)

# ============================================================
# 2단계: 소득 히스토그램 그리기 (평균선과 중앙값선 포함)
# ============================================================

# 그래프 만들기: figsize=(10, 6)는 가로10, 세로6 크기를 의미
# fig = 전체 그림, ax = 그래프 그리는 축(좌표 공간)
fig, ax = plt.subplots(figsize=(10, 6))

# 히스토그램 그리기
# - df['소득']: 소득 데이터만 추출
# - bins=20: 소득을 20개 구간으로 나누어 표시
# - color='skyblue': 막대 색깔을 하늘색으로
# - edgecolor='black': 막대 테두리를 검은색으로
# - alpha=0.7: 투명도 70% (0~1 사이 값)
# - label='소득 분포': 범례에 표시할 이름
ax.hist(df['소득'], bins=20, color='skyblue', edgecolor='black', alpha=0.7, label='소득 분포')

# ============================================================
# 3단계: 평균과 중앙값 계산하기
# ============================================================
# .mean(): 평균 계산 (모든 값을 더해서 개수로 나눔)
mean_income = df['소득'].mean()

# .median(): 중앙값 계산 (값을 작은 순서부터 나열했을 때 정가운데 값)
median_income = df['소득'].median()

# ============================================================
# 4단계: 평균선과 중앙값선을 히스토그램에 추가
# ============================================================

# 평균선 추가 (세로 직선)
# - mean_income: x축 위치 (평균값)
# - color='red': 빨간색
# - linestyle='--': 점선 스타일
# - linewidth=2: 선 두께
# - label=f'평균: {mean_income:.1f}': 범례에 평균값 표시 (.1f = 소수점 1자리)
ax.axvline(mean_income, color='red', linestyle='--', linewidth=2, label=f'평균: {mean_income:.1f}')

# 중앙값선 추가 (세로 직선)
# - median_income: x축 위치 (중앙값)
# - color='green': 초록색
# - linestyle='--': 점선 스타일
ax.axvline(median_income, color='green', linestyle='--', linewidth=2, label=f'중앙값: {median_income:.1f}')

# ============================================================
# 5단계: 그래프 꾸미기 (축, 제목, 범례, 격자 등)
# ============================================================

# X축 설명: "소득 (만원)"
# fontsize=12: 글자 크기 12
ax.set_xlabel('소득 (만원)', fontsize=12)

# Y축 설명: "빈도"
# 빈도 = 각 구간에 몇 개의 데이터가 들어가는가? (개수)
ax.set_ylabel('빈도', fontsize=12)

# 그래프의 제목
# fontweight='bold': 글자를 굵게 표시
ax.set_title('소득 분포 (평균선과 중앙값선)', fontsize=14, fontweight='bold')

# 범례 표시 (오른쪽 위 자동 배치)
# fontsize=11: 범례 글자 크기
ax.legend(fontsize=11)

# 배경 격자 추가 (읽기 쉽게)
# axis='y': Y축 방향 격자선만 그리기
# alpha=0.3: 투명도 30% (약하게 보이도록)
ax.grid(axis='y', alpha=0.3)

# ============================================================
# 6단계: 그래프 보여주기
# ============================================================

# plt.tight_layout(): 모든 요소가 겹치지 않도록 자동 배치
plt.tight_layout()

# 그래프를 화면에 출력
plt.show()

# ============================================================
# 7단계: 핵심 통계 결과 정리해서 출력
# ============================================================
print("\n소득 통계:")
# .2f: 소수점 2자리까지 표시
print(f"  평균: {mean_income:.2f}")
print(f"  중앙값: {median_income:.2f}")
# .std(): 표준편차 계산 (데이터가 평균에서 얼마나 퍼져있는가?)
print(f"  표준편차: {df['소득'].std():.2f}")
```

### 확장(Plus) 트랙

> 동일 데이터에서 교육수준 빈도표와 교육수준별 소득 박스플롯을 생성하고, 집단별 분포 차이를 비교하는 코드를 작성해줘


```python
# ============================================================
# 1단계: 교육수준 빈도표 생성
# ============================================================
print("=" * 60)
print("교육수준 빈도표")
print("=" * 60)

# .value_counts(): 각 카테고리가 몇 개씩 있는지 세기
# sort=False로 지정하면 원래 순서 유지 (생략하면 빈도 내림차순)
edu_freq = df['교육수준'].value_counts(sort=False)
print(edu_freq)
print(f"\n총 응답자: {edu_freq.sum()}명")

# ============================================================
# 2단계: 교육수준별 소득 박스플롯 그리기
# ============================================================
print("\n" + "=" * 60)
print("교육수준별 소득 기술통계")
print("=" * 60)

# 교육수준 그룹별로 소득 통계 계산
# .groupby(): 데이터를 교육수준별로 나누어 처리
# aggregation functions: 각 그룹별 통계량 계산
edu_income_stats = df.groupby('교육수준')['소득'].agg(['count', 'mean', 'median', 'std', 'min', 'max'])
print(edu_income_stats)

# 그룹별 통계 중 중앙값만 따로 빼서 표시
print("\n[그룹별 중앙값 비교]")
for edu in sorted(df['교육수준'].unique()):
    median_val = df[df['교육수준'] == edu]['소득'].median()
    print(f"  {edu}: {median_val:.1f}만원")

# ============================================================
# 3단계: 교육수준별 박스플롯 그리기
# ============================================================

# 그래프 만들기
fig, ax = plt.subplots(figsize=(10, 6))

# 교육수준별 박스플롯
# - data=df: 사용할 데이터 지정
# - column='소득': 소득으로 박스플롯 그리기
# - by='교육수준': 교육수준별로 그룹 나누기
bp = df.boxplot(column='소득', by='교육수준', ax=ax, patch_artist=True)

# 그래프 꾸미기
ax.set_xlabel('교육수준', fontsize=12, fontweight='bold')
ax.set_ylabel('소득 (만원)', fontsize=12, fontweight='bold')
ax.set_title('교육수준별 소득 분포 (박스플롯)', fontsize=14, fontweight='bold')
plt.suptitle('')  # 자동으로 생기는 제목 제거

# 격자 추가
ax.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

# ============================================================
# 4단계: 집단별 분포 차이 비교
# ============================================================
print("\n" + "=" * 60)
print("집단별 분포 특성 비교")
print("=" * 60)

# 각 교육수준별로 상세 분석
for edu in sorted(df['교육수준'].unique()):
    edu_data = df[df['교육수준'] == edu]['소득']
    
    print(f"\n【{edu}】")
    print(f"  표본수: {len(edu_data):3d}명 ({len(edu_data)/len(df)*100:.1f}%)")
    print(f"  평균: {edu_data.mean():7.1f}만원")
    print(f"  중앙값: {edu_data.median():6.1f}만원")
    print(f"  표준편차: {edu_data.std():6.1f}만원")
    print(f"  범위: {edu_data.min():.1f}~{edu_data.max():.1f}만원")
    print(f"  1사분위수(Q1): {edu_data.quantile(0.25):6.1f}만원")
    print(f"  3사분위수(Q3): {edu_data.quantile(0.75):6.1f}만원")
    print(f"  IQR(사분범위): {edu_data.quantile(0.75) - edu_data.quantile(0.25):6.1f}만원")
    
    # 평균과 중앙값의 차이 (치우침 여부)
    diff = edu_data.mean() - edu_data.median()
    if abs(diff) < 20:
        skew_status = "대칭에 가까움"
    elif diff > 0:
        skew_status = "우편향(오른쪽으로 치우침)"
    else:
        skew_status = "좌편향(왼쪽으로 치우침)"
    print(f"  분포 형태: {skew_status}")

# ============================================================
# 5단계: 전체 요약
# ============================================================
print("\n" + "=" * 60)
print("핵심 발견사항 요약")
print("=" * 60)

# 교육수준 그룹별 중앙값 비교해서 어느 그룹이 가장 높은지 찾기
edu_medians = {}
for edu in df['교육수준'].unique():
    edu_medians[edu] = df[df['교육수준'] == edu]['소득'].median()

highest_edu = max(edu_medians, key=edu_medians.get)
lowest_edu = min(edu_medians, key=edu_medians.get)

print(f"\n1. 소득이 가장 높은 집단: {highest_edu} (중앙값: {edu_medians[highest_edu]:.1f}만원)")
print(f"2. 소득이 가장 낮은 집단: {lowest_edu} (중앙값: {edu_medians[lowest_edu]:.1f}만원)")
print(f"3. 집단간 중앙값 차이: {edu_medians[highest_edu] - edu_medians[lowest_edu]:.1f}만원")

# 표준편차로 분포 퍼짐 정도 비교
print("\n[표준편차로 본 분포 퍼짐 정도]")
for edu in sorted(df['교육수준'].unique()):
    std_val = df[df['교육수준'] == edu]['소득'].std()
    print(f"  {edu}: {std_val:.1f}만원")
```

#### 해석

> 위 결과를 해석문 4단계 구조로 작성하세요.

**분석목적**

조사 대상자를 교육수준 별로 집단화하여, 교육수준이 소득 분포에 미치는 영향을 파악하고 집단 간 소득 격차를 비교하기 위해 빈도표와 박스플롯을 생성했다.

**핵심근거**

- **표본 구성**: 대졸 80명(40%) > 고졸이하 66명(33%) > 대학원이상 54명(27%)

- **집단별 중앙값 비교** (소득의 "전형적 값"):
  - 고졸이하: 203.5만원
  - 대졸: 311.5만원
  - 대학원이상: 485.5만원
  - → 교육수준이 높을수록 중앙값이 뚜렷이 증가 (최고-최저 격차: 282만원)

- **집단별 산포 차이** (표준편차):
  - 고졸이하: 98.6만원 (Q1:161, Q3:272, IQR:111)
  - 대졸: 180.6만원 (Q1:198.2, Q3:456.5, IQR:258.2)
  - 대학원이상: 176.2만원 (Q1:353.2, Q3:645.2, IQR:292.0)
  - → 교육수준이 높을수록 IQR과 표준편차 증가 (소득 다양성 증가)

- **분포 형태**: 모든 세 집단이 **우편향(right-skewed) 분포**
  - 각 집단 내에서도 소수의 고소득자가 존재 (평균 > 중앙값)
  - 박스플롯에서 대졸과 대학원이상 집단에 뚜렷한 **이상치(outlier)** 관찰

**결론**

조사 표본에서 **교육수준에 따라 소득 수준과 분포 특성이 명확히 다르게 나타난다**. 구체적으로:

1. **소득 수준의 차이**: 대학원이상 집단의 중앙값(485.5만원)은 고졸이하 집단(203.5만원)의 **2.4배** 수준으로, 교육수준이 높을수록 소득이 높음

2. **소득 불평등의 차이**: 대학원이상 집단의 IQR(292.0)은 고졸이하 집단(111.0)의 **2.6배**로, 고학력 집단 내 소득 격차가 더 큼

3. **이상치의 분포**: 고소득자(이상치)는 대학원이상과 대졸 집단에 더 많이 분포하며, 이는 **고학력이 고소득 기회를 더 많이 제공**함을 시사

**주의**

- 이 결과는 **기술통계이므로 교육수준이 소득을 인과적으로 결정한다는 것을 의미하지 않으며**, 두 변수 간의 "상관"을 보여줄 뿐이다.

- **표본의 한계**: 이 표본(n=200)은 특정 기간·지역·연령대에 국한되었을 가능성이 높으며, 전체 한국 인구으로 일반화할 수 없다.

- **역인과 관계 가능성**: 데이터의 횡단면 특성상, 높은 교육 수준이 높은 소득을 가져왔는지, 아니면 높은 소득을 가진 사람들이 더 높은 교육을 받은 것인지 구분할 수 없다.

- **누락된 변수**: 성별, 근무 산업, 경력 등 다른 중요한 변수들이 소득에 미치는 영향을 통제하지 않았으며, 이들이 교육수준-소득 관계를 매개할 가능성이 있다.

- **측정 오류**: 자기보고식 설문의 특성상 소득이 과소보고되었을 가능성이 있으며, 특히 고소득층에서 더 클 수 있다.

### 흔한 오류 3개

- 실습 중 자주 나타나는 오류 정리
- 자신의 해석문에 해당하는 항목이 없는지 점검

**오류 1: "평균 = 대부분의 값"이라는 오해**
- 치우친 분포에서 평균은 대부분의 관찰값과 다를 수 있음
- 소득 분포에서 평균(351.1만 원)보다 적게 버는 응답자가 절반 이상일 수 있음
- **반드시 중앙값과 함께 보고**하고, 평균과 중앙값의 차이가 큰지 확인 필요

```python
# 확인법: 평균과 중앙값 차이 비교
diff = df["소득"].mean() - df["소득"].median()
print(f"평균 - 중앙값 = {diff:.1f}만원")  # 차이가 크면 치우침 의심
```

**오류 2: 표준편차를 보고하지 않음**
- 평균만으로는 분포를 파악할 수 없음
- "평균 소득 351만 원"이라고만 쓰면 → 모두가 300~400만 원 사이인지 80~980만 원에 퍼져 있는지 알 수 없음
- **항상 평균 ± 표준편차 형태로 보고**
- 올바른 보고 예: "소득의 평균은 351.1만 원(SD = 189.6만 원)이다."

**오류 3: 시각화 없이 숫자만 나열**
- 기술통계 숫자만 보고 "대칭이다", "이상치가 있다"고 판단하는 것은 위험
- 히스토그램과 박스플롯을 **반드시 함께 확인**
- **"숫자가 말해주지 않는 것을 그래프가 보여준다"**
- 예: 동일한 평균·표준편차를 가진 전혀 다른 분포가 존재할 수 있음 (Anscombe의 사분주)

   - 소수의 고소득자가 평균을 끌어올린 것일 수 있음
   - "대부분이 안정적"이라는 판단은 근거가 부족

   - 특정 직업군이나 연령대만 포함된 표본 → 전체 국민에게 일반화 불가

   - 표준편차가 300만 원이라면 → 소득 격차가 매우 크다는 뜻
   - 퍼짐 정도를 확인하지 않고 "안정적"이라고 결론 → 과잉 해석
