> 데이터의 전체적인 모습을 어떻게 한눈에 파악하고, 말로 표현할 수 있을까?

**학습 목표**
1. **중심 경향**(평균, 중앙값, 최빈값)과 **산포**(범위, 분산, 표준편차)를 계산하고, 각 지표의 특성과 한계를 설명할 수 있다.

3. 기술통계 결과를 **해석문 4단계 구조**(분석목적 → 핵심근거 → 결론 → 주의)로 작성할 수 있다.

## 핵심 개념 학습

### 왜 필요한가

2주차에서 우리는 데이터를 정리하고 변수의 척도를 판별했다. 그러나 정리된 데이터를 바로 분석에 사용하기 전에, **데이터 전체의 모습을 먼저 파악**하는 것이 중요하다. "응답자의 소득은 대략 얼마인가?", "연령은 고르게 분포하는가, 아니면 특정 연령대에 몰려 있는가?"와 같은 질문에 답하는 과정이 **기술통계(descriptive statistics)**이다.

기술통계는 크게 두 축으로 이루어진다. 첫째, 데이터의 "대표값"을 구하는 **중심 경향**, 둘째, 데이터가 얼마나 퍼져 있는지를 보는 **산포**이다. 이 두 가지를 함께 보아야 데이터의 전체 모습을 파악할 수 있다. 오늘은 여기에 더하여, 분석 결과를 체계적으로 서술하는 **해석문 4단계 구조**를 처음 도입한다. 이 구조는 이번 주차부터 14주차까지 모든 과제에서 반복 사용하게 된다.

---

### 핵심 개념 1: 중심 경향(central tendency)

**중심 경향(central tendency)**이란 데이터의 "대표값"을 나타내는 지표이다. 데이터 전체를 하나의 숫자로 요약하려 할 때 사용한다. 대표적인 세 가지 지표는 다음과 같다.

- **평균(mean)**: 모든 값의 합을 개수로 나눈 것이다. 가장 널리 쓰이지만, **이상치에 민감**하다. 극단적으로 크거나 작은 값이 하나만 있어도 평균이 크게 변한다.
- **중앙값(median)**: 값을 크기순으로 나열했을 때 **정가운데**에 위치하는 값이다. 이상치에 **강건(robust)**하다. 분포가 치우쳐 있을 때 평균보다 더 대표적인 값이 된다.
- **최빈값(mode)**: 가장 빈번하게 나타나는 값이다. **범주형 데이터**에서 특히 유용하다.

핵심 포인트는 이것이다. **분포가 치우쳐 있으면 평균과 중앙값이 달라진다.** 이때 중앙값이 "전형적인 값"을 더 잘 대표하는 경우가 많다.

쉽게 말해서, 소득 분포를 생각해 보자. 대부분의 사람은 200~400만 원 사이의 소득을 올리지만, 소수의 고소득자(월 1,000만 원 이상)가 있다. 이 소수의 고소득자가 평균을 끌어올린다. 그래서 **평균 소득 > 중앙값 소득**이 되며, 중앙값이 "보통 사람의 소득"을 더 잘 보여준다.

---

#### 예측 투표 — 평균 vs 중앙값

> **투표 질문**: "한국 가구의 월 평균 소득과 중앙값 소득 중 어느 쪽이 더 클까?"
>
> ① 평균이 더 크다
> ② 중앙값이 더 크다
> ③ 거의 같다
>
> — 투표해 보자. (10초)

정답은 **① 평균이 더 크다**이다. 소득 분포는 오른쪽으로 치우쳐 있기 때문에(소수의 고소득자가 평균을 끌어올림), 평균 소득이 중앙값 소득보다 항상 높게 나타난다. 통계청 자료에서도 가구 평균 소득은 중앙값 소득보다 약 20~30% 높다.

---

### 핵심 개념 2: 산포(dispersion)

**산포(dispersion)**란 데이터가 중심(대표값)에서 **얼마나 퍼져 있는지**를 나타내는 지표이다. 평균이 같더라도 산포가 다르면 전혀 다른 분포일 수 있다.

- **범위(range)**: 최댓값 − 최솟값이다. 가장 간단하지만, 극단값(이상치) 하나에 크게 좌우된다.
- **분산(variance)**: 각 값과 평균의 차이를 **제곱**하여 평균 낸 것이다. 제곱 때문에 원래 단위와 다르다(예: 만원² → 해석 어려움).
- **표준편차(standard deviation)**: 분산의 **제곱근**이다. 원래 단위와 같아 해석이 쉽다. 가장 많이 사용하는 산포 지표이다.

핵심 포인트는 이것이다. **"평균만으로는 분포의 퍼짐 정도를 알 수 없다."** 예를 들어, 두 학급의 시험 평균이 모두 70점이지만, A반은 60~80점에 몰려 있고(표준편차 작음), B반은 30~100점에 퍼져 있다면(표준편차 큼), 두 학급의 성적 분포는 전혀 다르다. 평균만 보고하면 이 차이를 놓치게 된다.

**요약통계 해석 체크리스트:**
- 평균과 중앙값의 차이가 크면 → **치우침(skew)** 의심
- 표준편차가 평균에 비해 크면 → **이질적 집단** 가능성 (다양한 하위 그룹이 섞여 있을 수 있다)

### 핵심 개념 3: 해석문 4단계 구조(interpretation template)

통계 분석 결과를 체계적으로 서술하는 틀이 **해석문 4단계 구조**이다. 이 구조는 **이번 주차에서 처음 도입**하며, 이후 4주~14주까지 모든 과제에서 반복 사용한다.

- **(분석목적)** "___를 보기 위해 ___를 사용했다."
- **(핵심근거)** "출력에서 ___(수치/표/그래프)를 확인했다."
- **(결론)** "따라서 ___이라고 말할 수 있다."
- **(주의)** "이 결과는 ___를 의미하지 않으며, ___의 한계가 있다."

이 틀이 왜 필요한가? 통계 분석 결과를 보고할 때, 숫자만 나열하면 독자는 "그래서 무엇을 말하려는 건가?"를 알 수 없다. 반대로 결론만 쓰면 "근거가 무엇인가?"를 알 수 없다. 4단계 구조는 **왜 분석했는지 → 무엇을 봤는지 → 무엇을 말할 수 있는지 → 무엇을 말할 수 없는지**를 빠짐없이 포함하도록 안내한다.

### 예시: social_survey_w03.csv 기술통계

오늘 사용하는 데이터는 2주차에서 정리한 설문 데이터의 깨끗한 버전이다(n=200). 아래는 주요 변수의 기술통계 결과이다.

**표 W03.1** 주요 변수의 기술통계 (n=200)

| 변수 | 평균 | 중앙값 | 표준편차 | 최솟값 | 최댓값 |
|---|---:|---:|---:|---:|---:|
| 연령 | 38.2 | 37.0 | 11.8 | 20 | 65 |
| 소득(만원) | 348.5 | 285.0 | 182.3 | 80 | 1200 |
| 생활만족도(1-5) | 3.18 | 3.0 | 1.05 | 1 | 5 |

이 표에서 주목할 패턴은 다음과 같다.

- **소득**: 평균(348.5) > 중앙값(285.0)이다. 차이가 63.5만 원으로 크다. 이는 **오른쪽으로 치우친 분포**(소수의 고소득자가 평균을 끌어올림)를 시사한다. 표준편차(182.3)가 평균(348.5)의 절반 이상이므로, 소득 분포가 매우 넓게 퍼져 있다.
- **연령**: 평균(38.2) ≈ 중앙값(37.0)이다. 차이가 1.2세로 작다. 대체로 **대칭에 가까운 분포**로 보인다.
- **생활만족도**: 평균(3.18) ≈ 중앙값(3.0)이다. 5점 척도이므로 중간 수준이다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/social_survey_w03.csv", encoding='utf-8')

# 기술통계
print(df[["연령", "소득", "생활만족도"]].describe())

# 히스토그램 + 평균/중앙값 표시
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df["소득"].hist(bins=20, ax=axes[0])
axes[0].set_title("소득 분포")
axes[0].axvline(df["소득"].mean(), color='r', linestyle='--', label=f'평균={df["소득"].mean():.0f}')
axes[0].axvline(df["소득"].median(), color='g', linestyle='-', label=f'중앙값={df["소득"].median():.0f}')
axes[0].legend()

# 박스플롯
df.boxplot(column="소득", ax=axes[1])
axes[1].set_title("소득 박스플롯")
plt.tight_layout()
plt.show()
```

### 그래프 해석 가이드

시각화는 숫자만으로는 보이지 않는 패턴을 드러낸다. 오늘은 두 가지 그래프에 집중한다.

**히스토그램(histogram)**: 연속형 변수의 분포를 막대로 표현한 것이다.
- 막대가 한쪽으로 치우쳐 있으면 → **비대칭(skewed)** 분포. 빨간 점선(평균)과 초록 실선(중앙값)의 위치 차이를 확인한다.
- 막대가 좌우 대칭이면 → 평균과 중앙값이 거의 같다.

**박스플롯(boxplot)**: 분포의 요약 정보를 상자와 수염으로 표현한 것이다.
- 상자 안의 선 = **중앙값**. 상자의 아래/위 변 = **1사분위수(Q1)/3사분위수(Q3)**.
- 상자 바깥의 점 = **이상치(outlier)**. 상자가 한쪽으로 길면 → 비대칭.

**그래프 선택 가이드:**
| 변수 유형 | 권장 그래프 |
|---|---|
| 연속형(소득, 연령 등) | 히스토그램, 박스플롯 |
| 범주형(성별, 교육수준 등) | 막대그래프 |
| 두 연속형 변수의 관계 | 산점도 |

### 기본(Basic) 트랙

#### ① 실행

소득과 생활만족도의 기술통계를 출력하고, 소득 히스토그램을 그린다.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/social_survey_w03.csv", encoding='utf-8')

# 1. 기술통계 출력
print("=== 기술통계 ===")
print(df[["소득", "생활만족도"]].describe())

# 2. 소득 히스토그램 + 평균/중앙값 표시
df["소득"].hist(bins=20)
plt.axvline(df["소득"].mean(), color='r', linestyle='--', label='평균')
plt.axvline(df["소득"].median(), color='g', linestyle='-', label='중앙값')
plt.title("소득 분포")
plt.xlabel("소득(만원)")
plt.legend()
plt.show()
```

**완료 체크**: 기술통계(평균, 중앙값, 표준편차, 최솟값, 최댓값)가 출력되었는가? 히스토그램에 평균선과 중앙값선이 표시되었는가?

#### ② 확인

평균과 중앙값의 차이를 확인하고, Copilot 응답을 확인하고 검증한다.

**2-1. 평균 vs 중앙값 비교**: 소득 변수에서 평균과 중앙값의 차이를 확인한다. 차이가 크다면, 어느 방향으로 치우쳐 있는지 판단한다.

**2-2. Copilot 프롬프트 (필수 1)**: Copilot에게 다음을 질문한다.

> "이 변수의 분포가 정규분포에 가까운지 어떻게 판단하는가?"

Copilot의 답변을 읽고, **1가지를 채택 또는 거절**한다. 채택/거절 이유를 결과 근거로 2문장 이내로 작성한다.

*(예: "Copilot이 '히스토그램의 형태가 종 모양인지 확인하라'고 답했다. 채택한다. 실제로 소득 히스토그램은 종 모양이 아니라 오른쪽으로 길게 늘어진 형태이므로, 정규분포에 가깝지 않다.")*

**완료 체크**: 평균과 중앙값의 차이를 기록했는가? Copilot 채택/거절 기록이 있는가?

#### ③ 해석

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 해석문에 해당하는 항목이 없는지 점검해 보자.

**오류 1: "평균 = 대부분의 값"이라는 오해**

치우친 분포에서 평균은 대부분의 관찰값과 다를 수 있다. 소득 분포에서 평균(348.5만 원)보다 적게 버는 응답자가 절반 이상일 수 있다. 반드시 **중앙값과 함께** 보고하고, 평균과 중앙값의 차이가 큰지 확인한다.

```python
# 확인법: 평균과 중앙값 차이 비교
diff = df["소득"].mean() - df["소득"].median()
print(f"평균 - 중앙값 = {diff:.1f}만원")  # 차이가 크면 치우침 의심
```

**오류 2: 표준편차를 보고하지 않음**

평균만으로는 분포를 파악할 수 없다. "평균 소득 348만 원"이라고만 쓰면, 모두가 300~400만 원 사이인지 80~1200만 원에 퍼져 있는지 알 수 없다. 항상 **평균 ± 표준편차** 형태로 보고한다.

*(올바른 보고 예: "소득의 평균은 348.5만 원(SD = 182.3만 원)이다.")*

**오류 3: 시각화 없이 숫자만 나열**

기술통계 숫자만 보고 "대칭이다", "이상치가 있다"고 판단하는 것은 위험하다. 히스토그램과 박스플롯을 반드시 함께 확인한다. **"숫자가 말해주지 않는 것을 그래프가 보여준다."** 예를 들어, 동일한 평균·표준편차를 가진 전혀 다른 분포가 존재할 수 있다(Anscombe의 사분주).

> **(결론)** 소득 분포는 소수의 고소득자에 의해 평균이 끌어올려진 비대칭 형태이며, 중앙값(285만 원)이 "전형적인 응답자"를 더 잘 대표한다.

### 과제 3: 결과 종합 + 해석문 작성

Copilot의 답변 중 **1가지를 채택 또는 거절**하고, 이유를 결과 근거로 2문장 이내로 기록한다.

#### ③ 종합 해석문 작성

조원 전원이 합의하여 A-E 형식의 해석문을 완성한다.

**제출물 구성 (A-E)**:

- **(A) 분석 질문** (1-2문장): "이 데이터의 응답자 특성(소득, 연령, 교육수준)은 어떤 분포를 보이는가?"
- **(B) 사용 변수/척도**: 분석한 변수명 + 척도 + 수준 (표 W03.2 참조)
- **(C) 결과 근거**: Copilot이 생성한 코드 + 기술통계표(표 W03.3) + 히스토그램/박스플롯에서 발견한 패턴 2가지
- **(D) 해석문** (5-7문장): 아래 4단계 구조를 따른다.
  - **(분석목적)** 왜 이 분석을 했는가?
  - **(핵심근거)** 어떤 수치와 패턴을 확인했는가?
  - **(결론)** 데이터에서 무엇을 말할 수 있는가?
  - **(주의)** 이 결과의 한계는 무엇인가?
- **(E) Copilot 사용 요약** (2-3줄): 어떤 질문을 했고, 어떤 답변을 받았으며, 무엇을 채택/거절했는가?

**해석문(D항) 작성 시 주의**: 조원 전원이 초안을 읽고, **4단계가 모두 포함되었는지 반드시 교차 점검**한다.

| 빠지기 쉬운 부분 | 확인 질문 |
|---|---|
| (분석목적) 생략 | "왜 이 분석을 했는지 써 있는가?" |
| (핵심근거) 수치 누락 | "평균, 중앙값, 표준편차 등 구체적 숫자가 있는가?" |
| (결론) 모호함 | "그래서 무엇을 말할 수 있는지 명확한가?" |
| (주의) 한계 누락 | "표본, 자기보고, 기술통계의 한계가 명시되었는가?" |
