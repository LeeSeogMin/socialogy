### 왜 필요한가

핵심 질문은 이것이다: "표본에서 보이는 차이가 **진짜 모집단의 특성인가**, 아니면 **표본 추출 과정에서 우연히 발생한 것인가**?"

---

### 핵심 개념 1: 기술통계 vs 추론통계 전환

**기술통계(descriptive statistics)**란 "이 데이터에서 무엇이 보이는가"를 있는 그대로 요약하는 것이다. 평균, 표준편차, 분할표, 그래프가 모두 기술통계에 해당한다.

**추론통계(inferential statistics)**란 "이 패턴이 모집단에서도 성립하는가"를 **불확실성 하에서** 판단하는 것이다. 표본은 모집단의 일부이므로, 표본에서 관찰된 차이가 모집단에서도 존재하는지 확인하려면 **통계적 추론**이 필요하다.

쉽게 말해서, 기술통계는 "이 교실의 학생 30명 중 여학생 평균 키가 163cm이다"라고 서술하는 것이고, 추론통계는 "이 30명의 데이터를 근거로, **전국 여대생**의 평균 키가 160cm보다 크다고 말할 수 있는가?"를 판단하는 것이다.

> **이 전환이 통계학의 핵심 도약이다.** 3--4주차에서 배운 기술통계는 "기술(description)"에 불과하다. 오늘부터 배우는 추론통계가 "표본 → 모집단"의 다리를 놓는다.

---

4. "교육수준에 따른 생활만족도 차이는 모집단에서도 존재할 가능성이 높다." → *(추론)*

### 핵심 개념 2: 가설검정 프레임(hypothesis testing)

**가설검정(hypothesis testing)**은 "표본 데이터를 근거로, 모집단에 대한 주장을 판단하는 절차"이다. 법정 비유로 설명하면 이해가 쉽다.

> 법정에서 피고인은 **무죄로 추정**한다. 검사가 충분한 증거를 제시해야만 유죄로 판결한다. 가설검정도 마찬가지이다. "차이가 없다(무죄)"에서 출발하여, 데이터가 충분한 증거를 보여줄 때만 "차이가 있다(유죄)"고 판단한다.

가설검정의 절차를 5단계로 정리하면 다음과 같다.

1. **귀무가설(H₀, null hypothesis)** 설정: "차이/관계가 없다." (무죄 추정)
2. **대립가설(H₁, alternative hypothesis)** 설정: "차이/관계가 있다." (검사의 주장)
3. **검정통계량** 계산: 표본 데이터에서 차이의 크기를 수치로 요약한다 (예: t, χ², F).
4. **p값** 확인: "H₀가 참이라면, 이 정도 결과가 우연히 나올 확률은 얼마인가?"
5. **판단**: p값이 유의수준(α) 미만이면 H₀를 기각한다.

**그림 W05.1** 가설검정 의사결정 플로우차트

```
연구 질문 설정
    ↓
변수 척도 확인 (명목/서열/등간/비율)
    ↓
적절한 검정 선택 (가이드표 참조)
    ↓
귀무가설(H₀) / 대립가설(H₁) 설정
    ↓
검정통계량 계산 + p값 확인
    ↓
p < α → H₀ 기각 ("통계적으로 유의한 차이/관계가 있다")
p ≥ α → H₀ 기각 못함 ("충분한 증거를 찾지 못했다")
```

---

### 핵심 개념 3: p값과 유의수준(α)

**p값(p-value)**이란, "**H₀가 참이라고 가정할 때**, 현재 데이터에서 관찰된 것과 같거나 더 극단적인 결과가 우연히 나올 확률"이다.

동전 던지기 비유로 생각해 보자. 동전이 공정한지(H₀: 앞면 확률 = 0.5) 의심스러운 상황이다.

- 동전을 10번 던져 **6번 앞면** → "우연일 수 있겠다" (p값이 큼)
- 동전을 10번 던져 **8번 앞면** → "조금 의심스럽다" (p값이 작아짐)
- 동전을 10번 던져 **10번 앞면** → "이건 우연이 아닐 것이다" (p값이 매우 작음)

**유의수준(α, significance level)**은 "이 확률 이하면 우연이 아니라고 판단하겠다"는 기준선이다. 사회과학에서는 보통 **α = 0.05**(5%)를 사용한다.

> **판단 규칙**: p < α이면 H₀ 기각, p ≥ α이면 H₀ 기각 못함.

---

#### p값 오해 TOP 3

p값은 가설검정의 핵심이지만, 가장 많이 오해되는 개념이기도 하다.

**표 W05.1** p값의 흔한 오해와 올바른 이해

| 오해 | 올바른 이해 |
|---|---|
| "p = 0.03이면 H₀가 참일 확률이 3%이다" | p값은 H₀가 참일 확률이 **아니다**. "H₀가 참이라면 이 정도 결과가 나올 확률"이다. |
| "p값이 작을수록 효과가 크다" | p값은 효과의 크기를 말하지 **않는다**. 표본이 크면 작은 효과도 유의하게 나온다. |
| "p > 0.05이면 H₀가 참이다" | 기각하지 못한 것일 뿐, H₀가 참이라고 **증명**된 것이 아니다. |

#### 미니 디버깅 -- 틀린 해석 찾기

아래 해석문에서 잘못된 부분을 찾아보자. (30초 개인 → 30초 짝 토론)

> "p값이 0.02이므로, 귀무가설이 참일 확률은 2%이다. 따라서 두 변수 사이에 **인과관계**가 확인되었으며, 효과가 상당히 크다고 판단한다."

**문제점 3가지:**
1. p값 ≠ H₀가 참일 확률 (오해 1)
2. p값이 작다고 효과가 크다는 뜻이 아니다 (오해 2)
3. 통계적 유의성 ≠ 인과관계 (4주차 복습)

---

### 1종 오류와 2종 오류 (간략 소개)

가설검정의 판단에는 두 가지 오류가 있다.

- **1종 오류(Type I error)**: H₀가 사실인데 기각한 오류. 쉽게 말해서, "무죄인 사람을 유죄로 판결"한 것이다. 이 오류의 확률이 바로 유의수준 α이다.
- **2종 오류(Type II error)**: H₀가 거짓인데 기각하지 못한 오류. "유죄인 사람을 무죄로 풀어준" 것이다.

α를 너무 낮추면(예: 0.01) 1종 오류는 줄어들지만, 2종 오류(진짜 차이를 놓치는 경우)가 늘어난다. 이 둘 사이에 **트레이드오프**가 존재한다.

---

### 가설검정의 한계

가설검정은 강력한 도구이지만, 다음 두 가지 한계를 반드시 기억해야 한다.

1. **"유의하다" ≠ "중요하다"**: 통계적으로 유의한 결과가 실질적으로 의미 있는 것은 아니다. 표본이 충분히 크면 아주 작은 차이도 유의하게 나온다. 반드시 **효과 크기**(Cohen's d, Cramer's V, R² 등)를 함께 보고해야 한다.
2. **"유의하지 않다" ≠ "효과가 없다"**: p > 0.05라고 해서 두 변수 사이에 관계가 전혀 없다는 뜻이 아니다. "이 표본에서는 충분한 증거를 찾지 못했다"는 것일 뿐이다.

### 짝 점검 -- 가설과 검정 선택 교차 확인

> **점검 기준**: 상대방의 답안에서 다음을 확인한다.

> 2. 검정 선택이 가이드표(표 W05.2)와 일치하는가?
> 3. 독립변수와 종속변수의 척도가 정확하게 기재되었는가?
>
> 오류가 발견되면 이유를 적어 돌려준다.

## 정리/오류 패턴 공유 + 비판적 읽기

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 답안에 해당하는 항목이 없는지 점검해 보자.

**오류 1: 귀무가설과 대립가설을 뒤바꿈**

"남녀 소득에 차이가 있다"를 H₀로 설정하는 경우가 있다. 가설검정은 **"무죄 추정"에서 출발**한다. H₀는 항상 "차이/관계가 없다"이고, H₁이 "있다"이다.

**확인법**: H₀에 "없다", "같다", "독립이다" 같은 표현이 있는지 확인한다. 만약 H₀에 "다르다", "있다"가 들어 있으면 H₀와 H₁이 뒤바뀐 것이다.

**오류 2: 변수 척도와 검정의 불일치**

예를 들어, 독립변수가 명목(3수준)이고 종속변수가 비율인데 카이자승 검정을 선택하는 경우이다. 카이자승 검정은 두 변수가 **모두 범주형**일 때 사용한다.

**확인법**: 가이드표(표 W05.2)에서 자신의 변수 척도 조합을 찾아 해당하는 검정을 확인한다.

**오류 3: p값을 "H₀가 참일 확률"로 해석**

"p = 0.03이므로 귀무가설이 참일 확률이 3%이다"라고 쓰는 경우가 가장 빈번하다.

**확인법**: 해석문(D항)에서 "H₀가 참일 확률"이라는 표현이 있으면 수정한다. p값은 "H₀가 참이라고 가정할 때, 이 정도 결과가 우연히 나올 확률"이다.

---

### 해석문 모범 구조

오늘 과제의 (D)항 해석문은 다음 4단계 구조를 따른다. 아래는 연구 질문 1(성별과 월 소득)의 모범 예시이다.

> **(분석목적)** 성별에 따라 월 소득에 차이가 있는지 확인하기 위해 독립표본 t-검정을 실시했다.
>
> **(핵심근거)** t(198) = 2.45, p = 0.015이며, Cohen's d = 0.35이다. 남성 평균은 355만 원, 여성 평균은 310만 원으로, 약 45만 원의 차이가 관찰되었다.
>
> **(결론)** 유의수준 0.05에서 귀무가설(남녀 소득 평균에 차이가 없다)을 기각한다. Cohen's d = 0.35로, 작은~중간 수준의 효과 크기이다.
>
> **(주의)** 이 결과는 성별이 소득을 "결정"한다는 의미가 아니다. 직종, 경력, 근무시간 등 교란변수를 통제하지 않았으며, 표본(n = 200)의 대표성에도 한계가 있다.

### 비판적 읽기 (5--10분)

아래는 가상의 연구 보고서 발췌문이다. 읽고 문제점을 찾아보자.

> *"대학생 500명을 대상으로 설문한 결과, 운동 빈도와 학점 사이에 유의한 상관이 나타났다(r = 0.08, p = 0.04). 이는 운동이 학업 성취를 향상시키는 효과가 있음을 **입증**한다."*

> **토론 질문**: 이 해석에서 문제점을 찾아보자. (2--3분, 짝 토론 후 전체 공유)

**문제점 예시:**

1. **효과 크기 무시**: r = 0.08은 매우 약한 상관이다. p값이 유의한 것은 표본(n = 500)이 크기 때문이지, 효과가 크기 때문이 아니다. "유의하다 ≠ 중요하다"의 전형적 사례이다.

2. **인과 표현**: 설문 데이터(횡단면 조사)에서 "향상시키는 효과"나 "입증"이라는 표현을 사용했다. 상관은 인과를 의미하지 않는다. "운동 빈도와 학점 사이에 약한 양의 상관이 **관찰되었다**"가 적절하다.

3. **역인과/교란변수 가능성**: 학점이 높은 학생이 자기관리 능력이 좋아서 운동도 더 많이 할 수 있다(역인과). 혹은 사회경제적 배경이 두 변수 모두에 영향을 미칠 수 있다(교란변수).

### 오늘의 과제 특성

오늘 과제는 이전 주차와 다르게, **실제 데이터 분석 코드를 실행하지 않는다**. 대신 "연구 질문 → 가설 → 검정 선택 → 가상 결과 해석"이라는 **논리적 흐름을 문장으로 쓰는 연습**에 집중한다. 코드 실행은 6주부터 본격적으로 시작한다.

### 검정 선택 가이드표 (참조용)

| 연구 질문 | 독립변수 척도 | 종속변수 척도 | 검정 |

오늘은 코드 생성 대신, Copilot에게 **검정 선택의 적절성을 확인받는 용도**로 바이브코딩을 활용한다.

**바이브코딩 프롬프트 예시**:

| 순서 | 프롬프트 예시 |
|---:|---|

**질문 1 가상 결과**:
