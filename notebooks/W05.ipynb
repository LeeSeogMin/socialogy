### 핵심 개념 1: 기술통계 vs 추론통계 전환

- **기술통계(descriptive statistics)**
  - "이 데이터에서 무엇이 보이는가"를 있는 그대로 요약하는 것
  - 예: 평균, 표준편차, 분할표, 그래프

- **추론통계(inferential statistics)**
  - "이 패턴이 모집단에서도 성립하는가"를 **불확실성 하에서** 판단하는 것
  - 표본 = 모집단의 일부 → 표본에서 관찰된 차이가 모집단에서도 존재하는지 확인하려면 **통계적 추론** 필요

- **기술 vs 추론 예시**
  - 기술통계: "이 교실의 학생 30명 중 여학생 평균 키가 163cm이다"
  - 추론통계: "이 30명의 데이터를 근거로, **전국 여대생**의 평균 키가 160cm보다 크다고 말할 수 있는가?"

> **이 전환이 통계학의 핵심 도약이다.** 3~4주차에서 배운 기술통계는 "기술(description)"에 불과하다. 오늘부터 배우는 추론통계가 "표본 → 모집단"의 다리를 놓는다.

1. "설문 응답자 200명의 평균 소득은 330만 원이다." → *(기술)*

4. "교육수준에 따른 생활만족도 차이는 모집단에서도 존재할 가능성이 높다." → *(추론)*

### 핵심 개념 2: 가설검정 프레임(hypothesis testing)

- **가설검정(hypothesis testing)** 정의
  - "표본 데이터를 근거로, 모집단에 대한 주장을 판단하는 절차"

- **법정 비유**

> 법정에서 피고인은 **무죄로 추정**한다. 검사가 충분한 증거를 제시해야만 유죄로 판결한다. 가설검정도 마찬가지이다. "차이가 없다(무죄)"에서 출발하여, 데이터가 충분한 증거를 보여줄 때만 "차이가 있다(유죄)"고 판단한다.

- **가설검정 5단계 절차**:
  1. **귀무가설(H₀, null hypothesis)** 설정: "차이/관계가 없다" (무죄 추정)
  2. **대립가설(H₁, alternative hypothesis)** 설정: "차이/관계가 있다" (검사의 주장)
     - 연구가설은 보통 이 대립가설의 실질적 내용이며, 검정 절차에서는 결국 **H₀를 기각할 근거가 충분한지**를 판단한다.
  3. **검정통계량** 계산: 표본 데이터에서 차이의 크기를 수치로 요약 (예: t, χ², F)
  4. **p값** 확인: "H₀가 참이라면, 이 정도 결과가 우연히 나올 확률은 얼마인가?"
  5. **판단**: p값이 유의수준(α) 미만이면 H₀ 기각

- **통계적으로 유의하다**의 뜻
  - "대립가설이 절대적으로 증명되었다"는 뜻이 아니라, **설정한 유의수준 α 아래에서 귀무가설을 기각할 수 있다**는 뜻

- **그림 W05.1** 가설검정 의사결정 플로우차트

```
연구 질문 설정
    ↓
변수 척도 확인 (명목/서열/등간/비율)
    ↓
적절한 검정 선택 (가이드표 참조)
    ↓
귀무가설(H₀) / 대립가설(H₁) 설정
    ↓
검정통계량 계산 + p값 확인
    ↓
p < α → H₀ 기각 ("통계적으로 유의한 차이/관계가 있다")
p ≥ α → H₀ 기각 못함 ("충분한 증거를 찾지 못했다")
```

---

### 핵심 개념 3: p값과 유의수준(α)

- **p값(p-value)** 정의
  - "**H₀가 참이라고 가정할 때**, 현재 데이터에서 관찰된 것과 같거나 더 극단적인 결과가 우연히 나올 확률"

- **동전 던지기 비유** — 동전이 공정한지(H₀: 앞면 확률 = 0.5) 의심스러운 상황
  - 10번 던져 **6번 앞면** → "우연일 수 있겠다" (p값이 큼)
  - 10번 던져 **8번 앞면** → "조금 의심스럽다" (p값이 작아짐)
  - 10번 던져 **10번 앞면** → "이건 우연이 아닐 것이다" (p값이 매우 작음)

- **유의수준(α, significance level)**
  - 연구자가 **사전에 정하는 기각 기준선**
  - "이 확률 이하면 우연만으로 보기 어렵다고 판단하겠다"는 약속
  - 사회과학에서 보통 **α = 0.05**(5%) 사용

- **중요 구분**
  - 기각의 기준선은 **p값 자체가 아니라 α**이다.
  - p값은 데이터로부터 계산된 결과이고, α는 그 p값을 비교하는 판단 기준이다.

> **판단 규칙**: p < α이면 H₀ 기각, p ≥ α이면 H₀ 기각 못함.

---

#### p값 오해 TOP 3

- p값은 가설검정의 핵심이지만, 가장 많이 오해되는 개념

- **표 W05.1** p값의 흔한 오해와 올바른 이해

| 오해 | 올바른 이해 |
|---|---|
| "p값이 작을수록 효과가 크다" | p값은 효과의 크기를 말하지 **않는다**. 표본이 크면 작은 효과도 유의하게 나온다. |
| "p > 0.05이면 H₀가 참이다" | 기각하지 못한 것일 뿐, H₀가 참이라고 **증명**된 것이 아니다. |

**문제점 3가지:**

1. **p값 ≠ H₀가 참일 확률** (오해 1)
2. **p값이 작다 ≠ 효과가 크다** (오해 2)
3. **통계적 유의성 ≠ 인과관계** (4주차 복습)

---

### 1종 오류와 2종 오류 (간략 소개)

- 가설검정의 판단에는 **두 가지 오류** 존재

- **1종 오류(Type I error)**
  - H₀가 사실인데 기각한 오류
  - 비유: "무죄인 사람을 유죄로 판결"
  - 이 오류의 확률 = **유의수준 α**

- **2종 오류(Type II error)**
  - H₀가 거짓인데 기각하지 못한 오류
  - 비유: "유죄인 사람을 무죄로 풀어줌"

- **트레이드오프**
  - α를 너무 낮추면(예: 0.01) → 1종 오류 감소
  - 그러나 2종 오류(진짜 차이를 놓치는 경우) 증가

---

### 가설검정의 한계

- 가설검정은 강력한 도구이지만, **반드시 기억할 두 가지 한계**:

1. **"유의하다" ≠ "중요하다"**
   - 통계적으로 유의한 결과가 실질적으로 의미 있는 것은 아님
   - 표본이 충분히 크면 → 아주 작은 차이도 유의하게 나옴
   - ∴ 반드시 **효과 크기**(Cohen's d, Cramér's V, R² 등)를 함께 보고

2. **"유의하지 않다" ≠ "효과가 없다"**
   - p > 0.05라고 해서 두 변수 사이에 관계가 전혀 없다는 뜻이 아님
   - "이 표본에서는 충분한 증거를 찾지 못했다"는 것일 뿐

- **확인법**: 가이드표(표 W05.2)에서 자신의 변수 척도 조합 찾아 해당 검정 확인

### 비판적 읽기 (5--10분)

- **가상의 연구 보고서 발췌문**

> *"대학생 500명을 대상으로 설문한 결과, 운동 빈도와 학점 사이에 유의한 상관이 나타났다(r = 0.08, p = 0.04). 이는 운동이 학업 성취를 향상시키는 효과가 있음을 **입증**한다."*

> **토론 질문**: 이 해석에서 문제점을 찾아보자. (2--3분, 짝 토론 후 전체 공유)

**문제점 분석:**

1. **효과 크기 무시**
   - r = 0.08 = 매우 약한 상관
   - p값이 유의한 이유: 표본(n = 500)이 크기 때문 (효과가 크기 때문 ✕)
   - 전형적인 "유의하다 ≠ 중요하다" 사례

2. **인과 표현**
   - 설문 데이터(횡단면 조사)에서 "향상시키는 효과", "입증"이라는 표현 사용 → 부적절
   - 상관 ≠ 인과
   - 적절한 표현: "운동 빈도와 학점 사이에 약한 양의 상관이 **관찰되었다**"

3. **역인과/교란변수 가능성**
   - 역인과: 학점이 높은 학생이 자기관리 능력이 좋아서 운동도 더 많이 할 수 있음
   - 교란변수: 사회경제적 배경이 두 변수 모두에 영향을 미칠 수 있음

#### 질문 1 예시 답안

- 검정통계량: t = 2.31

>

>
> **(주의)** 다만 통계적으로 유의하다고 해서 그 차이가 반드시 큰 것은 아니며, 효과 크기와 자료의 맥락을 함께 해석해야 한다.
