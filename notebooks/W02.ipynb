### 핵심 개념 1: 데이터 리터러시(data literacy)

**데이터 리터러시(data literacy)**란 데이터의 출처·수집 목적·표본 특성·편향 가능성을 파악하는 태도이다. 쉽게 말해서, 숫자를 분석하기 전에 **"이 데이터를 믿을 수 있는가?"**를 먼저 질문하는 것이다.

예를 들어, "이 설문은 대학생만 대상이므로, 결과를 전체 국민에게 일반화할 수 없다"와 같이, 데이터의 한계를 인식하는 것이 데이터 리터러시의 핵심이다. 이 태도는 이후 모든 과제에서 해석문의 "(주의) 한계" 항목을 작성할 때 직접 활용된다.

데이터를 처음 받으면 다음 체크리스트를 확인한다.

**데이터 신뢰 체크리스트:**
- [ ] 수집 주체와 목적이 명확한가?
- [ ] 표본 추출 방법이 적절한가? (무작위 추출 vs 편의 추출)
- [ ] 응답률과 결측 패턴이 보고되었는가?
- [ ] 변수 정의(**코드북**, codebook)가 제공되었는가?

**코드북(codebook)**이란 각 변수의 이름, 정의, 측정 방법, 범주 코딩 규칙 등을 정리한 문서이다. 쉽게 말해서, "이 숫자가 무슨 뜻인가?"를 알려주는 설명서이다. 예를 들어, 성별 변수에서 1=남성, 2=여성이라는 정보가 코드북에 들어 있다. 코드북 없이 데이터를 분석하면 변수의 의미를 정확히 알 수 없으므로, 해석이 불가능하다.

### 핵심 개념 2: 변수의 척도 수준(level of measurement)

**변수의 척도 수준(level of measurement)**은 변수가 담고 있는 정보의 종류와 수준을 나타낸다. 척도에 따라 사용할 수 있는 분석 기법이 달라지므로, 분석 전에 반드시 확인해야 한다.

| 척도 | 특성 | 예시 | 가능한 연산 |
|---|---|---|---|
| **명목(nominal)** | 이름/범주만 구분 | 성별, 지역, 종교 | 빈도, 최빈값 |
| **서열(ordinal)** | 순서가 있음 | 교육수준, 만족도 등급 | 빈도, 중앙값, 순위 |
| **등간(interval)** | 간격이 동일, 영점이 임의 | 온도(섭씨), IQ | 평균, 표준편차 |
| **비율(ratio)** | 절대 영점이 존재 | 소득, 연령, 체중 | 평균, 비율 비교 |

척도를 판별할 때는 다음 **의사결정 트리**를 활용한다.

**표 W02.1** 척도 구분 의사결정 가이드

| 질문 | 예 → | 아니오 → |
|---|---|---|
| 범주(카테고리)인가? | 명목 or 서열 | 등간 or 비율 |
| 순서가 있는가? | 서열 | 명목 |
| 간격이 동일한가? | 등간 or 비율 | 서열 |
| 절대 영점이 있는가? | 비율 | 등간 |

**척도가 분석 기법 선택을 결정한다.** 예를 들어, 명목 변수끼리의 관계는 카이자승 검정(4주차)으로, 등간/비율 변수의 집단 간 평균 차이는 t-검정(6주차)으로 분석한다. 척도를 잘못 판단하면 부적절한 분석 기법을 선택하게 되며, 그 결과는 신뢰할 수 없다.

### 코딩(부호화): 설문지에서 데이터로

실제 사회조사에서 데이터는 대부분 설문지에서 출발한다. 설문 응답을 숫자로 변환하는 과정을 **코딩(coding)** 또는 **부호화**라고 한다.

흐름은 다음과 같다.

> 설문지 작성 → **코드북** 정의(변수명, 척도, 코딩 규칙) → 응답을 숫자로 변환 → CSV 파일 생성

예를 들어, 설문지에 "귀하의 최종 학력은?"이라는 문항이 있고, 보기가 "① 고졸이하 ② 대졸 ③ 대학원이상"이라면, 코드북에는 다음과 같이 기록한다.

| 변수명 | 척도 | 코딩 |
|---|---|---|
| 교육수준 | 서열 | 1=고졸이하, 2=대졸, 3=대학원이상 |

이때 숫자 1, 2, 3은 **순서**를 나타낼 뿐이다. 1과 2 사이의 간격이 2와 3 사이의 간격과 같다는 뜻이 아니다. 이 점을 잊고 평균을 구하면 해석이 어색해진다("평균 교육수준이 1.8이다"는 무슨 뜻인가?).

### 핵심 개념 3: 결측치(missing data)

**결측치(missing data)**란 데이터에서 값이 비어 있는 경우를 말한다. 설문에서 응답하지 않은 문항, 입력 오류, 시스템 장애 등 다양한 원인으로 발생한다. 결측치를 어떻게 처리하느냐에 따라 분석 결과가 달라질 수 있으므로, 결측의 **유형**을 먼저 파악해야 한다.

**표 W02.2** 결측치 유형별 처리 전략

| 유형 | 정의 | 처리 방법 | 예시 |
|---|---|---|---|
| **MCAR**(완전 무작위 결측) | 결측이 어떤 변수와도 무관 | 삭제 가능 | 설문지 분실, 입력 실수 |
| **MAR**(무작위 결측) | 결측이 다른 관찰된 변수에 의존 | 다중 대체(MI) 권장 | 고연령자가 SNS 이용시간 문항 미응답이 많음 |
| **MNAR**(비무작위 결측) | 결측이 결측값 자체에 의존 | 모델 기반 접근 필요 | 고소득자가 소득 미응답 |

쉽게 말해서, MCAR은 "아무 이유 없이 빠진 것", MAR은 "다른 변수와 관련 있게 빠진 것", MNAR은 "빠진 값 자체와 관련 있게 빠진 것"이다.

**핵심 원칙**: 결측이 있는 행을 무조건 삭제하는 것은 위험하다. 결측 비율이 높거나(20% 이상), 결측 패턴이 무작위가 아닌 경우에는 삭제 대신 **대체(imputation)**를 고려해야 한다. 결측치를 무시하고 분석하면, 표본이 편향되어 결론이 왜곡될 수 있다.

### 기본(Basic) 트랙

#### ① 실행

제공된 데이터의 변수 목록을 출력하고, 각 변수의 척도를 분류표로 작성한다.

```python
# 크로스플랫폼 한글 폰트 설정 (macOS / Windows)
import platform
import matplotlib.pyplot as plt

if platform.system() == "Darwin":  # macOS
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd

df = pd.read_csv("data/social_survey_w02_raw.csv", encoding='utf-8')

# 1. 변수 정보 확인
print("=== 변수 정보 ===")
print(df.info())

# 2. 결측치 현황
print("\n=== 결측치 현황 ===")
print(df.isnull().sum())

# 3. 변수별 고유값 수 확인 (척도 판별 참고)
print("\n=== 변수별 고유값 ===")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}개 고유값 — {df[col].dtype}")
```

```python
print("\n=== 변수별 결측 비율 ===")
print((df.isnull().mean() * 100).round(1))
```

```python
# 대체 전 소득 통계 저장
before_mean = df["소득"].mean()
before_std = df["소득"].std()
n_missing = df["소득"].isnull().sum()

print(f"소득 결측: {n_missing}건 ({df['소득'].isnull().mean()*100:.1f}%)")
print(f"대체 전 — 평균: {before_mean:.1f}, 표준편차: {before_std:.1f}")
```

**3-2. 결측치 보완 (중앙값 대체)**: 수치형 변수의 결측을 중앙값으로, 범주형 변수의 결측을 최빈값으로 대체한다.

```python
df_filled = df.copy()

# 수치형: 중앙값 대체
for col in ["소득", "생활만족도"]:
    median_val = df_filled[col].median()
    df_filled[col] = df_filled[col].fillna(median_val)
    print(f"{col}: 중앙값({median_val:.0f})으로 대체")

# 범주형: 최빈값 대체
for col in ["성별", "교육수준"]:
    mode_val = df_filled[col].mode()[0]
    df_filled[col] = df_filled[col].fillna(mode_val)
    print(f"{col}: 최빈값('{mode_val}')으로 대체")

print(f"\n남은 결측: {df_filled.isnull().sum().sum()}건")
```

**3-3. 대체 전후 비교**: 소득 변수의 평균·표준편차가 어떻게 변했는지 비교한다.

```python
after_mean = df_filled["소득"].mean()
after_std = df_filled["소득"].std()

print("\n=== 소득 변수 전후 비교 ===")
print(f"{'':>12} {'평균':>10} {'표준편차':>10}")
print(f"{'대체 전':>12} {before_mean:>10.1f} {before_std:>10.1f}")
print(f"{'대체 후':>12} {after_mean:>10.1f} {after_std:>10.1f}")
```

**3-4. 해석 질문**: 중앙값 대체 후 표준편차가 줄어드는 이유를 1문장으로 설명하라.

*(힌트: 결측값을 모두 "중심값" 하나로 채우면 데이터가 중앙으로 몰리기 때문이다.)*

**완료 체크**: 결측 현황 확인 → 중앙값/최빈값 대체 → 전후 비교가 완료되었는가?

### 흔한 오류 3개

실습 중 자주 나타나는 오류를 정리한다. 자신의 과제에 해당하는 항목이 없는지 점검해 보자.

**오류 1: 척도 혼동 — 리커트 척도를 등간으로 취급**

리커트 5점 척도(매우 불만족~매우 만족)를 등간 척도로 간주하고 평균을 구하는 경우가 많다. 그러나 "불만족"과 "보통" 사이의 심리적 간격이 "보통"과 "만족" 사이의 간격과 동일하다는 보장이 없다. 보수적으로는 **서열 척도**로 취급하며, 등간으로 취급할 경우에는 그 가정을 명시해야 한다.

```python
# 서열 변수에 평균을 구하는 것이 적절한지 생각해 보자
print(df["생활만족도"].mean())  # 이 값의 해석이 타당한가?
print(df["생활만족도"].median())  # 서열 변수에는 중앙값이 더 적절
```

**오류 2: 결측치를 무조건 삭제**

결측이 있는 행을 무조건 삭제(리스트와이즈 삭제)하면, 결측이 무작위가 아닌 경우(MAR, MNAR) 표본이 편향될 수 있다. 결측 비율이 20% 이상이면 **대체(imputation)**를 고려해야 한다.

```python
# 무조건 삭제 (주의 필요)
df_clean = df.dropna()
print(f"원본: {len(df)}행 → 삭제 후: {len(df_clean)}행")
# 얼마나 많은 데이터가 사라졌는지 반드시 확인
```

**오류 3: 코드북 없이 분석 시작**

변수의 의미를 모른 채 분석을 시작하면, 결과를 해석할 수 없다. 예를 들어, 성별 변수에서 1이 남성인지 여성인지 모르면, "1이 더 많다"는 말은 아무 의미가 없다. **코드북 확인이 분석의 첫 단계**이다.

아래는 가상의 뉴스 기사 발췌문이다. 읽고 문제점을 찾아보자.

> **〈○○일보〉 2026.03.10**
>
> *"○○대학교 학생 300명을 대상으로 한 설문 결과, 한국 청년의 82%가 정치에 무관심한 것으로 나타났다."*

> **토론 질문**: 이 기사의 해석에서 문제점을 찾아보자. (2–3분, 짝 토론 후 전체 공유)

**문제점 예시:**

1. **표본 대표성**: ○○대학교 학생 300명은 "한국 청년" 전체를 대표하지 않는다. 특정 대학, 특정 지역, 특정 학과의 학생일 가능성이 있으며, 이를 전체 청년으로 일반화하는 것은 **과잉 일반화**이다.

2. **변수 정의 불명확**: "정치에 무관심"이라는 개념을 어떻게 측정했는지 보고되지 않았다. 설문 문항이 "정치에 관심이 있다/없다"인지, "정치 뉴스를 주 ___회 이상 본다"인지에 따라 결과가 완전히 달라질 수 있다. 코드북 없이는 "무관심"의 정확한 의미를 알 수 없다.

3. **응답률/결측 미보고**: 300명이 설문에 참여했다고 했지만, 모집단(전체 학생 수)에서 얼마나 응답했는지, 미응답자가 어떤 특성을 가졌는지 보고되지 않았다. 정치에 관심이 없는 학생이 오히려 설문에 응답하지 않았을 수 있다(MNAR).
