### 분석 흐름 완주 체크리스트

모든 분석은 아래 7단계를 따른다. 과제 수행 시 이 체크리스트를 순서대로 점검한다.

| 단계 | 내용 | 확인 |
|---:|---|---|
| 1 | **연구 질문** 설정 (1문장) | [ ] |
| 2 | **변수 정의**: 독립/종속 변수, 척도(명목/서열/등간·비율) 명시 | [ ] |
| 3 | **기술통계**: 빈도표, 평균/표준편차, 히스토그램 등으로 데이터 탐색 | [ ] |
| 4 | **검정 선택**: 변수 유형과 질문에 맞는 분석 기법 선택 (아래 표 참고) | [ ] |
| 5 | **가정 확인**: 선택한 검정의 전제 조건 충족 여부 확인 | [ ] |
| 6 | **검정 실행**: 코드 실행 및 결과 확인 | [ ] |
| 7 | **해석문 작성**: 4단계 해석문 (분석목적 → 핵심근거 → 결론 → 주의/한계) | [ ] |

### 검정 선택 가이드

| 독립변수 | 종속변수 | 추천 기법 |

| 연속형 | 연속형 | 단순회귀/중다회귀 |

### 1쪽 요약 메모 구조

최종 제출물은 아래 구조의 1쪽 요약 메모이다. 사회과학 글쓰기의 기본 구조를 따른다.

1. **서론** (2-3문장): 연구 질문과 그 질문이 왜 중요한지
2. **방법** (3-4문장): 사용 데이터, 변수(척도 포함), 분석 기법, 가정 확인 결과
3. **결과** (3-4문장): 핵심 수치(검정통계량, p값, 효과크기 등) + 표/그래프 1개
4. **논의** (2-3문장): 결론, 한계, "이 분석이 말할 수 없는 것"
