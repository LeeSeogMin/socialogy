## A회차: 보고서와 발표 구성

### 사회과학 보고서 구조

- 최종 제출물인 2쪽 요약 메모는 아래 구조를 따른다.

| 섹션 | 분량 | 핵심 내용 |
|---|---|---|
| **1. 연구 질문** | 2-3문장 | 무엇을 알고 싶은가? 왜 중요한가? |
| **2. 데이터와 방법** | 4-5문장 | 어떤 데이터를 사용했는가? 변수와 척도는? 어떤 기법을 선택했으며 왜? |
| **3. 결과** | 5-7문장 + 표/그래프 | 핵심 수치는? 가설 검정 결과는? |
| **4. 해석과 논의** | 3-4문장 | 이 결과가 무엇을 의미하는가? "So what?" |
| **5. 한계** | 2-3문장 | 이 분석이 말할 수 없는 것은? 데이터의 제약은? |

---

### "So what?" — 핵심 메시지

- **핵심**: 분석 결과 나열만으로는 불충분
- **듣는 사람의 궁금증**: "그래서 무엇을 알 수 있는가?"

- **핵심 메시지 만드는 3단계**:
  1. **결과 요약**: "분석 결과, ___와 ___ 사이에 통계적으로 유의한 관련/차이가 있었다(또는 없었다)."
  2. **의미 부여**: "이는 ___를 시사한다."
  3. **한계 명시**: "다만, 이 결과가 ___를 의미하지는 않는다."

---

### 5분 발표 구성 가이드

| 시간 | 내용 |
|---:|---|
| 0-1분 | 연구 질문 + 왜 중요한가 |
| 1-2분 | 데이터·방법 요약 |
| 2-4분 | 핵심 결과 (표/그래프 1-2개로 설명) |
| 4-5분 | "So what?" + 한계 |

**발표 시 주의사항**:

- 슬라이드에 숫자 나열 (✕) → **핵심 그래프/표 1-2개**로 설명 (✓)
- 인과 표현 사용 금지
  - (✕) "A가 B를 초래한다"
  - (✓) "A와 B 사이에 유의한 관련이 있다"
- "모르겠다", "확신할 수 없다" = **학술적으로 올바른 솔직한 표현**

---

### 짝 팀 피드백 가이드

- 다른 팀의 발표를 듣고 **아래 3가지 질문**에 답:

1. 연구 질문이 **명확하고 구체적**인가?
2. 결과 근거(표/그래프/수치)가 결론을 **충분히 뒷받침**하는가?
3. 한계를 **솔직하게** 언급했는가?

- **피드백 형식**: "잘한 점 1개 + 개선 제안 1개"
