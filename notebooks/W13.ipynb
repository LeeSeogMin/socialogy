## A회차: 논의·한계 작성과 교차리뷰 방법

### 논의(Discussion) 섹션 — "So what?"

결과를 나열하는 것으로는 충분하지 않다. 독자는 **"이 결과가 무엇을 말하는가"** 를 알고 싶어한다.

3단계 구성:

1. **결과 요약** (1-2문장): "분석 결과, ___와 ___ 사이에 통계적으로 유의한 관련/차이가 있었다(또는 없었다)."
2. **선행연구와 비교** (1-2문장): "이는 [선행연구]의 결과와 일치한다(또는 불일치한다). 이유는 ___로 추정된다."
3. **대안적 설명 반박** (1-2문장): "이 관련성이 [제3변수]로 설명될 수 있다는 반론이 있으나, 본 분석에서는 ___를 통제했기에 그 가능성이 낮다."

**금지 표현**: "A가 B를 초래한다", "A 때문에 B가 증가한다", "A가 B의 원인이다". 상관관계 자료로 인과를 주장할 수 없다.

### 한계(Limitations) 섹션 — 솔직함이 학술성이다

한계를 숨기지 않는 것이 학술적 성실성이다. 최소 3가지 측면을 서술한다:

| 한계 유형 | 예시 |
|---|---|
| **데이터의 한계** | 2020년 단일 시점 조사, 표본이 특정 지역 편중 |
| **측정의 한계** | 소득을 11점 서열 척도로만 측정, 자기보고 편향 가능 |
| **분석의 한계** | 횡단 자료이므로 인과 방향 불명, 중요 통제변수(예: 가구 재산)가 없음 |

### 참고문헌 — 최소 5편, 한/영 병행

- 최소 5편 (학술지 논문 위주, 학위논문 최대 1편)
- 한국어 2편 이상, 영어 2편 이상
- **APA 7판** 형식 일관 적용
- 본문 인용(in-text citation)과 참고문헌 목록이 **정확히 일치**해야 한다.

**금지**: Copilot이 생성한 참고문헌을 그대로 쓰지 않는다. 반드시 원문을 직접 확인한다.

### 동료 교차리뷰 절차

같은 수업의 학생 1명과 초안을 교환한다. 아래 3가지를 확인한다.

| 체크 항목 | 확인 내용 |
|---|---|
| **재현성** | 노트북을 Restart & Run All로 실행하여 오류 없이 완주되는가? |
| **해석 정확성** | 인과 표현이 있는가? p값만 보고하고 효과크기를 빠뜨렸는가? |
| **서술 명확성** | 연구 질문과 결과가 일관된가? 모호한 문장이 있는가? |

리뷰 제출 형식:
- 잘한 점 1개 (구체적으로)
- 개선 제안 1개 (구체적으로)
- 재현성 확인 결과 (성공/실패 + 실패 시 원인)

### 5분 발표 구성

| 시간 | 내용 |
|---:|---|
| 0-1분 | 연구 질문 + 왜 중요한가 |
| 1-2분 | 데이터와 방법 |
| 2-4분 | 핵심 결과 (표/그래프 1-2개로) |
| 4-5분 | So what + 한계 |

**슬라이드 원칙**: 숫자 나열 금지. 표 1개 + 그림 1개로 압축한다.
